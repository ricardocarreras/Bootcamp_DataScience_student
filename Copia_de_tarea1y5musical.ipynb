```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
warnings.filterwarnings('ignore')
```

```python
desnormalizadafinal["acousticness"].hist()
```

```python
# creamos datos que incluyen slo al artista Ed Sheeran

def histograma_artista (nombre_artista, parametro):
  datosART=desnormalizadafinal[desnormalizadafinal.name_y==nombre_artista]
  sns.histplot(data = datosART[parametro], x = datosART[parametro], stat = "probability").set(title=nombre_artista+" y "+parametro) 
  return
```

```python
 nombre_artista="Ed Sheeran"
 parametro="acousticness"
```

```python
# vemos el histograma de la caracterstica de acstuca "acousticness" para esos datos
```

```python
histograma_artista(nombre_artista, parametro)
```
