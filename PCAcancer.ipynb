```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
cancer.target
```

```python
cancer.dtype()
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
df_std=StandardScaler().fit_transform(df)
df_std
```

```python
import numpy as np
```

```python
np.cov(df_std.T)
```

```python
np.linalg.eig(np.cov(df_std.T))
```

```python
valores_propios,vectores_propios = np.linalg.eig(np.cov(df_std.T))
```

```python
valores_propios.sum()
```

```python
parejas=[(np.abs(valores_propios[i]),vectores_propios[:,i]) for i in range(len(valores_propios))]
parejas
```

```python
parejas.sort(key=lambda x: x[0],reverse=True)
```

```python
parejas
```

```python
var_explicada=[i*100/valores_propios.sum() for i in sorted(valores_propios,reverse=True)]
```

```python
cum_var=np.cumsum(var_explicada)
```

```python
cum_var
```

```python
len(df.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=12)
```

```python
pca_df=pca.fit(df_std)
```

```python
pca_df
```

```python
pca_df.components_
```

```python
datos_cancerACP=pd.DataFrame(data=pca_df.components_,columns=df.columns)
```

```python
datos_cancerACP
```

```python
pca_df.explained_variance_ratio_
```

```python
pca_df
```

```python
pca_1=PCA()
```

```python
pca_1.fit(df_std)
```

```python

varianza=pca_1.explained_variance_ratio_
varianza_acumulada=np.cumsum(varianza)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(14, 8), dpi=80)

plt.bar(range(len(varianza)),varianza)
plt.plot(range(len(varianza)),varianza_acumulada)
plt.axhline(y=0.98)
plt.show()
```

```python
pca_df
```

```python
nuevos_datos=pd.DataFrame(pca_df.transform(df_std))
```

```python
nuevos_datos.shape
```

```python
#nuevos_datos["target"] = pd.Series(cancer.target)

df2 = pd.DataFrame(cancer.target, columns=["target"])
```

```python
nuevos_datos
```

```python
#nuevos_datos2=pd.nuevos_datos.join(df2)

nuevos_datos2 = pd.concat([nuevos_datos, df2], axis=1)
```

```python
nuevos_datos2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X=nuevos_datos2.
```

```python
X = nuevos_datos2.loc[:,nuevos_datos2.columns != 'target']
y = nuevos_datos2.loc[:,nuevos_datos2.columns == 'target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1234)
```

```python
modelo=LogisticRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
y_pred=modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
precision_score(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```
