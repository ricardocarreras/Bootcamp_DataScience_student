```python
text = """
Ana Rioja Jiménez
MIGUEL V. CARRERAS
De vocación, empresario
© Textos
Miguel V. Carreras Calvete y Ana Rioja Jiménez
Edita
Carreras Grupo Logístico
Autora
Ana Rioja Jiménez
Coordinación editorial
Heriberto J. Navarro
Fotografías
Archivo familiar Carreras-Lario y Grupo Carreras
Diseño y producción
Ediciones•94 (www.ediciones94.com // edic94@gmail.com)
Depósito Legal: Z/1613-2023
ISBN: 978-84-88921-82-6
Impreso en España / Printed in Spain
El papel utilizado para la impresión de este libro está calificado como papel ecológico y procede
de bosques gestionados de manera sostenible.
No se permite la reproducción total o parcial de este libro ni su incorporación a un sistema
informático, ni su transmisión en cualquier forma o por cualquier medio, sea éste electrónico,
mecánico, por fotocopia, por grabación u otros métodos, sin el permiso previo y por escrito, de
los titulares del copyright. La infracción de los derechos mencionados puede ser constitutiva de
delitos contra la propiedad intelectual (Arts. 270 y siguientes del Código Penal).
ÍNDICE
Prólogo
Noventa años del Grupo Carreras,
por Miguel V. Carreras ...................................................................... 7
Introducción
Una biografía, una larga entrevista ................................................. 9
Capítulo I
Miguel Carreras, una vida plena
El empresario aragonés a sus 89 años ............................................... 15
Capítulo II
Los pilares del éxito
El sentido común y otras cualidades del empresario ......................... 35
Capítulo III
Nace la Sociedad Carreras
Los orígenes. La primera generación ................................................ 59
Capítulo IV
Los comienzos de Transportes Carreras, S.A.
De ayudante de chófer a administrador único................................... 71
Capítulo V
Carreras Grupo Logístico
El buque insignia del holding familiar .............................................. 87
Capítulo VI
Eizasa Grupo Inmobiliario y Eizasa Hoteles
Llega la diversificación y el placer por construir .............................. 107
Capítulo VII
Explotaciones agrícolas
El amor a la tierra y el deseo de un grifo ......................................... 121
Capítulo VIII
La Fundación Carreras y otras obras sociales
Devolver a la sociedad parte de lo que se recibe de ella.................... 131
Capítulo IX
Sostenibilidad y Medio Ambiente
Proyecto Carreras 0,0: la logística en verde ..................................... 143
Capítulo X
Premios y reconocimientos
Nueve décadas y numerosas distinciones ........................................ 151
Epílogo
Lecciones de vida ......................................................................... 159

Miguel V. Carreras. De vocación, empresario7
PRÓLOGO
Noventa años del Grupo Carreras
Quien me conozca, se extrañará de que haya tenido la inmodestia de
publicar mi biografía, y no le faltará razón, pero la realidad es que mis
hijos se han empeñado en que se escriba y, de esta forma, dar a conocer
cómo se han ido desarrollando los 90 años de vida de Carreras Logística
y del resto de empresas del Grupo. El objetivo es que sirva de ejemplo y de
estímulo para mis trece nietos y también para los excelentes colaboradores
actuales de todas las empresas, para que se familiaricen con ellas, les
cojan cariño y conozcan el esfuerzo, la perseverancia generacional, el
amor al trabajo, el rigor, la lealtad, la responsabilidad y el tiempo que
han sido necesarios para llegar hasta donde están hoy. Quiero recordar
también a todas las personas que han trabajado durante estas nueve
décadas en las empresas, que están jubilados o han fallecido, a quienes
guardo mi mayor cariño y gratitud por su valiosa aportación.
En este libro narro la historia del Grupo Carreras, que no ha sido
un paseo, ni el maná, sino el relato de cómo se consiguen los objetivos
y proyectos: con trabajo y más trabajo, con jornadas de 14 y 16 horas
diarias, sábados y domingos, con humildad, austeridad y honradez.
Para ello, he contado con la colaboración de Ana Rioja, periodista
y escritora, que ha puesto toda su ilusión y oficio en este empeño, en
escucharme, ordenar y redactar todos los acontecimientos importantes
de mi vida en relación al Grupo Carreras, así como todo lo que he ido
aprendiendo a lo largo de estos años, para realizar mi biografía, la de
un hombre con una clara vocación: la de ser empresario.
No soy quién para aconsejar y, mucho menos, para dar ejemplo
a ningún colega empresario, ni a nadie. Durante mis 70 años de
Ana Rioja Jiménez8
trabajo en la empresa, todos los días he aprendido algo, muchas
veces de personas que consideraba inferiores a mí profesionalmente,
por formación o por el puesto que desempeñaban. Recuerdo que,
cuando iba a la finca, hablaba mucho con los pastores y me quedaba
maravillado de su inteligencia natural y sabiduría. Conocían a todas
las ovejas y sabían de quién era hija cada una, o cuando un cordero
le estaba robando la leche a una que no era su madre. Sabían cuándo
iba a llover observando las nubes, cuándo llegaba tiempo de sequía
porque había plantas que solo salían en época secas y las liebres criaban
menos porque auguraban menos hierba. Pero es que también sabían
de política. Conocían mejor que yo los nombres de los presidentes de
los países más importantes y todos los acontecimientos recientes: golpes
de estado, guerras, huelgas, accidentes... Sabían de economía y hasta
de bolsa, porque siempre les acompañaba su transistor, ese pequeño
aparato que mitigaba su soledad y les acercaba a otros mundos. Así,
todos los días me enriquecía con su conversación.
De este modo, fui aprendiendo de mecánicos, albañiles, financieros,
empresarios... y sigo aprendiendo todos los días de las personas que
tengo la suerte de tener cerca, colaboradores de mi equipo. Conociendo
mis limitaciones, no voy a ser tan osado de pretender dar lecciones
de empresa a nadie, Dios me libre, pero confío en que este libro y
mi relación personal con ellos refuercen la formación de mis nietos,
inteligentes y trabajadores. Seis de ellos, ya licenciados y con muy
buenas notas, ya me han superado en los estudios. Estoy seguro de que,
conforme vayan incorporándose a las empresas, van a hacerlo mejor
que yo. Esa es mi ilusión y la intención de esta publicación.
Miguel V. Carreras
Miguel V. Carreras. De vocación, empresario9
INTRODUCCIÓN
Una biografía, una larga entrevista
Este libro se comenzó a escribir un martes de primeros de abril de
2023, a las cinco de la tarde, una hora muy taurina, pese a que al
protagonista de esta publicación no le gustan nada los toros, el año
en el que el Grupo familiar de empresas que preside, con Carreras
Grupo Logístico como buque insignia, celebra su 90 Aniversario.
Cuando crucé las puertas de las oficinas de Eizasa, en la avenida César
Augusto, 27, de Zaragoza, lo confieso: estaba nerviosa.
Escribir un libro sobre este gran empresario aragonés: Miguel Valentín
Carreras Calvete, nacido en la comarca de Los Monegros (La Almolda,
Zaragoza, 1934), tan polifacético, reconocido y premiado, era un gran
reto, una aventura tan compleja como ilusionante. Las cifras de sus em-
presas, el Grupo Carreras, formado por Carreras Grupo Logístico y sus
filiales, Eizasa Grupo Inmobiliario, Eizasa Hoteles y varias explotaciones
agrícolas, sin olvidar uno de sus mayores orgullos: la Fundación Carreras,
son avasalladoras; pero detrás de cada número, de cada éxito empresarial,
hay un hombre liderando a un gran equipo, hay siete décadas de arduo
trabajo: mañana, tarde y noche; porque es cuando todos duermen cuan-
do afloran sus ideas. Y yo quería acercarme al ser humano, descubrir y dar
a conocer a los lectores de este libro de qué pasta está hecha este empresa-
rio, cuyo imponente apellido está impreso en numerosas naves, camiones
y rótulos que se pueden leer por toda España y buena parte de Europa.
Me presenté a la muchacha de la recepción. «He quedado a las cinco
con don Miguel Carreras», le dije. Y me preguntó que de qué banco
venía. «No, no —le contesté—, Soy una escritora y periodista. Vengo
a hacerle una entrevista». No quise pronunciar el nombre de biografía.
Aún era una palabra demasiado grande y hermosa.
Ana Rioja Jiménez10
Me anunció y me acompañó a una salita de espera. No pasaron
ni cinco minutos cuando otra persona me llevó hasta su despacho.
Aún no estaba decidido que yo fuera la persona que iba a escribir
este libro, su biografía, aunque a mí me gusta decir que es una larga
conversación que se ha sucedido durante cuatro meses, durante las
horas que he conseguido robarle a su trabajo, arrebatarle a su aún
apretada agenda. En ella relata su vida con su propia voz y desvela
con generosidad lo que ha aprendido del mundo de la empresa tras
muchos años de trabajo y esfuerzo, desde abajo, para llegar a lo más
alto, cimentando los pilares que sostienen sus empresas y su éxito.
Porque Miguel Carreras tiene 89 años, pero no está jubilado.
Renunció a esta pensión y se dio de alta en nómina para continuar
con su trabajo y sus proyectos, recibiendo un sueldo por ello. El
trabajo lo mantiene vivo, en plena forma, porque realiza una incesante
actividad, atendiendo a distintos temas de las empresas que preside,
una creada por su padre y sus tíos y las restantes por él, las que vio
afianzarse y crecer, y para las que tiene grandes planes de futuro.
Es un buen empresario, sí, pero también un ser humano, un hombre
casi nonagenario, un gran hombre con una lucidez extraordinaria,
una vitalidad y una ilusión por la vida que no me esperaba que tuviera
a su edad. ¿Pero qué es la edad? Tan solo un número, muchas veces
ingrato y cruel, que no hacía justicia a la persona que tenía enfrente.
Con una sonrisa y unos ojos vivaces que no delatan sus años, me
recibió con amabilidad, me dio mi lugar y me escuchó.
En nuestro primer encuentro le conté con más detalle quién soy,
los libros que he publicado (le llevé varios) y la idea que tenía para
escribir su biografía. Me dejó hablar y, pasados unos minutos, me
dijo: «Pero yo no hago teatro ni música. Creo empresas y pienso que
mi experiencia, todo lo que he aprendido en estos 70 años de trabajo,
puede ser interesante, sobre todo para mis nietos, para los jóvenes
emprendedores y para las personas que dirigen empresas y tienen que
enfrentarse a diario a decisiones importantes».
Miguel V. Carreras. De vocación, empresario11
Pensé: «Yo soy una escritora, una periodista especializada en temas
de cultura. Quizá no soy la persona más adecuada para escribir su
libro». Me repuse pronto, me armé de valor y le dije:
—Soy periodista y si algo he aprendido en todos estos años de
profesión es a escuchar, a sacar de los entrevistados lo mejor de ellos,
de su vida, su obra y su legado, y escribirlo de la forma más fiel posible,
pero también amena para el lector. Tengo cierta habilidad para que
las personas me abran sus mentes y sus corazones.
—Para escribir este libro tienes que pensar en clave empresarial —
añadió.
—Confíe en mí —le contesté—. Si algo sé es preguntar y seguir los
temas hasta que los conozco como la persona a la que tengo enfrente,
hasta que los hago tan míos como suyos.
Se rio. Miguel Carreras es una persona que sonríe mucho. Y me
recomendó la lectura de un libro que le marcó hace años: «Cartas de un
empresario a su hijo», de G. Kingsley Ward (Ediciones Deusto, 1992),
unas páginas fascinantes que hablan del empresario y del hombre
que, como Miguel Carreras, tiene en su vida varias constantes para
crear todo lo que ha construido, para llegar hasta donde está hoy: el
sentido común, la responsabilidad, la confianza en el factor humano y
la ilusión. Palabras que se repetirán con frecuencia en este libro. Pero
hay más, muchas más: el amor al trabajo y a la familia, la humildad, la
fe en uno mismo, el riesgo, la valentía, la ambición sana… y la suerte.
Y, sobre todo, la palabra familia, tan importante en la vida de Miguel
Carreras. De hecho, su Grupo está formado por empresas familiares en las
que han trabajado tres generaciones, desde su padre y sus tíos, a su hermano
y sus hijos, controlando el accionariado y marcando el rumbo de todas
las decisiones. Y su esposa María Natividad Lario, Marinati, una mujer y
compañera de vida sin la que no se podrían escribir las páginas de este libro.
Miguel Carreras tiene grandes cualidades que le hacen una persona
muy especial y carismática, con una filosofía de lo que es la vida y el
trabajo muy personal. Un modelo de empresario atípico, que empezó
Ana Rioja Jiménez12
su carrera desde abajo, como ayudante de conductor de camión, hasta
llegar a lo más alto; tan diferente a la imagen del empresario actual,
que llena su currículum de carreras universitarias y másteres, que
acapara páginas de los medios de comunicación y al que tan poco se
entiende cuando habla. Miguel Carreras es claro, transparente, mira
a los ojos a su interlocutor y contagia ilusión y pasión por su trabajo,
por sus empresas, por su legado, por la vida. Pero todo ello lo iremos
descubriendo en los diferentes capítulos de este libro.
A la segunda entrevista llegué con la obra que me había
recomendado leída y subrayada.
—Sí —le dije—, entre los dos vamos a escribir un buen libro.
—No sé yo, no sé si mi vida dará para tanto —me contestó.
—Pero esto no va a ser como un folleto publicitario lleno de gráficos
y datos —añadí—, para eso ya está la página web de la empresa. Va a
ser un libro, con un tono literario en el que hombre y empresario van
a ir de la mano.
Poco a poco lo fui convenciendo y arrancándole más horas para
vernos.
—Comenzamos a escribirlo ya y no lo dejaremos hasta que
terminemos. Ni usted ni yo tenemos mucho tiempo —me atreví a
decirle.
—Oye, que no pienso morirme aún —me contestó y se rio con
ganas, con esa risa suya que le hace parecer mucho más joven, con la
que me ha obsequiado tantas tardes de conversación y que a mí me
encanta.
—No, no, que no quería decirle eso —le expresé roja de
vergüenza—. Solo que si el Grupo Carreras celebra en septiembre su
90 Aniversario, me gustaría que el libro estuviese listo para esa fecha.
—Bueno, pues vamos a ver si entre los dos lo conseguimos.
Y aquí están nuestras conversaciones, vertidas en estas páginas,
llámese biografía o una larga entrevista. El subtítulo con el que
Miguel V. Carreras. De vocación, empresario13
finalmente aparezca da igual. A sus compañeros empresarios, a los
jóvenes emprendedores, a la gente del mundo de la economía y las
finanzas quizá les interesen más las cifras, los datos de sus empresas, las
divisiones de estas, su expansión a nivel nacional e internacional; pero
yo creo, y he apostado por ello, que a su familia, a las personas que
lo quieren y admiran, a sus clientes, colaboradores y trabajadores (su
segunda familia) les emocionarán más sus palabras y los sentimientos
que se perciben detrás de cada una de ellas.
Porque Miguel Carreras, en las conversaciones que hemos
mantenido para escribir este libro, se ha abierto en canal. No ha
escatimado anécdotas, sentimientos ni palabras para contarme su
vida, siempre en relación a su trayectoria como empresario.
Y, al final, de verdad creo que esto es una gran biografía, pero
sobre todo es una radiografía de su alma de hombre y de empresario,
porque en Miguel Carreras no se pueden separar ambas facetas.

Miguel Carreras y Ana Rioja, durante una de sus conversaciones.

Miguel V. Carreras. De vocación, empresario15
CAPÍTULO PRIMERO
Miguel Carreras, una vida plena
El empresario aragonés a sus 89 años
«Vamos a comenzar a escribir este libro por el final —le dije a
Miguel Carreras en nuestra segunda entrevista—. Quiero saber
cómo se siente hoy, qué balance hace de su vida». Se sorprendió,
pero me contestó: «De acuerdo, me dejo llevar». Deseaba
comenzar este libro con las palabras de un hombre que, cercanos
los 90 años, se muestra pletórico, lleno de ilusión, que todos los
días va a trabajar, a dar vuelta por sus empresas, y que aún le
brillan los ojos cuando habla de proyectos. ¿De dónde saca esa
energía y esa fuerza?
—Acaba de cumplir 89 años, 70 como empresario y se va a celebrar el
90 Aniversario del Grupo Carreras. ¿Cómo se siente? ¿Es feliz?
—No estoy en mi mejor momento físicamente, por supuesto,
porque la juventud es muy importante, pero soy un hombre feliz.
Tengo mucha ilusión por la vida y por mi trabajo, y aún tengo energía.
Amo a mi familia, es extraordinaria: mi esposa, mis hijos, mis nietos,
y quiero a mis amigos, a mis empleados. Tengo cantidad de amigos
que han sido o son empleados míos y estoy muy a gusto con ellos.
—No aparenta su edad. Usted dice que no sabe dónde se han metido
todos esos años que tiene.
—Mi cabeza no ha envejecido. Mi esposa me dice que siempre
pienso como si tuviera 20 años. No, no pienso que soy mayor. No
estoy tan ágil físicamente, pero la cabeza la tengo perfecta. Tengo
mucha memoria, mucho raciocinio y, sobre todo, mucha experiencia.
Ana Rioja Jiménez16
Eso hace que los problemas no me angustien. Otros ven grandes
inconvenientes en temas que veo sencillos, porque he resuelto
problemas mayores. Todo en esta vida es muy relativo. He sacado
adelante proyectos muy importantes y difíciles. Y ahora, no hay
problemas importantes que me hagan apearme de un proyecto. Los
miro a la cara, los estudio con tranquilidad y sin agobio, y así se
resuelven fácilmente.
—Va todos los días a sus empresas y sigue muy de cerca los nuevos
proyectos.
—Es lo que he hecho toda mi vida. Sí, voy todos los días a las
empresas y mi ilusión es esta: dirigir proyectos, ver cómo se van
cumpliendo y realizando, y se van acercando al objetivo.
—Setenta años trabajando… ¿no piensa en descansar?
—Sí, este año he cumplido 70 años de trabajo en la empresa. Entré
con 19 y no estoy cansado ni me he aburrido en ningún momento,
porque realmente no he trabajado nunca, me he divertido, es como
un hobby, he ido a jugar allí con la empresa, es como un juego,
conocer a más personal, motivarlos… No puedo dejar de ir a ver a
mi gente. Ojalá tuviera otros 70 años para continuar en la empresa y
estoy seguro que no me cansaría nunca, porque ya me encargaría yo,
y mi gente también, de marcar proyectos ambiciosos que tendrían
cultivada nuestra atención, desvelo, ilusión y esfuerzo, y siempre
tendríamos un proyecto u otro que sacar adelante, que es lo que hace
falta. Nunca nos cansaríamos.
—¿Le molesta que le digan que tiene adicción al trabajo?
—Dicen que soy muy trabajador, que tengo adicción al trabajo,
pero no la tengo. Lo que ocurre es que desde muy joven me ha
gustado mucho la empresa y estoy en la empresa para disfrutar. Sigo
trabajando porque eso me hace feliz. Y soy feliz, afortunadamente,
con los éxitos que tenemos y viendo cómo me apoyan mis hijos y los
Miguel V. Carreras. De vocación, empresario17
directivos, cómo hacen suyos mis proyectos y disfrutan ellos también.
¿Cómo no voy a ser feliz en la empresa si veo que cada día está mejor?
Voy a la empresa, pero no voy a trabajar, voy a disfrutar. Disfruto
viendo a mis empleados y la ilusión que tienen con las cosas que
hacemos, sabiendo que me acompañan, que comparten mi proyecto.
Eso me llena de satisfacción.
—Es cierto. Usted siempre comenta que necesita estar cerca de sus
trabajadores.
—Me encanta estar con los trabajadores, ver cómo te apoyan, cómo
te han comprado ese proyecto, ese objetivo y cómo disfrutan ellos
también, porque en las empresas lo importante no solo es trabajar y
tener éxito en los proyectos, lo importante es disfrutar, como disfruto
y mi gente también. Veo las sonrisas y el bienestar de los trabajadores,
cómo va aumentando el personal, porque cuando yo empecé había
solo un conductor contratado y hoy somos cerca de 3.000 personas
Miguel Carreras en su despacho de Carreras Grupo Logístico en PlaZa.
Ana Rioja Jiménez18
en las empresas, pues cómo no voy a disfrutar, cómo no voy a ir
todos los días a la empresa a ver a la gente, a saludarles, a hablar con
ellos, a comentar las cosas y a participar de su éxito; pero no soy
ningún adicto al trabajo, no voy por dinero. Nunca he trabajado por
dinero, puedo asegurarlo. Voy porque me divierte. Me gusta estar con
la gente que me sigue y que está sacando los proyectos adelante, que
comparten mi ilusión, mis ganas y mi ambición por hacerlo bien y
ser los mejores. ¿Con quién voy a estar mejor? Ese es mi estado actual.
La pena es que tengo ya muchos años y llegará un día en el que me
fastidiará no acompañarlos.
—¿Y los clientes? ¿Son la razón de ser de todo este trabajo?
—Satisfacer al cliente es el objetivo de las empresas. Una empresa
sin clientes no tiene sentido. Eso es lo que todos buscamos. El cliente
es el rey de la empresa, el último destinatario y todo gira alrededor de
él, del gusto, de la necesidad y de las exigencias del cliente. Trabajamos
para darle satisfacción en tiempo y forma. Y para dar este servicio
es necesario tener un personal leal, responsable y profesional. Con
su buen hacer y medios podemos comprometernos a dar un buen
servicio y conseguir fidelizar a grandes y buenos clientes, como los
que tenemos ahora.
Ya son 90 años desde nuestra fundación, nueve décadas de
historia, de experiencia y de servicio, y no podemos dejar de estar
orgullosos del largo camino recorrido, que no hubiera sido posible
sin la confianza de nuestros clientes. Todo el trabajo de la empresa,
del personal, de los sistemas, de los desarrollos logísticos… van
encaminados a dar el mejor servicio, de la manera más eficiente
posible y al coste más competitivo. De ese modo, les manifestamos
cada día nuestro más profundo agradecimiento. Nuestras relaciones
con clientes, proveedores y colaboradores son de largo plazo y se
basan en la honestidad, la confianza, la transparencia y la integridad.
—Siempre pensando en lo que se puede hacer, en ser mejores.
Miguel V. Carreras. De vocación, empresario19
Miguel Carreras, orgulloso de ser aragonés, junto a su esposa, su hija Cristina
y dos de sus nietas.
Ana Rioja Jiménez20
—Queremos ser cada vez mejores y aumentar nuestros medios
para cumplir nuestra vocación de servicio al cliente. Para ganar su
confianza es necesario darles buen servicio, ser serios y competitivos,
y procurar su satisfacción. Me llueven cantidad de ideas que
comento con mis hijos y aún me toca tomar decisiones importantes
sobre proyectos y, sobre todo, de inversiones muy fuertes. Siempre
me ha gustado arriesgar, pero no de una forma desordenada.
Estudiando los casos, he ido casi siempre al límite, pero ponderado.
Los apalancamientos en las empresas siempre los he hecho de forma
que pudiese dormir, porque cuando me comprometía a devolver un
préstamo o una inversión, nunca ha sido a un corto plazo, sino a
medio y largo plazo, porque yo sabía que era para un proyecto que
iba a tener su rentabilidad. Eso me hacía estar tranquilo porque
una buena parte o en su totalidad se iba autofinanciando. Y si
no se autofinanciaba tenía también otros proyectos, otros valores
importantes, tanto mobiliarios como inmobiliarios que me daban
tranquilidad por si hacía falta ayudar a los nuevos. Como los demás
estaban ya financiados, no había ningún problema en hacerlo.
Siempre he ido muy fuerte con eso.
—Y una constante: invertir y reinvertir
—Mi padre y yo siempre hemos reinvertido todos los retornos
que nos daban las empresas, todas ellas. De ninguna he cogido nunca
un céntimo, todo lo he reinvertido, pero no solamente el beneficio,
como mínimo invierto el 100 % del beneficio y en algunos casos
el 200 %, siempre en activos productivos y sumado al resto de los
activos. De este modo, no he tenido ningún problema en ir cubriendo
y pagando los créditos y sus intereses. Las inversiones las hago a largo
plazo no a corto plazo. Porque, para que una empresa prospere, no
solo hay que invertir los beneficios, sino que debe apalancarse en
la garantía de su patrimonio. Con una administración austera, si el
objetivo es correcto y el proyecto está bien calculado, los números
salen. Y detrás de un proyecto está otro, porque lo importante es
Miguel V. Carreras. De vocación, empresario21
tener ideas, estudiarlas, analizarlas, ponerlas en valor y realizarlas con
el fin con el que se proponen, que salgan bien, tener éxito, porque si
no lo tienen…
—Y siempre en Zaragoza y Aragón. ¿Ha tenido ofertas para trasladar
sus empresas?
—Sí, por supuesto. Nos han propuesto que trasladásemos nuestra
sede a Cataluña, Navarra, Madrid... pero nunca he pensado en hacerlo y,
al contrario, nosotros tenemos actividades en toda España, en Canarias,
en Baleares, en Portugal y en Francia, pero todos los impuestos se pagan
aquí, en la Diputación General de Aragón. Todo lo que facturamos, más
de 400 millones de euros en todas nuestras actividades del Grupo, se
liquida en Aragón. Esto también es un orgullo y una satisfacción.
—¿Cómo es para usted un día de trabajo?
—Me levanto a los 8 de la mañana, hago media hora o tres cuartos
de hora de ejercicio (bicicleta, cinta y un poco de gimnasia), me ducho,
desayuno y leo la prensa. A las 11 llego a la oficina, a mi despacho en
PlaZa, y estoy hasta las 14.30 horas. Tengo una persona que me lleva
la agenda y ya empiezo a recibir visitas y a mantener reuniones: recibo
a bancos, clientes, jefes de negociado que me explican cómo va su
división. Sobre todo a bancos, directores que quieren hablar conmigo
para pedirme, para darme…
Comento con mis hijos cómo van los proyectos, los éxitos,
los problemas que tenemos. Estoy con la directora de personal, el
responsable de conductores, los mecánicos de camiones, el responsable
de logística y también con mi asesor fiscal privado, comentamos cómo
van las finanzas. Y me voy a comer a casa, si no tengo comida con
algún cliente, proveedor o amigo.
Por la tarde voy a Eizasa, desde las 16.30 horas hasta las 20 horas
y aquí despacho con la directora de hoteles: cómo va la ocupación,
cómo ha ido el fin de semana, problemas de personal, reformas y
Ana Rioja Jiménez22
mantenimiento, compras, sugerencias… También despacho con
mi hijo Jesús y otra responsable de la inmobiliaria, comentamos
las ventas, los costes, las obras nuevas... Analizamos y estudiamos
nuevos proyectos, y los problemas que se presentan continuamente,
reclamaciones de algunos clientes, también recibo a bancos
relacionados con esta empresa para financiar hipotecas y demás.
—Escuchándolo, vida y trabajo van tan íntimamente unidos, que su
vida es su trabajo.
—No puede ser empresario el que quiere tener un trabajo
cómodo. La empresa te exige la vida entera, porque el trabajo no es
una condena, es una bendición. Y, si tienes salud, ya es lo más grande.
Sin trabajo no existiríamos. El trabajo te permite realizarte tú y lo
que llevas dentro. Te da satisfacciones a ti y a los demás. Gracias al
trabajo te distraes, te enfocas, te superas, prosperas, das ejemplo a los
hijos… y, en muchos casos, alcanzas el éxito. Trabajar no es un castigo
y es lo que nos hace personas. Todos los inventos y los avances se han
conseguido trabajando en los distintos sectores. El trabajo dignifica
a las personas. No puedo separar la vida de mi trabajo, porque es mi
ilusión. Las empresas las he creado yo y para mí todas son como hijos.
—Pero, ¿cómo ha podido dirigir todas estas empresas a la vez?
—Porque una me llevaba a la otra y me volcaba en los nuevos
proyectos sin desatender los ya iniciados, que se iban ampliando.
Aun hoy tengo en mi mente todos los proyectos y sus problemas:
del transporte, de la logística, de los hoteles, de la inmobiliaria, de la
agricultura y de la bolsa. Tengo grandes planes para ellos. Y todos los
días los repaso. Desde los 19 años tenía ideas nuevas que trastornaban
a mis padres y a mis tíos. He hecho siempre lo que he querido y
para mí es una costumbre, una forma de actuar, que hace falta ser
administrador único para obrar así y desde bien joven lo fui. Mi
padre y mis tíos confiaron en mí. He tocado todos los temas que me
apetecían y, conforme los iba creando, los iba asimilando. No dejaba
Miguel V. Carreras. De vocación, empresario23
uno para coger otro, iba sumando y sumando. Tuve la capacidad de
ir sumando todos. Y ahora si hiciera otra operación, también. Para mí
era fácil y ha sido muy importante escoger el equipo adecuado para
cada empresa.
—Asegura que nunca ha trabajado por dinero. ¿Ha sido un medio y
no un fin?
—No, nunca he trabajado por dinero, sino por proyectos, por
sueños, por ambición de la buena. Soy un ambicioso sano, pero
no por dinero, porque he vivido siempre de mi sueldo. Como he
comentado, mi empresa nunca ha repartido dividendos. Empezó con
esta política mi padre y la he seguido. Todo se reinvierte. La ilusión y
la satisfacción no la da el dinero, solo es un medio. La satisfacción la
da el éxito. Si lo tienes, te da dinero, pero no te preocupa. El dinero
es un medio necesario para ampliar y expandir el negocio.
Yo perseguía el éxito de los proyectos y con ello me venía todo.
Aportaba capital, lo invertía y, sin darme cuenta, iba aumentando
camiones, naves, delegaciones, empleados… hasta las treinta y tantas
delegaciones. He sido muy feliz con todo lo que he creado y lo sigo
siendo con los nuevos proyectos, con lo que te falta por crear. El
dinero es un medio, nunca un fin.
—¿Qué parte de este éxito se debe a su esposa? Hábleme de lo que ha
supuesto en su vida y en su familia.
—Marinati (María Natividad Lario) ha sido y es todo en mi vida,
no sólo porque juntos hemos formado una gran familia, que es mi
orgullo, mi mayor éxito, mucho más que el de mis empresas. Es dos
meses menor que yo, y los dos estudiábamos Peritaje Mercantil en
Zaragoza cuando la conocí, y me costó conquistarla, porque era una
muchacha muy seria y muy trabajadora. Era de Daroca, muy guapa
y con moño. Iba con una amiga mía y con esa excusa comencé a
acercarme y ya empezó a aceptarme. Pero lo que más me conmovió
fue su bondad y su entrega a la familia. En nuestros 25 primeros
Ana Rioja Jiménez24
años de casados, yo no paré de viajar y ella se quedaba en casa con los
siete hijos, que se llevaban 14 o 15 meses uno de otro, cuidándolos y
educándolos. Y después, cuando ya fueron mayores, preocupándose
por cada uno de nosotros. Nunca se ha cansado de nada ni me ha
reprochado que trabajara tanto. Ella es el alma de nuestra familia.
Menudo descanso he tenido con ella.
María Natividad Lario es aún, a sus 89 años, una mujer preciosa,
alta y esbelta. En mis encuentros con ella, relacionados con la
escritura de este libro, se ha mostrado acogedora, afable, cercana
y tremendamente humana. Dicen que detrás de cada hombre hay
una gran mujer. A mí me gusta más decir que junto a un gran
hombre hay una gran mujer. Es una frase hecha, pero totalmente
cierta. Allí ha estado ella siempre. Ha creado el clima propicio en su
hogar para que su esposo pudiera desarrollarse como empresario
y como padre. Con toda esa entrega y ese esfuerzo, ha alentado
cada proyecto, avance y éxito de su marido, acompañándolo a
todos los actos, aplaudiendo cada logro y cada premio. Ella crio y
educó a siete hijos, sin reproches pese a sus ausencias, y, cuando
estuvieron formados y con sus estudios terminados, vio con
orgullo cómo se iban incorporando a las empresas para trabajar
junto a su padre y seguir desarrollando su legado.
—¿El mayor patrimonio de su vida son sus hijos?
—Sí. Algunos han pensado que mi mayor éxito en la vida era
la empresa, otros el dinero o el poder, y yo siempre he tenido claro
que lo más importante era mi familia. Algún amigo me ha dicho:
«tú has acertado, yo me he equivocado con la familia». Llevo ya
en Zaragoza 70 años moviéndome en empresas, cogiendo gente,
despidiéndola, tratando con unos y con otros, en un montón de
actividades. Y mis hijos me dicen: «Nadie habla mal de ti, papá». Eso
también es un capital muy importante, más que el dinero, que tus
hijos se sientan orgullosos de ti.
Miguel V. Carreras. De vocación, empresario25
Miguel Carreras y su esposa María Natividad Lario,
toda una vida juntos.
Ana Rioja Jiménez26
—Y su mayor satisfacción, que seis de sus siete hijos trabajaran junto
a usted.
—Es muy importante para un empresario, que la empresa esté
unida a la familia. Tiene que estar ensamblada con la familia, que
haya unión y armonía entre ambas partes. Mi ilusión era que mis
hijos no tuvieran que salir fuera para trabajar. Primero, mi sueño era
crear empresas y proyectos para que ellos no tuvieran que marcharse;
y segundo, para que estuvieran al frente de ellas, como accionistas,
porque hace más ilusión trabajar como accionista que de empleado.
Diversifiqué en varios sectores. Y me ha salido bien, porque hoy
los dirigen todos ellos. Excepto la mayor, Cristina, que se marchó
muy joven a estudiar a Sevilla, y allí se quedó porque encontró su
vocación y es profesora de Universidad, el resto está trabajando con-
migo. En el transporte y la logística José Luis, Fernando y Ricardo;
en agricultura, Miguel Ángel; con los hoteles y la inmobiliaria, Jesús;
en la asesoría jurídica del Grupo y en la Fundación, Carmen. Por
desgracia Pepe (José Luis), que es el primer hijo que vino a trabajar
conmigo, falleció hace tres años a consecuencia del Coronavirus. Esa
es mi gran tristeza.... Hoy, cinco de mis hijos están trabajando en el
Grupo de empresas. Que, por otra parte, es doble satisfacción, que
vaya a las empresas y estén mis hijos. Hay compañeros míos que ven
a los hijos una vez al mes o una vez al año. Yo los veo todos los días,
esa es otra gran suerte que tengo, mi mayor éxito familiar.
—¿La vocación empresarial se hereda? ¿La tienen sus hijos?
—Otros empresarios no han conseguido esto con sus hijos, que han
preferido ser, por ejemplo, profesionales independientes o funcionarios;
pero la culpa la tienen ellos. Yo, cuando regresaba a mi casa después
de un duro día de trabajo, les hablaba con satisfacción de la empresa;
otros iban muy enfadados y hablaban mal de los trabajadores, de los
proveedores o de los clientes, y sus hijos pensaban: «pues cualquiera se
mete en esa empresa con los líos que tiene mi padre». A mí me veían
Miguel V. Carreras. De vocación, empresario27
que era feliz y que disfrutaba con mi trabajo por arduo que fuera. He
tenido mucha suerte con mis hijos, sin ningún esfuerzo terminaban sus
carreras universitarias y rápidamente se metían en la empresa, porque
tenían ganas de participar en ella, les apetecía porque veían la ilusión y
la felicidad que tenía con ellas. Eso ha sido importante.
Y muy importante ha sido que hayan aprendido a valorar lo que
se ha hecho y cómo se ha hecho, con mucho trabajo y céntimo a
céntimo, aquí no hay premios ni loterías.
—Y continúa al lado de sus hijos, preocupándose por sus proyectos,
avances, ¿podría decirse que tutelándolos?
—Siempre les ha venido bien que estuviera a su lado, para acon-
sejarles, para evitarles algún problema, error o riesgo; muy cerca, para
estimularles y celebrar los éxitos juntos. Sí, les he ido formando y todos
tienen vocación empresarial y disfrutan también en las empresas.
—¿No le da miedo que no se conserve lo que ha creado cuando ya no
esté o que no se terminen los proyectos que ha iniciado?
—He educado a mis hijos y los educo para esto, para que continúen
llevando las empresas a lo más alto. Algo muy importante. Me da
mucha pena ver cómo grandes empresarios han tenido que vender su
empresa porque no tienen sucesión, porque sus hijos no han querido
trabajar en la empresa de sus padres. Pero, como he comentado antes,
la culpa la tienen ellos, porque en lugar de ir a casa alegres, contentos
del trabajo, contando con ilusión lo que habían hecho y conseguido,
iban quejándose de los obreros, de los sindicatos, del Gobierno… Y
no motivaron a sus hijos.
Yo, con mis hijos, al revés. De siete que tenía, cinco están en las
empresas y disfrutan con ellas, y ya en el testamento vigilo todo eso.
No les dejo vender a ninguno nada. Las empresas solo se pueden
vender con mayoría absoluta, pero para reinvertir en otra cosa.
Tendrán que reinvertir la mayor parte de los beneficios, como hizo
Ana Rioja Jiménez28
Miguel Carreras con Fernando, Carmen y Ricardo,
los tres hijos que trabajan en Carreras Grupo Logístico.
Miguel Carreras con su hijo Jesús, que trabaja junto a él en Eizasa.
Miguel V. Carreras. De vocación, empresario29
Su hijo José Luis Carreras, fallecido en 2020,
fue consejero delegado de Carreras Grupo Logístico.
Miguel Carreras con su hijo Miguel Ángel, que dirige las explotaciones agrícolas.
Ana Rioja Jiménez30
mi padre y como he hecho yo. Y que los proyectos, que saben cuáles
son, los lleven con éxito, porque son muy importantes. Viven de sus
sueldos, que son acordes a sus cargos.
—Usted, pese a lo que pudiera parecer, ha vivido siempre de su sueldo.
¿Predica la austeridad también con sus hijos?
—También. A mis hijos los he educado con austeridad y viven
todos de su sueldo y el dinero que les deje en metálico de mi
patrimonio personal les sobrará, no saben gastarlo, como yo, igual.
Es como el árbol pequeño. Hay que educarlos con humidad y
austeridad. El destino es distinto según cómo los educas. La empresa
es la empresa, no eres tú. La empresa te empuja. Tienes que reinvertir
todo y te lo exige todo: más clientes, más camiones, más medios.
Conquistar el lugar que estamos ocupando a nivel nacional, también
te da una motivación, porque si no eres nadie, pues vaya, pero si eres
un referente… Te da satisfacción tener algo muy importante. Y no es
por el patrimonio que te da, sino porque la empresa vaya a más.
—Y los nietos, ¿esa cuarta generación?
—Tengo seis que ya están licenciados. Mi ilusión es que vengan a la
empresa, que trabajen en el Grupo familiar y que convivan bien entre
los primos. Esa es mi mayor ilusión. Quiero dejarles un buen medio
de vida con su trabajo en las empresas, que les permitirá disfrutar de
una posición social fuerte, y mi deseo es que, además, disfruten del
trabajo; que el día de mañana Carreras sea más que lo que es hoy,
el doble o el triple, que lleva camino de ello. Su mejor herencia y
mayor orgullo será participar en esta Sociedad. Eso es mucho más
importante que el dinero.
—¿Qué haría usted si la vida le diera otros diez años más en plenas
facultades?
—Pues quizá viva hasta los 100 años (y se ríe con ganas). La edad
no es la del carné. Empecé a cuidarme hace 15 años, que dejé de
Miguel V. Carreras. De vocación, empresario31
fumar, y he estado hasta los 75 años jugando al tenis. Si la vida me
diera diez años más seguiría tutelando nuestros grandes proyectos,
muy cerca de mis hijos y de mis nietos. Mi ilusión sería que Carreras
Logística estuviera como está en España, en Francia; que nuestra
cadena tuviera más hoteles; con Eizasa construir cientos de viviendas
más para que la gente las disfrutara y dijera qué bien construye esta
empresa; que en agricultura tuviéramos grandes extensiones de olivos
y que comercializáramos un buen aceite para todo el mundo. Y me
lo pasaría genial con todos estos proyectos, porque voy a trabajar y
todo son satisfacciones. Yo les digo a mis hijos que todo lo que hemos
creado en 90 años se puede multiplicar en unos pocos. Porque ahora
se avanza mucho y muy deprisa. Lo que antes nos costaba un año,
ahora se hace en un día. En diez años pueden avanzar más que en
toda mi vida.
Yo los empujo y aconsejo, pero no me puedo poner a liderar
proyectos a largo plazo. Tengo que dejarlos en sus manos. Y ojalá que
en sus corazones. Ya no puedo ser la mula de varas, la que lleva el
carro, el mayor peso, la más responsable. Los carreteros cuando iban
a acarrear, cuando iban a buscar la mies para llevarla a la era, llevaban
dos o tres mulas delante, y la de varas es la que llevaba el carro y lo
sacaba del barro si se atascaba. La mejor mula, a varas. Yo siempre he
ido a varas. He sido la mula de varas de todos los proyectos, la que ha
llevado el carro de todas las empresas. Ahora ya no me puedo meter
en varas, porque no terminaría los proyectos y sería peor, pero les he
dejado el camino marcado y allanado para que sean ellos los que tiren
ahora con fuerza del carro y espero que lo hagan mejor que yo.
—¿Un consejo para sus hijos relativo a las empresas?
—Que no tengan prisa por jubilarse mientras les acompañe la
salud. Un empresario con vocación no piensa en jubilarse. Ojalá
tuviera 50 o 60 años, como mis hijos, porque en los años que se
avecinan va a ir todo muy deprisa. Les digo que en 12 o 15 años se
duplicará el Grupo. Y dónde van a estar mejor que donde tienen éxito
Ana Rioja Jiménez32
y reconocimiento, donde disfrutan. Nada les dará tanta felicidad
como esto y pensarán: «estoy en el lugar en el que tengo que estar,
donde se toman decisiones importantes y se hacen cosas importantes,
donde se mueven cientos de millones». E irán a las empresas, no para
ganar dinero, sino para que los proyectos que haya en ese momento
se encaucen y realicen. Para disfrutar con ellos y con los empleados.
Si no se está encima de los proyectos mientras la fuerza te
acompañe es como traicionarlos, abandonarlos. Es como un hijo
tuyo, estás tutelándolo, asesorándole toda la vida en la medida en que
tú tienes el conocimiento. Y no puedes consentir que fracase, porque
sería el fracaso de varias generaciones. Hay que seguir al pie del cañón
para sostener y reforzar el Grupo para que sea más grande. Hay que
tutelarlo para que ocupe un puesto importante a nivel nacional e
internacional, y eso les motivará a seguir. «Seréis los arquitectos de un
gran rascacielos —les digo— ¿Vais a dejar de disfrutar de ello?».
Y luego están los clientes. Tenemos clientes de más de 40 años,
que no nos han abandonado, y sus sucesores también confían en
nosotros. Eso da mucha felicidad. No paro de darles vueltas a mis
empresas. Mañana, tarde y noche. Mi vida han sido las empresas, la
familia y los amigos.
Pero no todo en la vida son satisfacciones y alegrías, en la familia
de Miguel Carreras ha habido en los últimos años tres grandes
pérdidas. Este empresario sabe muy bien que la muerte forma
parte de la vida, aunque llegue antes de tiempo y por sorpresa. Su
hijo José Luis, consejero delegado de Carreras Grupo Logístico
falleció en octubre de 2020 a consecuencia de la Covid-19 y en
noviembre de 2022 murió su hermano Jesús, vicepresidente de
Carreras Grupo Logístico, al que tan unido estaba. Hace unos
años también falleció su sobrino José Antonio, hijo de Jesús.
—Grandes pérdidas en poco tiempo, ¿cómo se recupera uno de esto?
Miguel V. Carreras. De vocación, empresario33
Miguel Carreras con su esposa e hijos, en una de sus fincas con los almendros en flor.
Ana Rioja Jiménez34
—No esperábamos que muriera mi hijo José Luis con 58 años.
Era un hombre muy fuerte y tuvimos esperanza hasta el final de que
se podría recuperar, pero ese maldito virus no perdonó a nadie en
los primeros meses y ocasionó muchas muertes. Fue un golpe muy
duro para toda la familia, una tristeza muy grande que tendremos
siempre, pero hay que seguir y gracias al trabajo, a las empresas, al día
y al día y a la unión y el cariño de mi esposa y del resto de los hijos,
lo vamos superando. Siempre estará con nosotros. Su despacho en
Carreras aún está sin tocar. Cuando la familia se reúne lo recordamos
para que esté presente. Fue una muerte muy sentida también para sus
amigos, para los trabajadores del Grupo y para todo el sector empre-
sarial, que destacó su excelencia personal y profesional. Le han hecho
varios reconocimientos públicos. Todo el que lo conocía, ensalzaba
su bondad.
Respecto a mi hermano, estaba muy unido a él. Desde que se in-
corporó a la empresa en 1968, han sido más de cincuenta años juntos
en los negocios y en la vida. Comíamos juntos, viajábamos juntos,
toda la vida juntos, y por una mala caída fallece en dos días. Eso
me produjo también un trauma muy grande. Es más reciente y aún
no me he hecho a la idea de que mi hermano ya no está. El trabajo
del día a día ocupa mi mente y me hace bien. Ha sido también una
muerte muy sentida por todos, que así te lo expresen, y recibir las
condolencias y el afecto de tantos, como con la pérdida de José Luis,
te reconforta.

Miguel V. Carreras. De vocación, empresario35
CAPÍTULO II
Los pilares del éxito
El sentido común y otras cualidades del empresario
Miguel Carreras es un hombre de éxito, así lo avalan la mayoría de
proyectos que ha emprendido y la buena marcha de las empresas
que él preside. Pero, ¿qué es tener éxito? ¿Cómo, cuándo y por
qué llega? ¿Qué diferencia su éxito del de los demás?
Para ser empresario hay que ser de una madera especial y tener
una vocación, que debe alimentarse de un montón de cualidades
más, porque «ser empresario es un arte y con cierto éxito, mucho
más», afirma Miguel Carreras. Un arte que es el compendio
de diversas características que definen a un hombre y que le
llevan a conseguir el reconocimiento personal y el de los demás.
Reunirlas todas no es fácil. «Por supuesto, yo tampoco las tengo
a pesar de intentarlo, porque el éxito ni llega fácil ni llega rápido,
requiere grandes dosis de trabajo y años. He trabajado mucho,
pero otros también lo hacen y no logran llevar adelante sus
proyectos, alcanzar sus sueños. Tener éxito en una empresa es un
compendio de muchas cosas y, aun reuniéndolas todas, no está
garantizado», afirma.
Vamos a desarrollar algunas de estas cualidades con las palabras
de Miguel Carreras. Si ser un buen empresario es un arte, ¿qué
valores y herramientas ayudan a ello?
—¿Qué es para usted el éxito?
—El éxito es algo muy relativo. Por ejemplo, un aprendiz de zapatero
puede terminar cumpliendo su sueño: tener una tienda de zapatos y eso
Ana Rioja Jiménez36
ya es un éxito. Entonces, el éxito no compite con nadie, sino con uno
mismo. El éxito es fácil, es proponérselo y trabajar para conseguirlo.
Hay miles de éxitos, tan orgulloso está el aprendiz de zapatero con su
tienda, como el chófer que llega a ser propietario de un camión, como
el estudiante de Medicina que logra ser un gran cirujano. Depende de
lo que te propongas en la vida y en tu empresa. Ser líder es otro tema,
llevar a cabo la dirección empresarial de un Grupo como Carreras,
hoy la presidencia en mi caso, es más complejo. Entonces, el liderazgo
implica competir con los demás, con los buenos, y demostrar que eres
el mejor, y tener o adquirir una serie de cualidades.
—En su caso, para llegar a ese liderazgo, imagino que la primera
cualidad es la vocación de ser empresario.
—Por encima de todas las cualidades de un empresario está la
vocación. La empresa es mi pasión, de las profesiones más nobles, efi-
cientes, solidarias y creativas que te permiten desarrollar tus inquie-
tudes, la imaginación, medir tu capacidad con nuevos retos, conocer
a buena gente e incorporarla a tus proyectos, conseguir su amistad.
Pero, además, quiero evidenciar que en la cultura occidental, la em-
presa privada es la que crea puestos de trabajo, crea riqueza que, vía
impuestos, nutre las arcas del Estado. Una nación es próspera si su
tejido empresarial y comercial es amplio y rico.
—¿Cómo se preparó para ser empresario?
—Yo terminé Peritaje Mercantil y me matriculé en la carrera de
Profesorado Mercantil, pero nunca llegué a terminarla, porque la
empresa absorbió toda mi energía y mi tiempo desde los 19 años. Ya
de adolescente era un negociante nato y ayudaba a mi padre y a mis
tíos en verano. Nunca me he considerado muy inteligente. De hecho,
era un mal estudiante, pero tengo otras cualidades que me han servido
de mucho en la empresa y en la vida: sentido común, humildad,
responsabilidad, perseverancia, ambición sana, intuición, ilusión,
mucha energía, mucha capacidad de trabajo y de sufrimiento, con
Miguel V. Carreras. De vocación, empresario37
entrega y disciplina, no me cansaba nunca. Tengo mucha imaginación
y soy muy soñador, y no le tengo miedo al riesgo. Pero también soy
austero, inconformista y perfeccionista. Todas las noches cuando me
acostaba hacía una autocrítica del día. Yo sabía los errores de cada
éxito aparente y eso, lo rumias, te da experiencia y una cultura que te
hace tropezar menos. Pues con todo ello pude prosperar.
La responsabilidad y la fuerza de voluntad
Un ensayo titulado «Un mensaje a García», escrito por Elbert
Hubbard en 1899, narra la historia —con base real— de cómo a
finales del siglo XIX, el presidente estadounidense necesitaba
urgentemente contactar al líder rebelde cubano, el general Calixto
García, para coordinar esfuerzos durante la guerra contra España.
El presidente confió esta tarea a un joven teniente que aceptó la
misión sin hacer preguntas y sin demoras. Como decía mi padre,
«sin excusa ni pretexto». En lugar de cuestionar las dificultades de
la tarea, se dirigió a Cuba con el único objetivo de cumplirla.
Su travesía estuvo llena de obstáculos: terrenos hostiles, peligros
y barreras. Sin embargo, ni las adversidades ni las circunstancias
cambiaron su compromiso. A pesar de todos los desafíos, logró
entregar el mensaje a García y, en esencia, cumplió su deber. Esta
historia aporta lecciones profundas que me han guiado a lo largo
de mi vida:
Responsabilidad inquebrantable: asumir la tarea sin cuestionarla
ni buscar excusas. Esta actitud de tomar responsabilidad y actuar
con determinación es una cualidad que trasciende el tiempo y el
contexto.
Eficiencia y determinación: una vez que se acepta una tarea,
no importa cuán desafiante sea, es posible llevarla a cabo con
determinación y eficiencia.
Actitud positiva ante los desafíos: en lugar de detenerse en las
dificultades, hay que centrarse en encontrar soluciones y avanzar.
Lealtad al deber: la lealtad a un compromiso es una virtud
fundamental que puede inspirar y motivar a otros.
Ana Rioja Jiménez38
—¿El sentido común?
—Mi padre tenía los estudios básicos, a los 14 años ya salió a
trabajar, pero tenía mucho sentido común. Y yo he aprendido de él y
se lo transmito a mis hijos y a los empleados, les ayudo a discurrir. Mi
mujer se cree que soy muy listo por cómo aconsejo a los hijos, por el
éxito de las empresas, por los premios que me han dado, pero lo que
tengo es mucho sentido común, analizo los pros y los contras de todo.
Antes también, pero ahora más, veo que hay muy poco sentido común.
La logística, de hecho, es sentido común. La experiencia y el tiempo
lo aumentan, y te vas dando cuenta de las necesidades, las carencias,
de los problemas que tiene el cliente, los estudias, te pones en su lugar
porque estás más preparado que él, y resuelves el día a día, lo que se puede
mejorar. Si a ello le añades más imaginación y más ganas, lo completas.
—El factor humano, el personal, ¿es el verdadero patrimonio de la
empresa?
—Si los clientes son la razón de ser de la empresa, el principal objetivo,
el verdadero capital de la empresa es el personal, no es el dinero. Y
ese personal tiene que estar motivado, tiene que saber por qué trabaja,
para qué trabaja y cuál es el proyecto para el que está colaborando.
El personal tiene su corazón y hay que tratarle con humanidad, con
familiaridad, con amistad y con cariño, y eso se siente y se transmite
con hechos. El amor al personal te lleva a tener verdaderas amistades, a
Superación de barreras: la perseverancia puede superar barreras y
lograr resultados sorprendentes.
Esta lección trasciende las fronteras y se aplica en todas las
áreas de la vida, recordándonos que lo más importante es la
fuerza de voluntad necesaria para llevar a cabo nuestras tareas,
independientemente de las dificultades que puedan surgir.
Miguel V. Carreras. De vocación, empresario39
conservarlas en su jubilación, y te reúnes con ellos en casa, los abrazas
y los sigues apoyando mientras viven. Ellos son lo más valioso de la
empresa. Hay que cuidar los pequeños detalles, valorar lo más simple
e insignificante, eso me ha llevado siempre al corazón de las personas.
—Respecto al personal, es un hecho que cada vez se incorporan más
trabajadoras a su empresa.
—Sí, porque la mujer ha ido conquistando su puesto en el mercado
laboral, con los mismos derechos y obligaciones que los hombres. Son muy
Miguel Carreras en su despacho, en las oficinas de la Ciudad del Transporte (1990).
Ana Rioja Jiménez40
eficientes, fieles, trabajadoras y responsables, y dan un toque muy humano
a las empresas. En algún departamento, como el de Administración,
superan en número a los hombres. En Carreras intentamos que las
mujeres que son madres puedan conciliar su trabajo con la maternidad y,
cuando surge algún problema, intentamos auxiliarlas y buscar soluciones
de forma que puedan seguir trabajando, que no se vean obligadas a dejar
su puesto de trabajo. Tenemos, además, la Fundación, que se acerca a
ellas para paliar las carencias que puedan tener y que la Fundación pueda
resolver. Como Carreras es una empresa familiar, el trato es familiar hacia
la mujer. Y velamos por las familias de los empleados y, sobre todo, por
las de aquellas o aquellos que se encuentran en dificultades. Hacemos lo
posible por el bienestar, no solo del empleado, sino de su familia.
—¿Y la importancia de crear un buen equipo directivo?
—Las ideas pueden ser individuales, pero el desarrollo de las
mismas es un trabajo en equipo. Por eso es tan importante rodearse
de un buen equipo para trabajar junto a él con respeto, lealtad y
disfrutando con su trabajo bien hecho. Formar al personal es muy
importante. Los primeros colaboradores los hice yo, nunca se los
quité a ninguna otra empresa ni fui a buscarlos fuera. No podía pagar
más que el otro y no me gustaba hacerlo, porque el personal de la
competencia tiene sus cosas buenas, pero también una forma de hacer
que a lo mejor no coincidía con mis criterios.
Estuve 25 años viajando constantemente para abrir las 18
primeras delegaciones y a todos los delegados los escogí y formé yo.
Pasaron muchos hasta que tuve el equipo más adecuado. Quería
que trabajaran a mi forma, hacerlos a mi estilo. Por ejemplo, tuve a
cuatro sacerdotes que, por temas de celibato se salieron de curas, eran
buenísimas personas que no tenían ni idea de transporte. Los formé,
como a todos los responsables de delegaciones, y resulta que fueron
de los mejores empleados que he tenido: los más fieles, responsables,
y todos se han jubilado en casa. Uno de ellos me decía: «mira que yo
solo he tenido dos jefes: a Dios y a ti».
Miguel V. Carreras. De vocación, empresario41
Miguel Carreras se ha rodeado siempre de un gran equipo directivo,
entre ellos su hermano Jesús y sus hijos (en la primera foto, José Luis, de pie), 1990.
Ana Rioja Jiménez42
Todos esos primeros delegados se jubilaron en la empresa y son
amigos, aún hoy nos vemos y hacemos viajes juntos. Si se forman
en la empresa, aprenden a quererla como si fuera suya. Cuando
terminábamos un proyecto con éxito, lo celebrábamos, nos
sentíamos más unidos, aumentaba la autoestima para afrontar con
mayor seguridad el siguiente, que era más difícil. Hice una buena
selección de delegados y directores comerciales en 25 años, y luego
3 o 4 años más hasta que se asentó este personal. Uno me dice que
cuando oye Carreras se gira, y se apellida García. Otro me cuenta
que aún sueña con los camiones y que a sus 92 años, cuando ve un
camión de Carreras, se le ensancha el corazón. También me produce
gran satisfacción dar oportunidades a nuestros jóvenes talentos para
evitar que tengan que emigrar. Hay que velar por la gente joven con
preparación, con talento, con ilusiones, para que desarrollen todos
esos valores con nosotros. Tienen los medios para hacer una obra
importante, como es su aportación a la empresa. Hay que hacer que
la desarrollen, dándoles poco a poco más responsabilidad. Y con esa
ambición suya y con la nuestra, luego resultan ser grandes jefes. Esa
es nuestra ilusión y casi siempre lo conseguimos: que los mandos sea
gente formada en casa e iniciada de joven.
«Pensar en vaca»: La empatía
Otra metáfora útil es lo que llamo «Pensar en vaca». Dicen que en
un pueblo se perdió una vaca y se pusieron todos a buscarla. Nadie
la encontraba. Pero en estas, que un discapacitado del pueblo dijo:
«yo, si fuera vaca, iría a un sitio donde haya agua, hierba fresca y
sombra». Se pusieron a pensar dónde podía haber un sitio así. Lo
encontraron. Y, efectivamente, allí estaba la vaca.
Moraleja: hay que tener empatía, ponerse en lugar de los demás,
de los colaboradores o clientes. Y también, no hay que despreciar
ninguna opinión, hay que escuchar a todo el mundo. Esto es
«Pensar en vaca».
Miguel V. Carreras. De vocación, empresario43
La jubilación de los trabajadores de la empresa también era un momento
de celebración, de compartir. Y siempre, junto a Miguel Carreras, su hermano Jesús.
Los empleados de las empresas de Miguel Carreras son su segunda familia. Le encanta
estar siempre cerca de ellos. En Eizasa, celebrando la Navidad de 2014.
Ana Rioja Jiménez44
—¿Es importante también para un empresario ser un buen comunicador?
—Es primordial. Hay que comunicarse mucho con el personal:
valorarlos, involucrarlos en tu proyecto, animarlos, ha sido una
constante en mi trabajo. Tengo la sensibilidad de acercarme al personal
y de acercarlos. En eso también he tenido suerte. Tengo a mi gente
trabajando con ilusión y feliz. Importantes empleados y delegados
han tenido propuestas para irse ganando más, y no se han querido
marchar de esta empresa. A todos iban a tentar, pero ninguno se fue.
Cuando tienes éxito, la gente busca tu personal.
Hay gente que quiere conocerme, porque no intimido. Me tienen
excesivo respeto, pero por cariño. Me ven ya mayor, pero que no
disminuye mi interés por ellos y por las empresas. En mi cumpleaños,
las delegaciones me hacen fotos, vídeos, y los de Zaragoza me esperan
y me hacen hablar. El otro día estuve 17 minutos hablándoles. A la
gente le gusta que les hable y a mí me encanta hacerlo. Nos queremos.
La gente me acepta como soy, con mi humildad, sinceridad y nobleza,
siempre de cara con la verdad, pero también con mis carencias. A veces
saco mi genio fuerte en errores repetidos, aunque no soy rencoroso
y se me olvida inmediatamente, y trato de que aflore rápidamente el
cariño en un abrazo sincero.
—¿Hay que fomentar también la cercanía y la amistad con el cliente?
—La amistad es un don que nace de una empatía, de una
comunicación que implica al otro y a ti. Surge del contacto, de
la relación entre las personas. La amistad con el cliente es muy
gratificante. Hay dos fundamentos: primero, al cliente hay que
tratarlo y servirle bien; y segundo, si mi cliente está contento con mi
servicio, hay que tratar de reforzar esa relación. A mí me interesa esa
amistad, porque siempre he pensado que si el cliente es mi amigo,
me obliga más, todavía tengo más celo por cumplirle, le serviré mejor
sin darme cuenta, porque además de cliente, lo que implica que es
el objeto de mi trabajo, es amigo. No hay que pensar que, porque es
Miguel V. Carreras. De vocación, empresario45
amigo, le voy a servir peor o más tarde, eso es un error. Mi trato
y mi compromiso con los clientes amigos es superior, porque me
comprometo con un amigo, no solamente con un cliente. Además,
egoístamente, si el cliente es amigo, me fallará menos. Con el cliente,
cuanta más confianza e intimidad, mejor, aumenta la fidelización
Equipo directivo de Carreras Grupo Logístico en la convención celebrada
en el Hotel de Badaguás, a principios de 2020.
Miguel Carreras
siempre ha procurado
acercarse a sus trabajadores
y acercarlos.
Equipo de la empresa,
que jugaba en la liga
laboral hace unos años.
Ana Rioja Jiménez46
recíproca, más te obliga y más te puede perdonar un error. Y lo mismo
con el proveedor. Lo más hermoso es la amistad y para ello hay que
comunicarse con el otro, escucharlo, preocuparse por él, pero esto en
todos los órdenes de la vida.
—¿La amistad implica también generosidad?
—Los seres humanos somos cuerpo y alma, por lo que debe atenderse
también esa parte. Es responsabilidad de la empresa y del empresario
atender en este sentido a sus empleados, si tienen depresión, ansiedad,
problemas familiares… Hay que atender esas necesidades también.
Hay que ser generoso, fomentar la amistad y el amor al personal, ver la
valía de los demás y saber extraer lo mejor de cada uno. Debemos ser
mejores personas para ser mejores profesionales.
Intento ser generoso y soy extrovertido. Me gustan mis amigos,
mi gente, no soy parco en alabanzas, me gusta felicitar al personal
cuando se lo merece, dar las gracias a los míos, ensalzar sus cualidades,
muchos méritos dárselos a ellos. Y, ante la duda, favorecer al más
débil. La Fundación es un complemento para ejercitar la generosidad
y ayudar al que lo necesita.
—¿Y la responsabilidad?
—Yo les digo a mis hijos que la mejor motivación para los
trabajadores es el ejemplo y, no hacer las cosas a tontas ni a locas. Por
ejemplo, pido préstamos, pero los estudio: dónde los invierto, veo su
rentabilidad y hasta dónde puedo llegar. Me puedo equivocar, pero
también tengo un remanente por si algo falla. Lo tengo calculado, por
eso duermo. Porque sé que la exigencia que voy a tener de devolver
el dinero no es hoy, que será con un calendario y lo tengo estudiado.
Y eso es muy importante. Siempre he sido muy responsable con
cada decisión, con cada nuevo proyecto y reto, porque son miles de
personas los que han trabajado conmigo, cerca de 3.000 ahora y unos
5.000 en total los que han pasado por las empresas, 5.000 familias
que dependían de la buena marcha de mis empresas y de las que me
Miguel V. Carreras. De vocación, empresario47
he sentido responsable. Si hablamos de éxito, ese es mi mayor éxito,
mi mayor orgullo, haber creado esos puestos de trabajo. No el de
tener todas las empresas que tenemos.
—La imaginación también ha sido muy importante en su vida. ¿Es
un soñador?
—Yo tengo mucha imaginación y soy un soñador. Estuve 25 años
creando las 18 delegaciones que fundé en España y alguna en Europa.
Durante el día iba con los delegados y los comerciales haciendo
visitas a los clientes, y por la noche me trasladaba, a veces ni me
acostaba. Dormía poco. Como viajaba muchas horas en el coche,
con la imaginación hacía de todo, unas veces tenía una granja, otras
unos campos con riego, otras hoteles, sueños que se convirtieron en
proyectos y que posteriormente iba desarrollando.
Muchos de los proyectos que he hecho los he soñado y, cuando
en la realidad he llegado a ese lugar, ya lo tenía visto, no me era
desconocido, lo había visto con la imaginación, y lo buscaba y era tal
y como lo ha había imaginado. Eso me ha ocurrido muchas veces y
me ha facilitado el trabajo.
La imaginación: el guía de la caravana
Siempre me he sentido como el guía de las caravanas del Oeste
americano. Cuando tenían que ir a colonizar tierras, a buscar buenos
asentamientos y terrenos para tener una vida mejor y construir sus
casas y sus granjas, siempre había un guía que iba delante con su
caballo. Y por las noches, cuando todos acampaban y descansaban,
él montaba su caballo y se iba a inspeccionar el terreno para saber
si había agua, montañas, campamentos de indios, buenos caminos…
Y a la mañana volvía y les decía: el agua racionada, porque en varias
millas no había encontrado agua, y les indicaba el mejor camino,
el que tenía menos peligro, por dónde avanzar, dónde acampar
por la noche, cuál era el lugar más seguro. El terreno para él ya era
conocido. Pues eso me pasaba a mí con mi imaginación.
Ana Rioja Jiménez48
—Usted predica la austeridad.
—Un día me dijeron, pero oiga, es que usted tiene mucho dinero.
Si llevo 70 años trabajando, si soy austero en mi vida —no tengo yates
ni una gran mansión, ni me hacen falta— y con tantos años de trabajo,
por fuerza tengo que tener dinero. Si no, sería un idiota. Mejor dicho,
el dinero es de la empresa, que es la que lo necesita para expandirse.
Hay que dar ejemplo, ser austero y demostrarlo. Esa austeridad
personal me ha llevado a ser generoso con la empresa, porque es como
un ser vivo que necesita ser amado, alimentado y protegido, pero al
mismo tiempo necesita continuamente de inversiones y tecnología
para crecer, estar en el mercado y ser competitiva.
—¿Y el riesgo?
—Un empresario no solo tiene que invertir lo que gana y lo que tiene,
eso es muy cómodo, un empresario tiene que invertir el crédito que la
sociedad le da. En este caso, los bancos te dan unas garantías para dejarte
dinero, esas garantías no se pueden dejar ahí sin usarlas y el buen empre-
sario debe hacer uso de los beneficios que tiene y de esas garantías que la
sociedad le reconoce. Es incómodo, es un riesgo, pero bueno también se
habitúa uno a vivir con riesgo y ya es menos riesgo. Mi padre me decía:
«pero hombre, ¿cómo debes tanto, cómo puedes dormir?», y eso que él
no sabía muy bien lo que debía; y yo dormía con riesgos tremendos y te
vas familiarizando con ellos y te van saliendo bien. Yo cuando no dormía
era cuando no debía. Y entonces me preguntaba: «¿pero cómo no tengo
alguna idea y hago un proyecto nuevo para esa parte de garantía que no
utilizo?». Entonces, la utilizaba y la rentabilizaba, porque la obligación de
un empresario es rentabilizar todo eso. Ese egoísmo y falta de ambición,
de ideas y de dirección de empresas es no utilizar esa garantía que te con-
ceden. Si no, se va muy despacio.
—¿Cómo tiene que ser la relación con los bancos?
—Los bancos son necesarios y hay que hacer uso del crédito que
dan, de la garantía que otorgan en orden a la confianza que nos ofrecen.
Miguel V. Carreras. De vocación, empresario49
Miguel Carreras y su Grupo siempre han colaborado con la vida social
y cultural de Aragón con diversos patrocinios, siempre muy interesado por el deporte.
Ana Rioja Jiménez50
Para eso tenemos que cuidarlos. No se puede engañar a un banco, ni
defraudarlo, ni quedar mal con él, porque nos estaremos cerrando el
futuro. Es preferible vivir con lo básico, sin lujos ni dispendios, a dejar
de pagar a un banco. Lo importante es cumplir con el banco y habrá
futuro. De lo contrario, se cierra el futuro financiero y el de la empresa.
—La ambición sana.
—Hay que ser valiente, ambicioso y a la vez prudente. Nunca me
he conformado con cinco, y me preguntaba: «¿pero lo he hecho bien?
¿No podría producir siete u ocho?». La empresa debe ser ambiciosa
con nuevos proyectos, perfeccionar los actuales y saber diversificar
para crear oportunidades y seguridad.
Era ambicioso porque quería hacer una empresa fuerte, por eso
arriesgaba, por eso luchaba, sacaba fuerza suficiente para vencer los
obstáculos y para acometer proyectos, y para ello hay que tener fuerza,
energía mental. Para sacar adelante un proyecto tienes que verte
campeón y decir: «voy». Para mí, el trabajo no era trabajar. No me
hacía mella. Y salud, sobre todo salud, ambición, ilusión y confianza
en ti mismo, y con todo eso, adelante.
—Y la autoestima, no tener miedo al fracaso.
—Yo he arriesgado mucho. Todo lo que ganaba lo arriesgaba.
Cada vez tenía más garantías, pero las ponía todas. Y además con
alegría, sin temor, con confianza, a lo mejor, excesiva. Algunas veces
arriesgaba todo, porque los bancos me hacían firmar todo y avalar,
pero estaba seguro de lo que hacía.
Todo ello unido a que hay que ser optimista y tener autoestima, no
tener miedo. Cuando se estudia un proyecto y se va a por él, hay que
ser optimista, hay que verse ganador. Los chinos dicen que el primer
paso es medio camino. Hay mucha gente que nunca da el primer
paso. Hay que ser valiente y asumir riesgos para liderar una empresa.
Los grandes emprendedores ven el cambio como una oportunidad,
no como una amenaza.
Miguel V. Carreras. De vocación, empresario51
Sin riesgo no se gana dinero. Nunca pensaba en el fracaso. Nadie
puede ser campeón de nada si no se ve en el pódium, pues yo me veía
ganador y eso me daba confianza. Eso me ha valido para el desarrollo
de las empresas. He sido valiente, he tenido ilusión, he sido riguroso.
He valorado los pros y los contras con el sentido común, no he tenido
temor, y siempre he pensado que iba a ganar, tenía mucha confianza
y fe en mí y en mi equipo humano. Nunca he tenido miedo a fracasar
y sabía que si llegaba el caso, me levantaría.
—Pero el fracaso también existe. Hábleme de ello.
—También he tenido algún que otro fracaso, pero hay que saber
reponerse de él y aprender de los errores. En los años 70 empezamos
a abrir delegaciones en Inglaterra, Alemania, Polonia, Hungría, Italia,
en casi toda Europa, pensando que trasladaríamos allí el modelo de
negocio que tan bien funcionaba en España. Tuvimos que ir cerrando
muchas de ellas y nos quedamos con Francia, Italia y Portugal, que son
un éxito. ¿Por qué? No supimos gestionarlas como en España. Cada
país tiene su idiosincrasia, no estábamos preparados y fracasamos. No se
podía exportar tal cual el modelo de España porque cada país tiene sus
peculiaridades. Ahora volvemos ya, pero con otra preparación, con otra
experiencia y con otros medios. En Francia estamos con el transporte,
la logística y la distribución capilar, como hemos hecho en Portugal.
También operamos con Alemania e Italia en transporte general.
Otro fracaso fue que hace 20 años, con la ansiedad y la inquietud
que me caracteriza, creé una empresa de forwarder (un agente que
realiza operaciones de comercio internacional y que se ocupa de
gestionar el envío de las mercancías a nivel global por tierra, mar y
aire). Abrimos las oficinas en el Edificio Colón de Barcelona, con
más de veinte empleados y, a los tres años de perder dinero, la cerré.
Ahora hemos vuelto y tenemos otra, porque la preparación es distinta
y nuestra imagen también. Tampoco estábamos preparados. Nos
adelantamos veinte años.
Ana Rioja Jiménez52
Otros proyectos que comenzamos y no llevamos adelante fueron los
transportes especiales para cargas y dimensiones fuera de lo normal. Lo
tuve que dejar porque no teníamos los medios suficientes, había que
alquilarlos y era muy costoso. Además, había una empresa especializada
que hacía ese tipo de transporte y nos ganaba en infraestructura. No era
nuestro fuerte. Continuamos con las delegaciones, el transporte general
y luego la logística. Todo no se puede hacer bien.
También me agrupé con un amigo francés y con otro alemán, y
pusimos en marcha una línea directa con París y Berlín, que tampoco
prosperó. Destinamos camiones fijos para esa línea, una paquetería
directa. Pero no hubo suficientes clientes, los camiones iban a medio
cargar y tuvimos que abandonar el proyecto.
—Fortaleza y coraje para sobreponerse a los fracasos.
—Y luego hay que tener fortaleza y coraje para levantarse de los
errores y fracasos. No todo es un camino de rosas. Pero se vuelve a
empezar con más fuerza, las veces que sean necesarias si el proyecto
merece la pena y te ves capacitado para ello. El dinero es muy
conservador, es muy temeroso, pero nunca he tenido apego al dinero.
Ha influido también que he tenido mucha vitalidad, mucha energía y
salud. Eso ha contribuido mucho. Primero a hacer frente al proyecto,
a elaborarlo y a consolidarlo.
—No defraudar ni engañar a nadie.
—Toda mi vida he seguido el consejo de mi padre: no defraudar
a quienes confían en ti, ni engañar a nadie. Cuando fui cogiendo
personal, aprendí a quererlos, a mirar por ellos. Como había estado tres
años de chófer tenía espíritu obrero, sabía lo que gusta a los obreros, lo
que no gusta, cómo viven, cómo piensan y eso también me favoreció
mucho, me educó con humildad y no solamente de empresario, sino
como obrero. Esos tres años seguidos en un camión fueron importantes,
noche y día conduciendo y durmiendo allí, pero era joven y con ilusión.
Iba conociendo España y me encantaba. Fui muy feliz.
Miguel V. Carreras. De vocación, empresario53
Un gran hito para Miguel Carreras fue abrir delegaciones por toda España y crear la red
comercial que sentó las bases para ocupar la posición que tiene ahora en el sector.
Un fracaso. Miguel Carreras cuenta cómo tuvo que abandonar los transportes especiales
porque no tenían la infraestructura suficiente.
Ana Rioja Jiménez54
La intuición y el subconsciente
Otra virtud que tengo es que trabajo con el subconsciente. Todas
las soluciones a los problemas, todas las ideas de proyectos
nuevos han llegado siempre por la noche, en la cama, a las 4
o las 5 de la mañana, con la mente lúcida de la madrugada,
en silencio. Me las anotaba o me las grababa. Luego supe que
eso era trabajar con el subconsciente. El consciente no me daba
soluciones, pero sin querer, el subconsciente me despertaba y
empezaba a lanzarme el mensaje.
Y eso me ha ayudado mucho. Eran problemas de trabajo, a los
que no encontraba solución durante el día y luego, por la noche,
me venía sola, sin buscarla. Eso es muy importante. De hecho,
durante un tiempo estuve trabajando este tema con Antonio
Blay, un célebre psicólogo, para desarrollar ese potencial, para
aprender a consultar al subconsciente y que te conteste. Trabajar
con el subconsciente me dio serenidad y calma. Los seres
humanos somos energía y amor. No hay que dejar que la tristeza
o las desgracias las anulen. A mis hijos intento enseñarles esto,
pero no es fácil.
El consciente es la memoria, lo que sabes. El subconsciente es
aquello que no recuerdas, que no has retenido en el consciente,
pero has visto y oído. Y graba todo sin que tú te enteres, es un
ordenador que ante un problema, estudia todas las posibilidades,
pros y contras, y encuentra la mejor solución, porque él conoce
todo. Con esa información te da un resultado. El subconsciente
es el que te ayuda a tomar decisiones, sacando las respuestas
que tenemos dentro de nosotros y no queremos o podemos ver.
Según los expertos, el proceso cognitivo intuitivo se nutre de
información a través del subconsciente. La mente se abastece
de toda la información que se acumula en nuestro cerebro y
de la que no somos conscientes. Cuando llega la noche es
cuando actúa el subconsciente, de ahí la frase: «Consúltalo con
la almohada».
Miguel V. Carreras. De vocación, empresario55
—¿Cómo es su relación con la competencia?
—La competencia es necesaria. Gracias a ella hemos avanzado
en tecnología, en la ciencia… Si no hubiera habido competencia
nos hubiéramos relajado y no hubiera nacido en nosotros ese
espíritu de lucha. La competencia nos hace despertar lo que
llevamos dentro, nace un deseo de ganar, de competir con los
demás. Y para ganar, invento y realizo cosas, hago sacrificios, me
esfuerzo por estudiar y por hacerlo mejor. Si la competencia no
existiera, la civilización hubiera avanzado menos. La mejora viene
por la competencia, es la que nos hace superarnos y tener éxito. La
competencia es lícita, se puede competir en hacerlo mejor y, como
somos profesionales, hay que llevarse bien con la competencia,
nos respetamos y somos amigos, y luego, vamos a ver quién lo
hace mejor, pero eso no quiere decir que seamos enemigos. Es un
profesional y un colega que está en la misma profesión o sector que
yo y ambos disfrutamos de los avances, de las herramientas que
los técnicos nos preparan. Estamos evolucionando en la misma
dirección, uno lo hará mejor que otro o más deprisa, pero estamos
en lo mismo. La competencia es sana y hay que ser amigo de tus
competidores, te hace ser mejor.
—La importancia del marketing.
—La buena ropa no se vende en el baúl, como antes. Hay que
hacer un buen producto, pero hay que darlo a conocer, hay que
explicarlo, tienes que venderlo, que conozcan sus cualidades, su
eficacia, su precio, y comunicarlo como que es lo mejor que hay
en el mercado para que la gente lo conozca y crear la inquietud y
la necesidad de compra. Hay que despertar el interés del cliente
y del consumidor. Y para ello el marketing se sirve también de la
comunicación y de la publicidad. Hay que demostrar la bondad del
producto, darlo a conocer y ganarte la confianza del que te escucha,
que te crea y confíe en ti. Ganarte, en suma, la confianza del cliente
o consumidor.
Ana Rioja Jiménez56
—Respeto con el entorno y el medio ambiente. Sostenibilidad.
—Está en nuestro ADN: somos aragoneses, cultivamos la tierra,
tenemos aerogeneradores y, en todas nuestras naves, ponemos
placas solares. Adquirimos los camiones de última generación
para contaminar menos. Estamos reduciendo paulatinamente las
emisiones de dióxido de carbono. En Cataluña vamos a producir
hidrógeno, que consumirán los camiones y las carretillas de allí.
También tenemos vehículos eléctricos y con gas. Y nuestra intención
es llegar a no emitir nada de CO 2 en unos años. Apostamos por la
sostenibilidad.
—Y además de todo ello hay que tener suerte. ¿Se considera un hombre
con suerte?
—La suerte existe y hay que creer en ella. Siempre me he
considerado un hombre muy afortunado en todos los aspectos de
la vida. Hay gente que ha nacido con mala estrella y otros, como
yo, con buena estrella. Compraba una cosa y al poco tiempo se
revalorizaba, y todo el mundo pensaba que era muy inteligente,
pero no soy listo ni era un visionario, he tenido mucha suerte,
sobre todo con mis colaboradores, son ellos los que han sacado los
proyectos adelante. La suerte tiene sus reglas, llega cuando menos
te lo esperas, pero siempre, siempre, se obtiene con el esfuerzo.
¿La suerte en la empresa es tener éxito? Puede ser. Pero ya digo que
el éxito en las empresas viene de un cúmulo de muchas cosas que
tienes que manejar.
A veces pienso que no me merezco lo que tengo, que tengo más
de lo que me corresponde. Otros han trabajado mucho y no tienen
nada. Casi todos los proyectos que he empezado han prosperado. De
ahí la necesidad de crear una Fundación y de ser solidario, porque
la empresa también tiene una función social en beneficio del bien
común.
Miguel V. Carreras. De vocación, empresario57
La familia ha sido siempre el gran pilar de la vida de Miguel Carreras.
En esta imagen, con su esposa y sus siete hijos.
Los padres de Miguel Carreras vieron los éxitos de su hijo y estaban muy orgullosos de él.
Su padre, Jesús Carreras Pomar, siempre lo apoyó y estuvo a su lado.
Ana Rioja Jiménez58
Veinte pilares para la casa de un empresario
1. Vocación (dedicación y entrega constante).
2. La familia. Unión y apoyo familiar.
3. El sentido común.
4. El verdadero capital de la empresa es el factor humano.
5. Generosidad, amistad y cuidado del personal.
6. Los objetivos deben ser los proyectos, nunca el dinero.
7. Austeridad.
8. Reinversión.
9. Responsabilidad.
10. Ambición sana.
11. Humildad.
12. No defraudar a quien confía en ti.
13. Implantar una ética en la empresa.
14. Confianza y autoestima.
15. Dar ejemplo: siempre con la verdad, nunca con la mentira.
16. Honestidad y admitir los errores.
17. Suerte e intuición.
18. Ser valiente, arriesgado y, a la vez, prudente.
19. Fortaleza y salud.
20. Respeto con el entorno y el medio ambiente.

Miguel V. Carreras. De vocación, empresario59
CAPÍTULO III
Nace la Sociedad Carreras
Los orígenes. La primera generación
Miguel Carreras ha heredado de su padre el sentido común, su carácter
austero y la ilusión por tener algo propio, creado por él y verlo crecer:
una empresa que va a más, que hay que cuidar y seguir muy de cerca.
El amor por el trabajo, en suma. De su madre, la simpatía, su don
de gentes, esa sonrisa que no escatima y que ofrece a su interlocutor
desde el primer momento, cuando estrecha su mano. Y de los dos
aprendió a ser humilde, honrado, responsable, respetuoso y leal. Su
padre nació en Peñalba y su madre en La Almolda, localidad natal
también de Miguel Carreras, ambas en Los Monegros, una comarca,
un paisaje aragonés que imprime carácter y que forjó su personalidad.
—Hábleme de su padre y de su familia paterna.
—Mis abuelos paternos eran de Peñalba. Mi abuelo estudió en el
seminario, pero se salió antes de cantar misa y trabajó como guarda
forestal. También era juez de paz del pueblo. Además tenían un bar
de comidas y bebidas que lo regentaba con mi abuela. Tuvo siete
hijos: seis chicos y una chica.
El mayor hizo carrera militar y mi padre, Jesús Carreras Pomar,
que era el sexto, se fue voluntario a la mili y se quedó un tiempo en
el Ejército. Llegó a ser sargento. Se salió para formar una sociedad
con otros dos hermanos porque le gustaba más la empresa que el
Ejército, aunque le quedó una impronta de esta formación, ya que
era muy disciplinado y muy serio. Mandaba comunicados, cuando ya
estábamos en la empresa, que decían: «sin excusa ni pretexto. Esto es
de obligado cumplimiento». Son términos militares.
Ana Rioja Jiménez60
Los padres de Miguel Carreras con tres de sus hijos.
Aún nacería después un cuarto.
Miguel Carreras con su esposa, sus padres, hermanos y cuñadas en los años 90.
Miguel V. Carreras. De vocación, empresario61
—Y en 1933 funda Carreras junto a dos de sus hermanos.
—En 1933, junto a dos de sus hermanos: Antonio y Teodoro, fundan
Carreras, una pequeña empresa de transportes, que era una sociedad con
un documento privado. Compraron un camión viejo, un Bedford de
segunda mano y de tres toneladas, y comenzaron a transportar cereales y
paja. Por aquel entonces, mi padre se casó con mi madre, una muchacha
de La Almolda, y vivía allí, aunque el negocio lo tenía en Peñalba.
—¿Por qué le llamaban a esa sociedad La Canadiense?
Camiones adquiridos
por la Sociedad Carreras a
partir de 1939:
una camioneta y un
camión FIAT, utilizados
durante la guerra civil.
Detalle de un sobre
de los años 40, en el que
se leen los servicios de
transporte que ofrecía Jesús
Carreras Pomar.
Ana Rioja Jiménez62
—Uno de los transportes que hacía el camión era a Mequinenza,
localidad en la que estaba la empresa minera La Canadiense.
Transportaban paja y cebada para las mulas, porque había por lo
menos doscientas y consumían mucho, claro. Entonces los del pueblo
comenzaron a decir que los Carreras, en plan de chunga, habían
formado La Canadiense. Aunque viejo, pero los Carreras tenían un
señor camión para la época. También trasportaban trigo para las
fábricas de harina.
—Llegó la guerra civil y les requisaron el camión.
—La guerra interrumpió esa actividad y truncó durante tres años
sus vidas, como las de todos. Les confiscaron el camión. Nada más
terminar la guerra reanudaron la actividad de transporte.
—¿Cómo recuerda a su padre con sus ojos de niño?
—Mi padre era serio, humilde, muy trabajador, inteligente,
responsable y muy presumido, le llamaban «el mediquer» porque
era tan pincho como el médico y llevaba sombrero y todo. Era muy
honrado y justo. Por su conducta y su humanidad se salvó de morir en
la guerra civil. Cuando entraron los rojos en La Almolda, obligaron
a todos los del pueblo que tenían escopetas de caza a entrar en la
ermita de san Antonio y a entregar las armas. No los dejaron salir.
Los encerraron allí y más tarde los bajaron en un camión a Bujaraloz
y los fusilaron a todos. Al salir mi padre de la ermita, uno de los
soldados que los vigilaba, dijo: «no, a este no, que es buena persona».
El soldado había sido peón de mi padre, era el único que le había
dado un trabajo porque el hombre llevaba fama de rojo. Y le salvó la
vida. Mi padre era muy querido en La Almolda.
—¿Cómo era su madre?
—Mi madre se llamaba Carmen, era muy guapa y bondadosa,
muy entregada a sus hijos, era muy simpática y hacía amistades muy
pronto. Sus padres se dedicaban a la agricultura. En la época en la que la
Miguel V. Carreras. De vocación, empresario63
agricultura no necesitaba tanta dedicación, porque ya se había sembrado
y demás, su padre solía ir con el carro a traer vino y aceite, luego los
vendía por los pueblos y bajaba trigo. Hacía transporte con el carro.
Cuando la conoció mi padre, vivía con su madre, que era ya viuda.
Acababa de perder a su padre y a su hermano, que falleció muy joven,
con 21 años, iba de luto y apenas salió de casa durante tres años.
Eran tres hermanas y ese hermano que falleció. Mi padre se enamoró
de esa muchacha hermosa y triste. Cuando se casaron tenía 20 años,
mi padre 27. Al contraer matrimonio, mi padre subió a vivir a La
Almolda, pero trabajaba con sus hermanos en Peñalba y comenzaron
a transportar cereales, sobre todo trigo, cebada y paja. Tuvieron cuatro
hijos. Yo soy el mayor.
—¿Cómo reanudaron la actividad tras la guerra? ¿Les devolvieron el
Bedford?
—No, de ese camión ya no supieron nada más. La empresa reanudó
su actividad, porque les dieron una camioneta FIAT, de la guerra, que
habían traído los italianos. A los dos o tres años compraron un camión
FIAT, de vagón más grande, también de la guerra. Y comenzaron a
hacer más negocio.
—¿Qué se transportaba entonces?
—Sobre todo albardín, esparto, cuya actividad fue muy importante
en estos pueblos durante los años cuarenta porque les supuso un
medio para ganarse la vida, porque servía para hacer papel. España
estaba bloqueada entonces. No había comercio con Europa y la pasta
la tenían que fabricar de albardín. Estuvieron 10 o 12 años con el
esparto. Contrataban los montes comunes donde había esparto y
entonces se ponían unos rastrillos y las mujeres peinaban el esparto, se
quitaba la cabeza, se quedaba limpio y se hacían pacas que se llevaban
en camiones a las fábricas de papel para hacer pasta de papel, sobre
todo a Tolosa (Guipúzcoa) y a Burriana (Castellón).
Ana Rioja Jiménez64
—¿Qué tareas desempeñaba su padre en la empresa?
—Era el administrador, el que llevaba la empresa. Mi padre recorría
los pueblos donde teníamos arrendados los campos del albardín y
organizaba todo. En La Almolda teníamos los rastrillos donde se
peinaba y empacaba el esparto, que daba trabajo a unas 30 mujeres.
Luego se reunía en Zaragoza con los socios del albardín y con otros
transportistas comisionistas que colocaban el albardín en las fábricas.
Y aún se compraba y se vendía cebada y trigo.
Mi padre en «La Canadiense» coordinaba todo. Solíamos llevar
esparto y traíamos patatas. Mi padre se encargaba de buscar a los
patateros, para que los camiones retornasen con mercancía, o bien de
Castellón o de Guipúzcoa. Transportaban a la vuelta patatas, cebollas
y demás. Se encargaba de la contabilidad, del trato con los clientes, de
hacer las facturas. La organización de esa sociedad la llevaba mi padre.
Cuando se abrieron las fronteras, la pasta, que ya era de madera,
venía de fuera y se quedaron con poco trabajo. Mi padre tomó la
decisión de venir a Zaragoza para hacer transporte desde la capital
aragonesa. Fue en 1951, cuando yo tenía 17 años.
—¿Qué recuerdos tiene de su infancia?
—Pasé la guerra en La Almolda con mi madre y mis hermanos.
Íbamos a la escuela y los recuerdos del pueblo son bonitos. Teníamos
unas tierras, pocas, y nos las cultivaban. En la época de la siega y la
trilla, íbamos los chicos al campo, primero a llevar el botijo, luego a
respigar, a coger las fajinas... En la era, a trillar con las mulas, luego
llegaron las trilladoras y las cosechadoras. En verano apoyábamos esa
agricultura. A mí ya me gustaba el campo y la agricultura.
También recuerdo cuando salí por primera vez del pueblo, a los 8
años, a casa de mi tío Manuel, un hermano de mi padre que vivía en
Manresa y era sastre. Y a Burgos, con mi tío Víctor, el militar, que era
brigada. Y a Barcelona con la hermana de mi padre, que se llamaba
Luisa, también pasé alguna temporada. A los 14 años vi por primera
Miguel V. Carreras. De vocación, empresario65
Carmen Calvete con sus cuatro hijos.
Miguel Carreras era el mayor.
Ana Rioja Jiménez66
vez el mar. Mis padres lo hacían para que saliera del pueblo, para que
viera otras ciudades. A los 10 años ya hice el ingreso en Bachiller y me
metieron interno en los Escolapios de Zaragoza.
—Los Monegros es una comarca dura, que imprime carácter. ¿Marcó
el suyo?
—Ya lo creo. En los pueblos de Los Monegros la vida es muy
dura. Es un territorio hermoso, pero árido y seco, donde se pasaban
tres y cuatro años sin coger cosechas y había que sobrevivir, donde
se valoraba tanto el agua, que no se podía desperdiciar ni una gota,
porque el agua que se bebía era la que se recogía en las balsas y a veces
ni había, y se tenía que ir a buscarla con mulas y una cuba de La
Almolda a Caspe, que hay treinta y tantos kilómetros.
En estos pueblos duros y en los fronterizos la gente es más viva,
más austera y solidaria, vive con menos, aprende a sacrificarse y es
más despierta. Algunas veces, si se quería comer, se tenía que ir a cazar
un conejo o a venderlo, por lo que había que aprender a cazar, y con
tirachinas se mataban pajaricos para comer. Si no se cogía cosecha un
año, se tenían que empeñar para poder sembrar, rezando para que se
pudiera coger al año siguiente. Y la consecuencia era que la gente se
marchaba fuera a trabajar. Nosotros con el esparto aguantamos unos
años. Los chicos éramos despejados. Eso forja un carácter especial de
fortaleza y sacrificio, no eres exigente ni con la bebida ni con la comida.
Sabías dónde dormían los pájaros, dónde estaban los conejos, poner
lazos para cazarlos, de quién era cada perro, cómo se cruzaban los
animales y cómo parían. Sabíamos mucho de la vida y de la naturaleza.
—¿Estuvo interno tres años en los Escolapios?
—En los Escolapios estuve interno hasta tercero de Bachiller,
desde los 10 a los 13 años, que cogí una pulmonía y ya no pude
continuar en el internado, perdí un trimestre. Cuando me repuse
comencé los estudios de Peritaje Mercantil, porque a mí me gustaban
los camiones y el negocio. En verano, a los 14 y 15 años, ya iba con
Miguel V. Carreras. De vocación, empresario67
una moto comprando por los pueblos cebada, trigo y pesando. Y a
regatear. Eso ya me gustaba. Quería una carrera corta porque a mí me
gustaba el negocio: comprar, vender, los camiones… Por eso estudié
Peritaje y Profesorado Mercantil.
—¿Cómo eran los internados entonces? ¿Fue también un aprendizaje
de vida?
—El internado fue también una escuela de vida. Los vigilados (o
externos) eran los de la capital y los internos, que éramos todos de
pueblos, les dábamos veinte vueltas y con vivencias extraordinarias en
contacto con la naturaleza y la tierra. Eso también es cultura y forma
tu carácter. La educación que recibíamos los internos era muy estricta,
tenías que obedecer y aguantar lo que te dijeran los curas, no podías
protestar, comerte lo que te daban sin malas caras, y a la fuerza, sin
poder elegir otra cosa. Solo podías ir a casa en Navidad, Semana Santa y
en las vacaciones de verano. De ahí no se movía nadie. Alguna vez venía
a verte un familiar, bajabas a las sala de visitas, los veías media hora y
adiós. Los domingos nos dejaban salir un rato, con los sacerdotes, de
dos en dos y en fila. Pero te acostumbras a eso y ahí está la diferencia
con otras personas que no han pasado por ello. Al principio es un
sacrificio, luego no. Te acostumbras a las contrariedades y ya todo en
la vida es más llevadero. Aprovechas mejor las oportunidades cuando
salen, porque situaciones así te hacen ser muy espabilado, agudizar el
ingenio para ir sobreviviendo. El que estaba externo tenía en casa todo,
aquí había que aguantar a los curas y a los compañeros. Me debía portar
bien y ser bueno, porque querían que fuera Escolapio. «Valentinico
—me decía un sacerdote— tú serías un buen cura, ya te llevaremos al
seminario». Pero no, yo no tenía esa vocación.
No podías salir, aunque pasase algo en la familia, y eso te va cociendo
y haciendo fuerte. Yo estaba deseando que llegara el verano, porque traía
mi libertad, iba al campo y hacía los trabajos acordes a mi edad: llevar el
botijo, luego hacer gavillas, luego coger los tractores, y cuando a los 19
años me monté en el camión, para mí no era ningún esfuerzo. Luego
Ana Rioja Jiménez68
a los chóferes les retaba a cargar y descargar el camión, a ver quién lo
hacía antes. Era un chaval acostumbrado a todo. Guardo muy buenos
recuerdos de mis compañeros. Éramos solidarios y nos queríamos e
hice grandes amistades. A los curas los recuerdo con respeto y cariño.
—De los 13 a los 19 años continuó sus estudios en Zaragoza, ¿ya fuera
del internado?
—Cuando me repuse de la pulmonía, mis padres decidieron
que no volviera al internado y fuera a vivir a un piso que teníamos
en Zaragoza junto a dos hermanos, una señora del pueblo que nos
cuidaba y sus dos sobrinos. Entonces me matriculé por libre en Peritaje
Mercantil e hice el ingreso, primero y segundo en el mismo año.
Después ya, de forma oficial, cursé de tercero a quinto. Lo terminé y
me matriculé en la carrera de Profesorado Mercantil, pero no llegué a
terminarla. La idea era compaginarla con el trabajo, pero no se pudo.
Me faltó voluntad. Fueron años muy felices, en los que estudié poco
y me divertí mucho. Y recuerdo, sobre todo, las tardes en el cine.
—¿Se arrepintió de no haber acabado los estudios?
—Yo he soñado mucho con eso, que tenía que estudiar, que tenía
examen y nunca me examinaba. Tenía una espinita ahí clavada…
Como estaba en la orla, todo el mundo pensaba que había terminado
la carrera, hasta que un día lo dije en público, en un discurso a
unos graduados del IESE y ya descansé. «Yo he tenido éxito y no he
terminado los estudios —les comenté—, así que vosotros, que estáis
graduados, con más motivo tendréis éxito». Al confesar ese secreto,
descansé, y ya no he soñado más con exámenes.
—¿Cuándo se vino toda la familia a vivir a Zaragoza?
—En 1951 mis padres se vinieron a vivir con nosotros a Zaragoza
y trasladaron la empresa, junto a mis tíos, de Peñalba a la capital
aragonesa para retomar aquí el transporte de mercancías. Para
Zaragoza compraron dos camiones, un Pegaso nuevo y otro viejo, un
Miguel V. Carreras. De vocación, empresario69
Hispano-Suiza, con un motor Lelilan de 125 CV, que es en el que,
dos años después, comencé a trabajar yo.
—¿Pudieron ver sus padres sus logros, sus primeros éxitos?
—A los 28 años ya comencé a tener éxito. Fue también cuando
ganamos el primer millón de pesetas de beneficios. Recuerdo la alegría y
el orgullo de mi padre. Ese dinero se reinvirtió, claro. A los 30 ya llevaba
la empresa, y mis padres y tíos me dejaban hacer. Los dos vivieron mi
éxito. Mi padre me escribía cartas en las que me decía: «Ten cuidado,
que estás muy alto y es más difícil mantenerse que llegar».
Conoció la empresa, las delegaciones, la parte comercial, Eizasa, la
agricultura… Sí que disfrutó mi padre conmigo. Estaba siempre con
él. Mi padre era un hombre muy inteligente. Me enseñó el camino
con el sentido común, a ser riguroso, a no mentir. Muchos de mis
valores se los debo a él, porque nos educó a mis hermanos y a mí, no
solo con la palabra, sino con el ejemplo. Mi padre me introdujo en el
trabajo bien hecho, y me enseñó a disfrutar con él.
En esa adolescencia zaragozana, tras salir del internado, una de
las pasiones de Miguel Carreras era ir al cine, fascinado por las
películas americanas e italianas que veía en El Monumental (calle
Conde Asalto) y en el Frontón (calle Cinco de Marzo), en los bancos
corridos de madera de arriba, destinados a los militares y estudiantes,
al precio de 50 céntimos la entrada más barata. Las vidas de cine de
los protagonistas de esos filmes, con aquellos galanes que fumaban
Chester e iban acompañados de hermosas mujeres rubias y conducían
espectaculares coches, también le hicieron soñar. Soñaba con tener
algún día un traje blanco, un descapotable y un gran amor.
No pasaron muchos años hasta que todos estos deseos se cumplieron.
A los 19 años se enamoró de la mujer que ha caminado junto él toda
la vida: Marinati (María Natividad) a la que conoció estudiando
Peritaje Mercantil y con la que ha formado una familia de siete
hijos y trece nietos. También tuvo un traje blanco de Ermenegildo
Ana Rioja Jiménez70
Zegna, ya cumplidos los cuarenta, que se compró en Jaca y que se
puso para una boda con una pajarita y unos zapatos azules. «Parecía
un actor italiano», recuerda. Ya no ha vuelto a tener más trajes
blancos. También fumó Chester. De chico, cigarrillos sueltos que
compraba en El Tubo y, a partir de los 19 años, ya se pudo comprar
cajetillas, que compartía con sus amigos y su compañero de camión.
Y, finalmente, se compró un BMW descapotable.
Pero hay otro sueño, aún más hermoso, que le acompañaba de niño
cuando veía las tierras de secano de Los Monegros, agostadas cuando
llegaba el verano y en las que muy pocas veces se cogía cosecha
porque no llovía. Deseaba tener un campo con grifo, al que nunca
le faltase el agua. En 1983, Miguel Carreras compró su primera
finca de regadío.
Miguel Carreras a los 19 años.
Miguel V. Carreras. De vocación, empresario71
CAPÍTULO IV
Los comienzos de Transportes Carreras, S.A.
De ayudante de chófer a administrador único
El espíritu emprendedor de Jesús Carreras Pomar y su afán de
superación anidaron también en su hijo, que creció escuchando
sus consejos y admirando a su padre. Ya de pequeño, Miguel
Carreras era un negociante nato y, muy pronto, se sintió atraído
por el mundo del transporte, la compraventa y la empresa. Nunca
ha escatimado esfuerzo y tiempo en conseguir sus proyectos, esos
que soñaba y que se fueron haciendo realidad, multiplicados por
diez, por cien, incluso por mil. Su vida está jalonada de pequeños
y grandes éxitos, esas metas que se ponía en su camino y que iba
alcanzando, primero a pequeños pasos y luego a grandes zancadas.
Ha escrito la historia del Grupo Carreras, página a página. Hoy
parece que fue rápido y fácil, pero son 70 años de trabajo, uno
tras otro, siete décadas de correr riesgos, de inversiones, de saber
hacer con intuición, pero también con mucho sentido común y
rodeándose de un gran equipo. Le costó 25 años sentar las bases
de la empresa, abrir 18 delegaciones por toda España. La mayor
de las inversiones fue su tiempo, su vida. Todo, o casi todo lo
importante, comenzó en el bajo de una vivienda de Barcelona,
en una pequeña portería.
—Y con 19 años se subió a un camión como ayudante de conductor.
—Tenía 19 años cuando me incorporé a la pequeña sociedad familiar
fundada por mi padre y sus hermanos. Un 1 de julio me subí por
primera vez a un camión como ayudante de conductor. Solo tenía el
Ana Rioja Jiménez72
carné de coche. Entonces, la empresa tenía dos camiones. Uno de mis
tíos conducía el Pegaso y mi primo Víctor, que era el chófer y quien me
enseñó a conducir, y yo íbamos en el Hispano-Suiza. Nos llevábamos
muy bien. Éramos como hermanos y nunca discutimos. Comencé como
quería mi padre, desde abajo, como aprendiz, para conocer a conciencia
la filosofía del negocio. Me gustaba conducir y lo hacía bien, porque
muy pronto comencé a llevar yo el camión por las noches. No había
controles y, si alguna vez nos hacían parar, mientras el camión frenaba,
nos cambiábamos de asiento y, cuando llegaba el guardia, ya estábamos
cada uno en su sitio. El conductor confiaba en mí y hasta se acostaba a
dormir. Recuerdo que yo cobraba 2.000 pesetas al mes.
—¿Cuándo pasó a ser conductor?
—Al año siguiente me saqué el carné de conductor de primera
especial y ascendí a jefe de camión. A los 20 años ya era chófer.
Durante tres años, recorrí toda España. Me gustó mucho, estaba
muy ilusionado. Conocer las carreteras de España, sus ciudades,
las costumbres, las comidas de cada lugar, las personas... Era una
maravilla descubrir todo eso. Tuve varios compañeros conductores
y disfrutábamos con el trabajo. A todo esto ya tenía novia. La veía
poco, una vez al mes. Ella terminó sus estudios y volvió a Daroca.
Yo tenía mucha paciencia. Me respetaban porque era buen
conductor, humanamente era igual que ellos, cargaba y descargaba,
y cuando teníamos averías en los puertos, de noche, con nieve y bajo
cero, el primero que se echaba al suelo era yo. Poníamos unos sacos
en el pavimento y a cambiar lo que hiciera falta, ruedas y todo. Era
joven y fuerte, tenía mucha energía y ganas.
—¿Qué transportes hacía entonces?
—Cuando me incorporé transportábamos de todo, igual llevabas
patatas, que hierro, que cemento, que muebles, chatarra, lo que te
encargaban, transporte discrecional. Hasta pescado con hielo y con
lona desde Madrid y descargábamos por el camino, en los restaurantes
Miguel V. Carreras. De vocación, empresario73
Diferentes modelos de camiones utilizados por Carreras durante los años 50 y 60.
El de la fotografía en color fue restaurado y se expone en los talleres de PlaZa.
Ana Rioja Jiménez74
de Alhama, Calatayud…, y llegábamos a Zaragoza al mercado y el
producto estaba en buenas condiciones. Nos costaba siete u ocho horas
llegar de Madrid a Zaragoza. Con el pescado, menos, porque era más
ligera la carga. A Barcelona, otras ocho horas y ahora nos cuesta menos
de cuatro con 40 toneladas. Antes el camión tenía 125 CV y ahora
tienen 650 CV.
—¿Era muy dura la vida del camionero en esa época?
—Yo iba en el camión 18 horas y a veces 24, todos los días,
solo bajaba para cargar y descargar. Nos turnábamos para conducir.
Trabajábamos los sábados y los domingos por la mañana. Dormíamos
poco y no en la cama, sino en la litera del camión, cargábamos y
descargábamos con frío, con hielo o con mucho calor. Hacíamos 80
y 90 horas a la semana y nadie se lesionaba. Esa fortaleza me hacía
aguantar el sueño, el hambre, la sed, el sol, el frío, dormir a ratos
sin comodidades, con ruidos y bocinazos. Te acostumbras a todo,
también al riesgo. Conduces tanto noche y día que convives con
el riesgo. Por eso luego no he sido temeroso ni en la vida ni en la
empresa. Después de esos tres años, me dije: «ahora que venga lo que
sea, que seguro que es todo más fácil».
—¿Qué aprendió en esos tres intensos años como chófer?
—Fue un gran aprendizaje. Me vinieron muy bien humanamente.
Si estás tres años de conductor, te pones en su piel, sabes cómo piensan,
qué hacen, conoces sus necesidades, sus anhelos. Las conversaciones
con los otros chóferes eran muy aleccionadoras. Aprendí a convivir
con la gente humilde. Era difícil que me engañaran sobre cómo se
producían las averías, por ejemplo. Cuando se sentaba uno en el
camión ya sabía si valía o no, solo por la forma de sentarse, de coger
la palanca, el volante…
Eso me sirvió de mucho, sobre todo después, para saber mandar,
cómo se manda y qué es lo que mandas. Eso, lo sabía, porque primero
me mandaba a mí mismo, lo que creía que había que hacer. Y desde
Miguel V. Carreras. De vocación, empresario75
Miguel Carreras, a los 22 años paseando por Madrid,
ciudad en la que hizo el Servicio Militar.
Ana Rioja Jiménez76
luego, me enseñó a vivir con esfuerzo. Eso te da una fortaleza y
unas ganas de todo y una gran seguridad en ti mismo. La dureza de
esos trabajos hace que confíes en ti. Mi actividad para emprender
negocios desconocidos y superarlos viene de mi preparación, de la
confianza en mí. Estaba preparado para todo y no me ha dado miedo
nada. He cogido siempre todo lo que iba viniendo con fuerza, con
energía, con convicción. No he dudado en que iría bien cualquier
negocio que comenzaba. Sabía que tarde o temprano saldría. Luego,
también, he sido muy fuerte y he tenido salud.
Esa fortaleza y esas vicisitudes de las que hablábamos antes hacen
que convivas con el riesgo. Por eso, luego no he sido temeroso en las
inversiones, en pedir préstamos, en invertir con dinero que no era
mío, porque amo el riesgo, a mí me gusta el riesgo, he convivido con
él sin sobresaltos.
—¿El Servicio Militar interrumpió ese trabajo?
—De los 22 a los 24 años hice el Servicio Militar, pero también fue
una experiencia muy buena. Aprendí a obedecer órdenes hasta de los
veteranos, disciplina, limpieza, cocina... Hice buenos amigos y tengo
grandes recuerdos. Estuve seis meses en Madrid y estudié Gonio y
Morse. Luego me destinaron a Zaragoza, a Capitanía, y descifraba los
partes, mandábamos mensajes a Madrid. Se me daba bien. Trabajaba
un día y al otro fiesta, que era cuando iba a la empresa, daba vuelta
por los camiones y hacía viajes cortos.
—Terminó la mili, se casó y volvió al negocio, pero con una nueva
idea, que lo cambiaría todo.
—Me casé cuando vine de la mili, a los 24 años, y continué con mi
trabajo en el negocio, entonces ya cobraba 3.000 pesetas y teníamos
tres camiones. Yo conducía uno, lo llevaba por la noche hasta
Barcelona mientras el otro chófer dormía y luego descargábamos.
Además, me preocupaba de buscar la carga para todos los camiones,
y pasados unos meses pensé: «Esto no puede ser, que esté toda la vida
Miguel V. Carreras. De vocación, empresario77
en el camión viajando a Barcelona y Madrid e ir buscando cargas
siempre en estas condiciones no tiene futuro». Cambié el modelo de
negocio para tener relación directa con los clientes. Sabía que cuando
hiciera eso se me cerrarían las puertas de las agencias, pero se me
abrirían otras. A los primeros clientes que busqué fue a los que ya me
conocían, y como sabía lo que me cobraban de comisión las agencias,
pues les ofrecía ese descuento.
Era una decisión arriesgada, muy difícil el éxito y muy fácil el
fracaso, pero no pude frenar mis impulsos. Al principio me costó
un poco, pero fue la mejor decisión para expandir el negocio.
Pasábamos de ser flotistas y transportistas, a ser empresa. Siendo
más barato no era fácil dar servicio cuando había que contratar
más camiones. Me costó mucho ese equilibrio de ofrecer un precio
más barato de lo que se estaba pagando y encontrar un camión
Miguel Carreras se casó con María Natividad Lario en 1958.
Ana Rioja Jiménez78
que hiciera ese viaje y que nosotros ganáramos también algo. Tuve
que pedalear más rápido, buscar clientes más rápido y hacer de
agencia. Luego ya tuve clientes a precio de mercado.
Sabía que, si me daban a mí el viaje, ya no podía trabajar con las
otras agencias. Eso se correría: «A Carreras no le des un viaje porque
te quita el cliente», porque yo se lo quitaba. Y pensé: «Lo voy a pasar
mal un tiempo, pero si no hago esto nunca tendré mi empresa». Así
me fui haciendo con los primeros clientes que tuve. Pero claro, me
decían: «Yo tengo cuatro viajes para Zaragoza y uno para Madrid. Y
ya que estás tú, te doy todos». Eso me obligaba a buscar camiones de
fuera. Con los tres camiones hacía pocos viajes. Tenía que contratar
otros y para ello necesitaba tener una oficina, un contacto, a alguien
que me ayudara con ello.
—Y así nació la primera delegación de Transportes Carreras, ¿en una
portería de Barcelona?
—Sí. La primera delegación que abrí fue en Barcelona. Me ayudó
un amigo que se llamaba Hurtado y que trabajaba de portero en un
edificio de la calle Comercio, cerca del bar Colón, al lado de la estación
de Francia, en La Barceloneta. Es curioso que en un garito debajo de
la escalera de un edificio comenzara todo. Fue mi primer delegado,
un hombre de confianza que cobraba y pagaba los viajes. Lo hice mi
empleado. Fue la primera delegación, en una portería. Empecé a tener
más viajes y ya alquilé una oficina porque tenía más clientes.
—Aquella portería lo cambió todo. ¿Fue el primer paso de un gran hito?
—Cuando me decidí a dar ese paso, todo cambió, fue el hito más
importante de la empresa. Todo lo que hay hoy es obra de lo que
comenzó en aquella portería. Hay muchos transportistas que son
flotistas, que tienen camiones y trabajan para una empresa fija o para
agencias, pero no son empresa. Nosotros somos empresa. Damos
un servicio completo al cliente y con nuestros medios. Otros, solo
colocan sus camiones.
Miguel V. Carreras. De vocación, empresario79
La aventura de la minería
Paralelamente al desarrollo de esa primera agencia en Barcelona,
hice mis primeras incursiones en el mundo de la minería. Desde
esa delegación hacíamos muchos viajes de Barcelona a Madrid y a
Bilbao. Ida y vuelta. Como Madrid no era industrial no había carga de
vuelta para los camiones. Y eso me preocupaba. Por aquel entonces,
un amigo de Ontiñena, que se llamaba Calvo, me dijo: «Yo tengo
un primo que es el director general de Blanco Nevil —que era una
fábrica de Barcelona que utilizaba sulfato de bario para fabricar sus
productos— y compra mucha barita». Entonces había minas de este
mineral en Tobed, Calatayud, Córdoba, Flix… Con esa idea, se me
ocurrió que podríamos comprar la mercancía a esas minas, venderla
nosotros y transportarla, claro. Empecé a buscar dónde estaban
esas minas, compraba la existencia y se la vendía a la fábrica.
Pero surgió la posibilidad de que me hiciera socio de Baritosa, una
empresa que explotaba una mina que estaba ubicada en Tobed,
y cuyo propietario era Salvador Figueroa, una gran persona con la
que tuve una buena amistad. Y fuimos a medias. Así comenzó el
negocio de explotador de minas. En Aragón teníamos yacimientos
en Tobed, con unos 30 mineros, y en Herrera de los Navarros, otra
mina muy importante con 40 mineros. Todas las semanas íbamos
un día a la mina y hasta nos metíamos en la galería para ver el filón.
Cargábamos unos 3 o 4 viajes diarios de producción y con eso ya
tenía las vueltas a Barcelona aseguradas. Y además fabricamos
unas tolvas para que los camiones se cargaran solos. Llegaban de
Madrid después de descargar, se ponían debajo de la tolva y por la
mañana estaban ya en Barcelona.
Otra gran experiencia en la que conocí de cerca la minería y la
profesión tan dura que es ser minero. Había que estar dentro de la
mina, muy adentro, para extraer la barita. El terreno no era firme y
se nos perdía con frecuencia el filón y había que buscarlo, mientras
toda la gente estaba parada. Todos los mineros estaban contratados
por nosotros. Estuvimos unos 10 años, hasta que cerró Blanco Nevil y
ya no sabíamos a quién vender la barita, y dejamos las concesiones.
Lo curioso del tema es que trabajé esas minas para dar mercancía a
los camiones que volvían a Barcelona. Aproveché esa oportunidad.
Ana Rioja Jiménez80
Miguel Carreras abraza a su padre en el homenaje que la empresa Pegaso
rindió a Jesús Carreras Pomar.
Directivos de Pegaso, entre ellos Ramón Cebrián, y de Carreras,
el día del mencionado homenaje.
Miguel V. Carreras. De vocación, empresario81
Su hermano, Jesús Carreras Calvete,
se incorporó a la empresa en 1968.
Transportes Carreras en el Polígono de Cogullada, sede de Transportes Carreras,
antes de trasladarse a PlaZa.
Ana Rioja Jiménez82
—¿Cuatro años después abrió delegación en Madrid?
—Cuando empecé con Madrid, ya vi que con 3 camiones no
podía hacer nada y dimos un paso más: comprar 25 camiones.
Aquello fue una revolución para mis tíos y mi padre. Ellos compraban
siempre al contado, pero yo conseguí que Pegaso me vendiera 25
camiones para pagar en 36 letras por cada uno. Realicé esa operación
con Ramón Cebrián, que era el concesionario. Creyó en mí. Me
dieron 25 camiones sin pagar nada entonces, pero con esas letras
conseguimos pagar todas en esos tres años. Yo tendría ya unos 30
años, sería 1964-1965. Mis padres y mis tíos no se lo podían creer.
No se devolvió ni una letra. Unos años después compré otros 25
Barreiros y luego era fácil ya. Esa fue una operación tremenda.
Teníamos volumen para eso y mucho más. Cargaba de 30 a 50
viajes en cada delegación.
—1968 fue un año clave. Transportes Carreras se convierte en Socie-
dad Anónima. ¿Y usted en administrador único?
—Yo quería que esa sociedad privada creada por mi padre y mis
tíos se convirtiera en una gran empresa. Lo veía muy claro, así como
los pasos que había que dar, porque yo tenía más ambición que mis
tíos, quería ir más rápido porque sabía que se podía. Ellos eran más
cautelosos y conservadores. Mi padre me apoyaba en todo, por eso,
cuando empecé a abrir las delegaciones le dije: «Mis tíos son mayores,
diles a ellos y a sus hijos que yo quiero y necesito ser administrador
general único, para tomar decisiones que son claves para la compañía».
Aceptaron y en 1968 se fundó Transportes Carreras como Sociedad
Anónima. De este modo, pude crear la red comercial que sentó las
bases para ocupar la posición que ahora tenemos en el sector. Me
responsabilicé de Carreras, Sociedad Anónima. Conforme se iban
jubilando mis tíos, iba comprando sus acciones, convirtiéndome
poco a poco en socio mayoritario. Mis tíos siguieron trabajando en
la empresa hasta que se jubilaron, mi padre hasta que falleció. Mi
primo Víctor, que pasó a ser socio de la empresa, también me apoyó
Miguel V. Carreras. De vocación, empresario83
en este proyecto familiar. Fue un gran colaborador, coordinando
los servicios de los conductores, especialmente en la delegación de
Barcelona.
—En 1968, su hermano Jesús también se incorpora a la empresa.
—Mi hermano se incorporó a la empresa en 1968. Estudió
Peritaje Industrial y formó una empresa con otros compañeros, que
tuvieron que cerrar. Mi padre y yo lo incorporamos a la empresa y
se responsabilizó de la flota de vehículos, de las compras y del taller,
más tarde de los conductores, ayudando a asentar la especialización
de tareas de los miembros del equipo directivo. Hemos trabajado
siempre muy bien juntos, nos queríamos mucho. Fue vicepresidente
de Carreras Grupo Logístico hasta su fallecimiento, en 2022. Dos de
sus hijos también se incorporaron posteriormente a la empresa: José
Antonio, tristemente fallecido, y Jesús, que aún trabaja con nosotros.
—La empresa comenzó a crecer. ¿Fueron 25 años de duro trabajo
para abrir delegaciones por toda España?
—Sí, la empresa comenzó a crecer. Cada año casi formaba una
delegación nueva, tenía que mantener los clientes y hacer nuevos.
Eso era muy costoso. Viajaba toda la semana y volvía los viernes a Za-
ragoza. Durante el día visitaba a los clientes con el delegado y el co-
mercial, y por la noche viajaba de delegación en delegación. Dormía
poco, a veces unas tres o cuatro horas, en el coche. Pero tras Madrid,
llegó Bilbao, Valencia, Sevilla, Galicia, Asturias… 25 años me costó
abrir 18 delegaciones y encontrar el personal adecuado. Pasarían 70 u
80 delegados, comerciales, administrativos. Quedarme con lo mejor
supuso un gran esfuerzo, que mereció la pena.
Entonces, ya dejé de viajar y tenía 1.150 clientes. Teníamos alre-
dedor de 60 camiones, pero eso no fue nada. Eso fue hacer empresa,
sentar sus bases: 18 delegaciones y 60 camiones. Fue duro aquello. De
ahí la suerte de tener una gran mujer y una buena esposa a mi lado,
que se hiciera cargo de la casa, de los hijos, y de recibirme con amor
Ana Rioja Jiménez84
y comprensión los fines de semana y escucharme. La mayor inversión
de mi vida fue abrir delegaciones, sobre todo en años, en tiempo y
esfuerzo. Luego apareció la logística y fue otro gran reto y otro hito.
Un hombre con carisma, un joven líder a los 28 años
No había cumplido aún los 30 años, cuando intervine en dos
huelgas nacionales de la Patronal del Transporte. Por una serie
de circunstancias, nos querían subir el precio del gasoil y en otra
ocasión, nos ponían un canon de coincidencia con RENFE.
La primera de ellas surgió en Zaragoza. Vimos que no podíamos
continuar con la actividad del transporte en esas condiciones
y los empresarios decidimos movilizarnos, porque veíamos la
necesidad de parar los camiones porque no se podía desarrollar
la actividad con beneficios. Alguien tenía que dirigir eso. Yo tenía
28 años y empecé a hablarles allí y, sin darme cuenta, lideré esa
huelga. Había que hacer piquetes y parar los camiones. Tuve que
salir a la carretera y parar el primer camión y así, uno detrás de
otro. No se dejaba pasar de Zaragoza a Barcelona, tan solo los que
transportaban medicinas y alimentos para niños.
Secundaron la huelga en todas las provincias y se paralizó el
transporte en toda España. Así estuvimos seis o siete días hasta
que, tras sucesivas reuniones en Madrid con el ministro de
Transportes, entonces Rodolfo Martín Villa, se resolvieron nuestras
reivindicaciones.
A raíz de eso, me nombraron vocal nacional del Sindicato de la
Patronal de Transportes y representaba a Zaragoza, Soria y Navarra.
Era el más joven de todos. Poco tiempo después de esa huelga
quisieron implantar a nivel nacional unas tarifas obligatorias para
el transporte discrecional. El sindicato nacional confeccionó unas
tarifas, se aprobaron y las publicaron. Antes de aprobarlas para hacer
la Ley, se hizo una asamblea nacional en Madrid donde se explicaban
las tarifas y el desarrollo de las mismas. Me leí esa publicación por la
noche y me di cuenta de que no estaban bien, que eran erróneas. Al
día siguiente, en la asamblea, en la que había más de 600 personas,
empresarios de toda España, cuando las estaba explicando Manuel
Boj, de la Patronal de Barcelona, que era muy conocido y era
economista, yo dije que creía que no estaban ajustadas. Y así lo
Miguel V. Carreras. De vocación, empresario85
expresé, que eran erróneas. Me pidieron que subiera al escenario
para explicar por qué. Estaba el ministro, el jefe de sindicatos,
asesores, 8 o 10 personas en aquella mesa presidencial.
Me temblaban las piernas. Subí allí, comencé a hablar, me pusieron
una pizarra y empecé a desarrollar mis argumentos. Estuve dos
horas explicándoles por qué no eran correctas. «¿Quien piensa
como Carreras?», preguntó Manuel Boj. Y comenzaron a contestar:
«Málaga está con Carreras, Bilbao está con Carreras…». Después
de dos horas, estaba tan claro, que Boj lamentó no habérmelas
consultado antes. Era todo sentido común.
Se suspendieron esas tarifas y se aprobaron otras más justas y
me hice muy popular en el sector en toda España y se me valoró
como persona que entendía del tema de transportes. Son dos
hitos importantes en mi vida y en mi carrera, los de una persona
modesta, pero que entendía del tema y que estaba en lo cierto:
hacer la huelga y pararla en su momento, y echar por tierra unas
tarifas de carácter obligatorio que eran un desastre.
Con estos precedentes, era un joven empresario conocido y
apreciado, y tenía unos pensamientos muy sociales. Por ello, en
1964, tenía 30 años, estábamos en una época de crisis y salió la
oportunidad de agrupar a varios transportistas para hacer una
Cooperativa de la Patronal. El objetivo fue darles trabajo directo
a todos. Se unieron 30 o 40 camiones a esta iniciativa. Luego vino
una etapa de mayor bonanza y se deshizo.
Solo una reflexión más de esta etapa: entonces habría en España
unas 300 empresas de transportes, con 30 o 40 muy destacadas.
Hoy quedan muy pocas. Sesenta años después de aquello, en
que Carreras era una pequeña empresa de 3 o 4 camiones, somos
líderes en España de distribución de alimentos. Detrás de cada
uno de esos años hay mucho trabajo, ilusión, iniciativas, ganas y
riesgo, con esa máxima de no reparto de dividendos y de inversión
constante, el doble o el triple de los retornos obtenidos, a base de
un apalancamiento muy fuerte. Durante 30 o 40 años los bancos
me hacían avalar a mí, hasta que llegó un día que dije que ya se
había acabado, y luego avalaba con la garantía de la empresa. Y
siguió todo igual sin arriesgar mi patrimonio. Era valiente a favor de
la empresa. Pocos empresarios harían eso durante 40 años.
Ana Rioja Jiménez86
Al mismo tiempo que sentaba las bases de esa red comercial,
en 1965 Miguel Carreras inició una actividad de promoción y
construcción de viviendas y polígonos industriales, que serían
el embrión de otra gran empresa: la inmobiliaria Eizasa (1974).
Comienza otra aventura maravillosa, pero de ello hablaremos en
el capítulo VI de este libro. Tan solo comentar la capacidad de
este entonces joven empresario para emprender varios proyectos
a la vez de muy diversa índole, consciente de la importancia de la
diversificación, y es que, como su padre muy bien le había dicho:
«No hay que poner todos los huevos en la misma cesta».

Miguel V. Carreras. De vocación, empresario87
CAPÍTULO V
Carreras Grupo Logístico
El buque insignia del holding familiar
Carreras Grupo Logístico, el buque insignia de este holding
familiar, es un operador logístico integral de referencia
internacional. Es líder en el sector de gran consumo en la
península Ibérica y avanza imparable en la logística industrial.
Cuenta con filiales internacionales en Portugal, Francia, Italia y
Rumanía. Ofrece servicios en todos los eslabones de la cadena
de suministro: transporte nacional e internacional, almacenaje,
distribución, manipulación, copacking, paletería, consultoría
logística, tránsitos aéreos y marítimos y gestión de aduanas y
externalización comercial.
Este año, 2023, Carreras está de celebración. Cumple 90 años.
Nueve décadas de trabajo, esfuerzo y dedicación que le han
llevado a implantar 46 naves logísticas en España, Portugal y
Francia, con una red de almacenaje de más de 600.000 m 2 . Tiene
una flota controlada de más de 1.000 vehículos.
En la empresa están empleados más de 2.000 trabajadores, y da
trabajo a más de 3.000 personas, contando con los contratos
temporales y los transportistas autónomos. Carreras cerró 2022
con una cifra de facturación de 350 millones de euros y en 2023
espera superar esta cifra.
Pero, desde aquel camión Bedford de segunda mano que, en los
primeros años 30 del siglo pasado realizaba transportes locales
y comarcales, hasta hoy, hay un largo recorrido. La dedicación
de Miguel Carreras durante 70 años a este gran proyecto, con la
Ana Rioja Jiménez88
incorporación en la dirección de la empresa de su hermano en 1968
y a partir de los años 80 de sus hijos y de una importante red de
colaboradores, ha hecho posible este logro. Hoy, a diario cargan unos
1.300 tráilers y 300 o 400 camiones más pequeños. Siempre con
una constante: no cesan las inversiones en tecnología, innovación y
digitalización. Todo ello alineado con su estrategia medioambiental,
a través del proyecto Carreras 0,0, que busca la reducción paulatina
de las emisiones de CO2 de sus procesos logísticos.
Miguel Carreras hizo de esta empresa familiar su vida para
llevarla a lo más alto y ha disfrutado realizando cada uno de estos
proyectos, muchos de ellos pioneros. En este capítulo nos relata
paso a paso cómo ha sido todo ese proceso, siempre innovando
en logística.
—Veinticinco años abriendo delegaciones de Transportes Carreras por
toda España. ¿Cuándo llegó la expansión internacional?
—En 1968 me hice propietario y administrador único de la
empresa, y comencé a tomar decisiones en solitario para crear una
gran empresa, apoyándome en la experiencia, en la intuición, y
en grandes colaboradores. Una vez creadas las delegaciones en las
principales capitales de las provincias españolas para el transporte de
mercancías, se creó una tupida red nacional y ya teníamos una gran
empresa. Había ya 260 camiones, consolidándonos como una de las
principales empresas de transporte de España. En 1970 comenzamos
la expansión internacional, abriendo delegaciones en Europa, algunas
se cerraron y otras siguen. Tres décadas después de abrir la primera
delegación en España, teníamos 1.200 clientes y ya era una empresa
muy fuerte. Cuando creé estas delegaciones, tras 25 años de mi vida
viajando y seleccionado personal, ya vi que tenía el soporte de una
empresa que iba a ir a más. Luego se han ido abriendo y comprando
naves por demanda de esos clientes, ampliando servicios y comprando
más camiones. Nosotros tenemos unos 600 camiones (con chóferes
en nómina) más unos 500 (autónomos que trabajan con nosotros).
Miguel V. Carreras. De vocación, empresario89
Carreras construyó su primer centro logístico en
la Ciudad del Transporte de Zaragoza a finales de los años 80.
Nave de la Ciudad del Transporte dedicada a logística,
entonces unas instalaciones pioneras.
Ana Rioja Jiménez90
—En 1986, con la llegada de su hijo José Luis, se incorpora la tercera
generación de la familia.
—José Luis estudió Derecho y Económicas. Completó su formación
en la Escuela de Negocios IESE de Barcelona. Al terminar vino a
Carreras y trabajó con su tío en Transportes y luego de comercial y,
como ya estaban las delegaciones, viajaba para visitarlas y ya empezó
con la relación con los clientes. Más tarde, siendo consejero, se ocupó
de poner en marcha ExpressPallet, una división muy importante de
la empresa.
—¿Cuándo aparece la logística?
—Carreras seguía abriendo delegaciones. Estamos ya en los años
80, momento que coincide con que vamos a la Ciudad del Transporte
y construyo el primer centro logístico en Zaragoza. Tras la expansión
del transporte de mercancías, aparece la logística como una novedad
en Francia e Inglaterra. Y decidí hacer logística de alimentación.
Empecé con ello con ilusión y fuerza. Contraté a personal que
sabía del tema y comenzamos a coger clientes para almacenar su
mercancía y repartirla por Aragón. Si antes el fabricante tenía que
tener delegaciones para vender y suministrar su producto en toda
España, y eso le obligaba a tener delegados, comerciales, camiones,
chóferes…, una infraestructura muy fuerte que le resultaba costosa,
con la logística, Carreras comenzó a asumir esos servicios para ese
cliente, con lo cual ahora solo tenía que tener un comercial y era un
éxito seguro. Comenzamos con la logística en las naves de la Ciudad
del Transporte y entonces distribuíamos para Aragón. Era dar todo
ese servicio al cliente en una región.
—Otro hito, otra carrera contrarreloj… ¿Veinticinco años más para
implantar la logística?
—A los dos o tres años ya se puso de moda que el cliente, en
lugar de tener 14 delegaciones y 14 transportistas locales en diferentes
puntos de la geografía nacional, tuviera una sola en España. Entonces
Miguel V. Carreras. De vocación, empresario91
comenzamos a hacer, a marchas forzadas, delegaciones de logística,
igual que las otras, igual que ocurrió con el transporte, a hacer naves
por toda España y llegó otra carrera fuerte con la logística.
—Nace el primer Centro de Logística en Zaragoza. ¿Cómo ha sido el
proceso hasta llegar a los 46 almacenes que tienen hoy?
—Al poco tiempo de abrir el Centro de Logística en la Ciudad
del Transporte, comprendimos que el cliente no quería tener tantos
distribuidores, sino uno para toda España. Eso nos obligó a salir
de Zaragoza y abrir delegaciones de logística por todo el territorio
nacional: naves donde el cliente pudiera tener el stock y suministrar
sus productos desde su fábrica a nuestro almacén, y desde allí nos
decía dónde teníamos que llevarlo. Había que tener naves para dar
servicio a los clientes, para que tuvieran un solo operador logístico
para toda España y fue una corrida continua para dar servicio. Fue un
proceso muy rápido y no hemos parado. Hoy tenemos 46 almacenes
de logística.
—¿Coincidió con la incorporación de su hijo Fernando?
—Mi hijo José Luis estaba en Transportes, por eso a Logística
se incorporó Fernando en 1990, que hoy es consejero delegado
del Grupo. Hizo Derecho y el Máster MBA del IESE. Al terminar
este Máster se incorporó a la nueva empresa de Logística, Carreras
Almacenaje y Distribución, S.A. Yo había hecho ya la sociedad, que
llevaba un año funcionando, y estaba esperando a que viniera él para
apoyar ese proyecto y expandirlo. Y así lo hizo.
Empezamos a captar clientes y desarrolló e impulsó la logística
junto a otros directivos: compramos naves y abrimos almacenes en
Barcelona, Madrid, Sevilla y por toda España. Entonces teníamos
separadas las empresas. En el polígono de Cogullada estaban los
Transportes y en la Ciudad del Transporte, la Logística.
Ana Rioja Jiménez92
Era más conveniente tener dos empresas distintas para ganar
la confianza de los clientes. Una transportaba cemento, madera,
hierro, chatarra… y la otra, alimentos como galletas, mantequilla,
chocolate… Eran clientes distintos y mercancía diferente. Una vez
que la logística tuviera entidad y competitividad, las agruparía para
una mayor identidad. Y así lo hice años después en el polígono PlaZa,
donde se fusionaron ambas física y legalmente, y ya todo es Logística
Carreras.
—Y en 2003 comienzan a operar también en Portugal.
—Sí, hace 20 años también comenzamos con la logística en
Portugal. Carreras integra así sus servicios en toda la península
Ibérica, siendo líderes en gran consumo; también en las islas, Canarias
y Baleares. Hace 5 años que estamos también en Francia. Es un país
muy grande y va a costar, pero queremos crecer allí para reproducir
en el país vecino el éxito obtenido en España y Portugal. Empezamos
al lado de París, en Orleans. Ahora vamos a bajar al sur.
—Y en 2006 se trasladaron a PlaZa, la Plataforma Logística de
Zaragoza.
—Se creó PlaZa y allí que nos trasladamos en 2006. Compré
90.000 m2 para que no necesitásemos cambiar de sede en 50 años o
más. Comenzamos a construir por los lados. En un extremo el taller,
de 8.000 m2
, que es de los mejores que hay en España, donde tenemos
todos los servicios para reparación y mantenimiento de camiones, los
propios y los de los autónomos: electricidad, chapa, pintura, carrocería,
carpintería, mecánica… Hay una treintena de trabajadores, pues lo
hacemos todo. Eso es importante. Nos permite tener un mantenimiento
preventivo, de forma que los camiones desde el principio se cuidan
mucho (dos millones de kilómetros sin reparar) y esto nos favorece.
En el otro extremo hicimos las naves de los almacenes de 25.000 m2
. Y
así, conforme fuéramos necesitando más espacio, iríamos ampliando.
Hemos hecho otra ampliación de otros 25.000 m2 al lado de las naves
Miguel V. Carreras. De vocación, empresario93
y ahora vamos a hacer otra más de 10.000 m2
. Además de los 90.000
m2
, recientemente hemos comprado otras dos parceles enfrente de
30.000 m2 y haremos esa nueva nave de 10.000 m2 y los 20.000 m2
los destinaremos a parking. Y donde está actualmente el parking,
haremos naves. Fue un gran acierto y ahí nos quedaremos. Con estas
instalaciones, para Zaragoza estamos servidos.
—¿Estas instalaciones permitieron tener en la misma sede Transportes
y Logística?
—Sí y fue un gran avance. Además de haber ganado mucho
espacio, lo mejor es que en el mismo lugar están todas las empresas.
Unificamos ya todo y se fusionó en la misma sede Carreras Transportes
y Carreras Logística. Este es el cuarto cambio. Cuando vinieron mis
padres a Zaragoza, estuvimos primero en la calle Latassa. Compramos
un edificio y en los bajos se instalaron los talleres de camiones, que
eran pequeños. Después nos trasladamos a la avenida Valencia, que
Fernando Carreras Lario se incorporó a la empresa en 1990.
Hoy es consejero delegado de Carreras Grupo Logístico.
Ana Rioja Jiménez94
Sede de Carreras Grupo Logístico en PlaZa (Zaragoza)
con sus diversas instalaciones y parking.
Miguel V. Carreras. De vocación, empresario95
Carreras Grupo Logístico tiene una flota controlada
de más de 1.000 vehículos.
Ana Rioja Jiménez96
construimos nosotros el edificio junto con mi suegro. Después fuimos
al polígono de Cogullada con 4.000 m 2. Cuando nació la Ciudad
del Transporte, reservamos una parcela de 14.000 m 2 y construimos
unos 6.000 m 2 para albergar camiones y oficinas. Pero llegó la logística
e iniciamos todo el proceso allí y en Cogullada continuamos con los
talleres. Con el traslado a PlaZa, hemos destinado las naves de la
Ciudad del Transporte para logística industrial. Cogullada ahora está
cerrado, quizá la empleemos para acoger más logística industrial. Está
por ver.
—¿En qué consiste la logística industrial?
—Dentro de la logística, hemos comenzado con la industrial,
que ya no es alimentación, y hay menos empresas de logística que se
dedican a esto. La comenzamos hace unos años. Grandes empresas
industriales que tienen instalaciones por toda España, cuando sufren
averías, deben llevar las piezas a esos lugares y tener existencias. Antes
eran ellos los que desde sus almacenes transportaban esas piezas.
Ahora, nos dicen: «Aquí tenéis un almacén entero para gestionar con
todas las piezas que solemos cambiar y os avisarán de Málaga, La
Coruña, o de donde sea y tendréis que llevar las piezas que os pidan
allí». Son piezas especiales, hay algunas que pesan dos toneladas. Estas
industrias necesitan a alguien que esté capacitado para estocar sus
productos y dar servicio a las instalaciones que tengan en toda España.
De momento, tenemos cinco distritos ya de logística industrial con
naves propias: Zaragoza (Ciudad del Transporte), Madrid, Barcelona,
Sevilla y Málaga. Con ello ofrecemos servicios de externalización de
almacenes monocliente. Entramos en las fábricas de nuestros clientes.
Son empresas que mueven miles las piezas. Los almacenes para
esta logística son diferentes. Tenemos solo siete u ocho, pero habrá
que avanzar más, porque cada día son más las empresas que nos lo
encargan, conscientes de que estamos operando con buen servicio.
Y ahora estamos también en otro proceso que es la logística a
temperatura controlada, y que es más complejo. Estamos colocando
Miguel V. Carreras. De vocación, empresario97
cámaras en las distintas naves para poder dar este servicio, porque en
Portugal y Francia ya lo estamos haciendo. No se puede parar.
—¿Cuándo nació ExpressPallet?
—Es otro servicio que creamos en 2006. Antes llevábamos
camiones enteros con los productos del cliente, pero cada vez este es
más exigente. Todos los productos alimentarios tienen una caducidad
y los fabricantes quieren que los clientes que los consumen se beneficien
de ellos en su mejor momento, cerca de la fecha de la fabricación. Y el
que lo compra lo quiere del día uno para tener en su despensa y que le
queden días para vencer. Como consecuencia, los pedidos no pueden
ser tan grandes y hay que fraccionarlos por pallets.
Los clientes solicitan que no les mandes un camión, que son
25.000 kilos, sino un pallet, que son unos 500 o 600 kilos. Hubo que
crear otra empresa, filial nuestra, especializada en paletería, y realizar
el almacenaje y la distribución, apoyándonos en nuestras naves,
vehículos y rutas. De este modo, Carreras ofrece servicio urgente para
entregar de 1 a 33 pallets en plazos de 24 a 48 horas para España y
Portugal, y a partir de 72 horas para parte del resto de Europa. Hoy
distribuimos 8.000 o 9.000 pallets diarios, que son muchos. Requiere
otra organización y otro sistema, pero al utilizar las mismas redes que
la logística nos hace ser muy competitivos.
—Tan solo un año más tarde se incorpora el Copacking.
—Este nuevo servicio se incorpora en 2007. Ahora en nuestros
almacenes podemos finalizar operaciones productivas y completar
los productos del cliente con diversas maquinarias para su
acondicionamiento en cuanto a su manipulación y embalaje para
ofertas especiales y promociones de su interés.
—¿Cuándo llega su hijo Ricardo a la empresa?
—Hace diez años, en 2013, Ricardo se incorpora a la empresa como
director de Estrategia, Marketing y Comunicación. Hoy es consejero y
Ana Rioja Jiménez98
desarrolla actividades como la atención personalizada a determinados
clientes, la supervisión de varios departamentos y el impulso de las
nuevas tecnologías y la digitalización, tan importante hoy para el
desarrollo del Grupo. Estudió Ciencias Empresariales (primero en
la Universidad de Zaragoza y luego en EAP-ESCP en París, Londres
y Berlín) y Ciencias Políticas en la UNED, ha realizado diversos
másteres con becas en el Colegio de Europa (en Brujas, Bélgica) y la
Universidad de George Washington (en Washington DC) y es Doctor
por la Universidad Complutense. Aprobó unas oposiciones para ser
funcionario de la Comunidad Europea, aunque finalmente decidió
no incorporarse. Estuvo trabajando como consultor en Washington
y allí fundó una empresa de asesoría en comunicación. Regresó a
España después de 12 años estudiando y trabajando en varios países
y fundó la empresa Iberanalisis (Top Position, ID Digital School, ID
Bootcamps, Selecta) y es director del Máster en Marketing Digital,
Comunicación y Redes Sociales de la Universidad Camilo José Cela
(UCJC), tareas que compagina con su trabajo en Carreras. De hecho,
está la mitad de la semana en Zaragoza y la otra mitad, en Madrid. Su
bagaje en marketing, comunicación y nuevas tecnologías le ha venido
muy bien a nuestra empresa, ya que ha impulsado nuestra renovación
tecnológica y la digitalización.
—En 2016 se compra Metafase y Carreras ya ofrece servicios de gestión
del punto de venta.
—Este servicio es muy interesante. Nuestros clientes tienen un
subcliente, por ejemplo una gran superficie, y esta empresa hace un
seguimiento de su producto: dónde está ubicado, si está en almacén
o en estantería, cómo está expuesto… Esta empresa lo vigila y envía
al cliente emisiones continuamente a tiempo real sobre su producto
porque está monitorizado. Y lo más interesante es que, además de la
gestión del punto de venta, ofrece servicios de asesoría, consultoría,
formación e investigación de mercado.
Miguel V. Carreras. De vocación, empresario99
—Otro hito importante sucede en 2017, que se adquiere la transitaria
Carreras Forwarders.
—Forwarders, con sede en Barcelona, realiza tránsitos interna-
cionales marítimos, aéreos y terrestres, así como gestión de aduanas,
lo que nos permite prestar servicios en todo el mundo. Es una em-
presa que lleva un container en un tráiler, barco o avión a cualquier
parte del mundo, con medios subcontratados. Se brindan servicios
Almacenes de logística del Grupo Carreras en PlaZa (Zaragoza).
Ana Rioja Jiménez100
de logística, ya sea como agente aduanal, exportadores o realizando
asesoría para importar, de forma personalizada y atendiendo las ne-
cesidades específicas de los clientes. Por ejemplo, en la pandemia tra-
jimos un avión de mascarillas de China. Y podemos llevar un barco,
dos barcos, un naviero, lo que se necesite porque se subcontrata. Un
cliente nos pide que llevemos un container a Nueva York o a Hong
Kong y se lo organizamos todo.
—También en 2017, Carreras se instala en Francia.
—Compramos la empresa francesa Transeco (hoy Carreras Francia),
con sede central en Orleans. Además de un equipo de profesionales
con dilatada experiencia en el sector, también ha incorporado naves
propias con capacidad de almacenaje de 40.000 m2 y una extensa
red de colaboradores repartidos por toda Francia, lo que nos permite
ofrecer servicios de almacenaje, manipulación, distribución, transporte,
paletería o copacking en cualquier parte del país vecino. Actualmente,
nuestro negocio internacional ya cuenta con filiales propias en Portugal,
Italia y Rumanía, además de Francia. De esta manera, avanzamos en
nuestro plan estratégico de ser un operador relevante en la logística
integral de todo el suroeste de Europa.
—Y se ha lanzado una nueva filial especializada en depósitos fiscales.
—Sí, Carreras Grupo Logístico ha lanzado una nueva filial
especializada en depósitos fiscales, que ofrece a nuestros clientes
una capacidad de 4.000 m 2 en nuestras instalaciones de Masquefa
(Barcelona). La empresa, Carreras Servicios Integrales (CSI) gestiona
todos los procesos relacionados con los depósitos fiscales, y ya cuenta
con varios clientes importantes en el sector de bebidas y licores.
Otra constante del Grupo ha sido la apuesta por los sistemas,
la tecnología y la digitalización. Carreras está culminando con
éxito un plan de modernización de sus procesos y sistemas que
arrancó en 2016 e incluye 21 proyectos, entre los que destacan la
Miguel V. Carreras. De vocación, empresario101
fusión de sus programas operativos, un nuevo ERP (Planificación
de Recursos Empresariales) en la nube y una vanguardista
arquitectura de datos. Así consolida su posición tecnológica en el
sector, aliado de gigantes como Microsoft o IBM. Un ambicioso
plan de transformación digital que refuerza su liderazgo en
logística.
—Para llevar a cabo todos estos servicios, muy importante es también
la tecnología y la digitalización.
—Eso ha sido una revolución. Incorporamos tecnologías
avanzadas en todos los eslabones de la cadena logística. Llevamos
décadas desarrollando una gran cantidad de herramientas
tecnológicas específicas para nuestros clientes, que nos permiten
gestionar eficientemente nuestros centros de stock, la caducidad de
los productos y la logística inversa. También nos ayuda a gestionar la
distribución. Nos permite seguir en tiempo real las actividades de la
flota y de los pedidos.
Como operador de cuarta generación, ofrecemos módulos de
gestión vinculados a la actividad logística que nos han permitido
aumentar la calidad de la información de nuestros clientes, lograr una
mayor integración entre sus sistemas de información y los nuestros,
y reducir los costes y errores de manera conjunta. También hemos
desarrollado tecnología específica para las plantas de producción de
nuestros clientes. Y ofrecemos servicios específicos de consultoría de
sistemas logísticos.
—Para ello necesitarán un gran equipo de ingenieros informáticos.
—Nuestro equipo está formado por un centenar de informáticos.
En España trabajan con nosotros alrededor de 80 y el resto en otros
países, pero aún tendremos que contratar más a medio plazo.
Son millones los apuntes diarios que se hacen con el transporte de
los camiones, millones, y ya no hay papel, todo está informatizado.
Ana Rioja Jiménez102
En 2017 se crea la transitaria Carreras Forwarders,
que permite la prestación de servicios por todo el mundo.
La tecnología y la digitalización son muy importantes para el Grupo.
Carreras cuenta con más de un centenar de ingenieros informáticos.
Miguel V. Carreras. De vocación, empresario103
Nave del Grupo Carreras en Illescas (Toledo),
uno de los centros logísticos más modernos y sostenibles de Europa.
Carreras Grupo Logístico crece en Francia. Recientemente ha abierto
dos nuevas instalaciones, una de ellas en Fleury-les-Aubrais (distrito de Orleans).
Ana Rioja Jiménez104
Cada apunte de referencia de un producto, que tiene 14 referencias
distintas y un cliente que tiene un montón de productos cada uno
con sus referencias. Y con estas referencias se realiza un seguimiento
a tiempo real de este producto, dónde está, dar de baja de dónde
estaba, de alta de dónde está, y comunicarle al cliente que se ha llevado
y que lo han recibido bien. Eso son millones de apuntes diarios y
millones de movimientos, porque todos esos productos están en
nuestras naves, en las estanterías. Hay que ir sacándolos físicamente,
llevarlos al camión y repartirlos, dejar constancia de todo el proceso.
Y conocer la caducidad que tienen las existencias, para ver qué tienen
que sacar antes, si tienen que aplicar algún descuento, etc. Tenemos
que informar de todo a nuestros clientes, de las incidencias, de cuándo
se ha servido y a qué hora para que puedan facturar.
—¿Cuántos trabajadores están contratados en Carreras Grupo
Logístico para dar todos estos servicios?
—Tenemos en nómina a más de 2.000 personas, más luego
trabajadores que nos proporcionan las ETT (Empresas de Trabajo
Temporal) que no están en nómina, pero que subcontratan para
nosotros, y autónomos. Más de 3.000 personas en total. En
distribución, por ejemplo, hay unos 400 autónomos que trabajan
con nosotros, en la capilar, en la última milla, y que es el proceso
final del transporte y reparto.
—¿Cuáles son los proyectos o retos más inmediatos?
—Francia es un proyecto muy ambicioso. Intentaremos conquistar
Francia. Queremos captar a clientes franceses para suministrar a
clientes franceses, porque el servicio de clientes españoles a Francia
ya lo tenemos. En logística de alimentación, principalmente, y ya
comenzaremos también con logística industrial. Y seguir con Italia y
Alemania. En estos países ya estamos con transporte, desde hace 30
años. Por eso queremos comenzar también con la logística para las
grandes multinacionales.
Miguel V. Carreras. De vocación, empresario105
En España aún podemos crecer más, tener más clientes. Los clientes
que ya tenemos están muy bien. Ahora ya es cuestión de precio y
mejor servicio para la captación de nuevos clientes en Logística. Que
no se vaya ninguno y captar y trabajar con otros. Este es nuestro
propósito y en ello estamos.
Cuando nosotros nos enterábamos de algún tender importante
(el momento de cambiar de proveedor), íbamos a por él, ahora nos
llaman. El cliente nos dice: «Voy a hacer un tender, si os interesa, podéis
concursar». Nosotros concursamos, exponemos nuestras condiciones,
el precio y te lo adjudican o no. Al cliente le interesa que esté Carreras
en esa pugna, porque tener un operador de nuestra categoría es una
garantía, y nos llaman claro. Somos una referencia en España porque,
en la actualidad, estamos continuamente ampliando naves y servicios,
tratando de ser competitivos, informando al cliente lo más posible,
que en cada momento sepa dónde está su mercancía, qué tratamiento
ha llevado; mimando al cliente, en suma, para que nos transmita la
satisfacción con su operador. En este sentido, también queremos
reforzar el servicio de la última milla, el último eslabón de la cadena,
para mayor garantía y seguridad de servicios a nuestros clientes.
—Usted afirma que la logística es sentido común, pero también en su
caso intuición y anticipación.
—Hacer todo esto, para estar en vanguardia, es un compendio
de imaginación, ansiedad, ambición, sentido común, riesgo, decisión
y convicción. Es estar informado continuamente sobre qué ocurre
en nuestro sector, cuáles son los avances, las últimas novedades a
nivel internacional, sobre todo en los países más desarrollados que el
nuestro. Y rodearse de un buen equipo que esté preparado para ello
y que te enseñe también a ti, que es la forma de aprender, formarte
profesionalmente.
Hay que echarle más imaginación, más ganas, más tiempo. Y te
vas dando cuenta de las carencias, de los problemas y necesidades que
Ana Rioja Jiménez106
tiene el cliente, de lo que se puede mejorar, de lo que demanda, y lo
estudias, te pones en su lugar porque estás más preparado que él, y los
resuelves día a día.
Cuando vi que surgía en Europa una nueva forma de transporte,
la logística, fui a por ella. Soy un empresario muy inquieto, atento a
todas las innovaciones, por eso también he tenido algún fracaso. Pero
esto salió muy bien. De hecho, la mayor parte de nuestra facturación,
además del transporte general, viene de la logística especializada en
alimentación y de ExpressPalet.
Y mientras Carreras avanzaba a pasos agigantados con la logística,
Eizasa no se quedaba atrás. Estaba construyendo en cinco provincias
viviendas y hoteles. Pero aún iba a llegar una nueva aventura: la
agricultura. En 1983, Miguel Carreras empezó con el regadío,
compró unas tierras con ese grifo que tanto había deseado. Creó
dos sociedades agrícolas y cumplió otro de sus sueños: cultivar la
tierra que tanto amaba, saciar paulatinamente la sed de sus fincas
en Los Monegros.

Miguel V. Carreras. De vocación, empresario107
CAPÍTULO VI
Eizasa Grupo Inmobiliario y Eizasa Hoteles
Llega la diversificación y el placer por construir
En los años 60 y 70, a Miguel Carreras le surgió la posibilidad de
comprar solares y comenzar la actividad de promotor inmobiliario
y constructor de polígonos industriales y viviendas. Comenzaba
la diversificación del negocio, y, siguiendo los consejos de
su padre, eso le parecía entonces lo más adecuado. Y así, casi
también por azar, un buen día le pidieron que levantara un hotel
en tres meses y lo hizo. Llegaba la General Motors y había que
dar alojamiento a centenares de personas. Para este empresario
inquieto, aventurero y arriesgado no había nada imposible. Hoy,
Eizasa ha construido cerca de 4.000 viviendas en cinco provincias
y aún tiene suelo para levantar otras 1.000 más. Al mismo tiempo,
ha creado una cadena con nueve establecimientos hoteleros que
espera ampliar.
—¿Es cierto que la venta de un camión le llevó al negocio de la
inmobiliaria?
—Todo comenzó cuando vendí un camión viejo, que valía
500.000 pesetas. El comprador no disponía de todo el dinero, pagó
250.000 y para compensar el resto, me dio una parcela que tenía
en la calle Privilegio de la Unión, en San José (Zaragoza). Al poco
tiempo, busqué un constructor para que edificara y me diera a mí
obra. Me entregó un piso y unos bajos. Él se encargó de venderlos y
por mis propiedades me dieron 750.000 pesetas. Pensé: «Esto es un
buen negocio». Prácticamente había triplicado lo invertido. Entonces
empecé a comprar solares y casas para derribar. Se los vendía a los
Ana Rioja Jiménez108
contratistas y ganaba dinero. Era un negocio muy rentable, tan
interesante que llegó un momento en que me dije: «¿Y por qué
no voy a construir yo?». No tenía ningún conocimiento del tema,
pero pensé: «Ya aprenderé». Y lo hice, porque tenía que construir
viviendas, promoverlas, financiarlas y venderlas. Así comencé a
hacer obras directamente en esos solares que iba adquiriendo. Como
consecuencia de aquella primera parcela llegó la Inmobiliaria.
—¿Cuáles fueron sus primeras obras como promotor?
—El siguiente reto fue comenzar en el mundo de la inmobiliaria.
Creé una empresa llamada Sercosa con la que realicé varias promociones
en Zaragoza. También hicimos viviendas en Lérida. Tenía muchos
amigos comisionistas, que son los que me vendían solares y casas
para derribar. Uno de ellos me aconsejó que tenía que hacer algún
polígono, porque era una gran oportunidad, ya que iba a haber leyes,
por las que todos los transportistas y talleres que estaban en los bajos
de las viviendas iban a tener que salir fuera de las ciudades. Entonces,
compré un terreno muy grande en Utebo, de unas 20 hectáreas (me
reservé otras 20 con derecho a compra), para hacer un polígono
industrial e hice unas primeras naves, un restaurante, Asepeyo, y
urbanicé todo aquello.
—¿Cómo surge asociarse con la CAI y cuándo nace Eizasa?
—El proyecto del polígono El Águila era muy ambicioso y me
asocié con la CAI (Caja de Ahorros de la Inmaculada) para poder
financiarlo. Compramos las otras 20 hectáreas y lo hicimos a medias.
Creamos una sociedad, Irusa, para construir un centenar de naves
más. También compramos otro terreno e hicimos 250 viviendas en
Casetas. Llegó un momento en el que las cajas de ahorro no podían
participar en sociedades, me vendió las acciones y me quedé con toda
la sociedad. Cuando ya era toda mía la llamé Eizasa, en 1974, y seguí
construyendo solo, eso fue otro hito muy importante.
Miguel V. Carreras. De vocación, empresario109
—¿Y cómo fue eso de construir un hotel, El Águila, en 90 días?
—Cuando estábamos haciendo con Sercosa las naves de Utebo,
construimos un restaurante y dejé preparada una estructura arriba
por si el día de mañana pensábamos hacer un hotel. Aún no estaba
previsto que se hiciera, pero un día un amigo mío, compañero de
estudios, Carlos Fernández Guaza, que era uno de los pocos que
en los años 70 hablaba inglés a la perfección, me dijo que como se
instalaba en Figueruelas (Zaragoza) la General Motors (GM), iban a
venir 700 técnicos de Alemania para poner todo en marcha. Y él, que
se encargaba de estas gestiones para la multinacional, no sabía dónde
alojarlos. Había que construir un hotel en el menor tiempo posible.
A la mañana siguiente, me levanté con las ideas claras: podía hacer
ese hotel en tres meses. Firmé un contrato con la GM en el que me
comprometía a que estuviera listo en 90 días, con 50 habitaciones. Y
lo hice. Así nació mi primer hotel: El Águila. Pero allí no se podían
alojar todos, tan solo 100, así que como acabábamos de hacer las
viviendas de Casetas, destinamos ciento y pico pisos, que aún estaban
sin vender, para este personal. Los amueblé y, entre estas viviendas y
el hotel, pude alojar a cerca de 500 personas de las 700.
—Un contrato con GM por dos años para el hotel El Águila muy
ventajoso.
—Sí, el contrato era de utilización de El Águila durante dos años,
con el compromiso de que las habitaciones iban a estar ocupadas
por GM durante este tiempo. Pero al año empezaron a marcharse
algunos trabajadores y otros querían alojarse en el centro, y GM me
propuso rescindir el contrato. Llegamos a un acuerdo y no rompieron
el compromiso. Les rebajé el precio y, a cambio, me permitían ocupar
las habitaciones que ellos no utilizasen, comunicando con la debida
antelación qué necesitaban. Eso propició que el hotel tuviera doble
ocupación y así se realizó mejor la amortización. Con GM aún continuó
la relación un tiempo más y le alquilé una de las naves del polígono, de
10.000 m2
, cerca del Hotel El Águila, durante varios años.
Ana Rioja Jiménez110
—¿Y en Lérida, construyó otro polígono industrial?
—Sí. Surgió la oportunidad de realizar este polígono. Hice una
sociedad con tres personas y nació Alcarrás. Compramos 50 hectáreas
de terreno. Por esas vicisitudes de las empresas, al final me quedé
solo con la sociedad. Es un polígono que ha funcionado muy bien y
que a día de hoy todavía está en activo y todo completo. En Lérida
he construido mucho. Junto a otros socios, creamos también Ayssa
e hicimos chalés en Lérida. Empezamos con 50 con esta sociedad.
Se disolvió y yo hice otros tantos ya con Eizasa, además de otras 80
viviendas. También edifiqué un hotel: el Hotel Real de Lleida, situado
en pleno centro comercial y de negocios, junto al conjunto histórico-
artístico de la ciudad; y, años más tarde, Apartamentos Real Lleida,
un bloque nuevo de 29 apartamentos, algunos dúplex, totalmente
equipados. Y seguimos promoviendo viviendas, acabamos de entregar
44 en Lérida y hemos terminado 28 en Fraga.
—¿Cómo se desarrolló la cadena de hoteles de Eizasa?
—Tras la experiencia de El Águila, pensé que esto de los hoteles no
era tan difícil, e hice otro hotel en Jaca, Hotel & SPA Real Jaca, en un
enclave perfecto para relajarse y disfrutar del maravilloso paisaje del
Pirineo; luego el hotel de Lérida ya mencionado. Después, Los Girasoles,
un apartahotel en el centro de Zaragoza del que estoy muy orgulloso;
luego 60 apartamentos turísticos en el valle de Ezcaray (La Rioja); los
apartamentos de Lérida, y finalmente el de la Ciudad del Transporte.
Además, compré dos más en el Pirineo, que reformé completamente:
el Hotel Real Villa Anayet, en Canfranc Estación; y Hotel & SPA Real
Badaguás. Tenemos actualmente nueve establecimientos hoteleros.
—Usted ha manifestado que le gustaría llegar a los 12 o 15 hoteles.
—Mi intención era llegar a tener una cadena con 12 o 15 hoteles,
porque ya tenemos toda la estructura. Tenemos los sistemas, la
administración, lo último en informática y conocemos el negocio.
Quería comprar algún hotel más para reformarlo e incluirlo en la
Miguel V. Carreras. De vocación, empresario111
El Hotel El Águila se construyó en 90 días,
por un compromiso de Miguel Carreras con General Motors.
El Apartahotel Los Girasoles es una de las grandes obras de Eizasa.
Ubicado en el centro de Zaragoza, posee unas maravillosas vistas desde su terraza.
Ana Rioja Jiménez112
cadena, pero no se presentan buenas oportunidades para invertir.
Con eso de que España es el país en el que más sube el turismo,
los hoteles están muy demandados. Aún estamos pendientes, tanto
en playa, como en grandes ciudades. Nos gustaría tener hoteles en
Madrid, en Barcelona, en Málaga... La intención es esa.
—¿Por qué ese afán por los hoteles?
—Los hoteles son difíciles de gestionar, sobre todo los nuestros,
que están en Lérida, en el Pirineo, en Zaragoza y en La Rioja. Pero
los proyectos difíciles son los que más te enganchan, más tienes que
pensar e inventar. Las cosas que van bien, van solas. Los negocios
que cuestan más, les dedicas más tiempo. No están en zonas donde
más se cobra por habitación o más ocupación hay, sobre todo los de
Zaragoza, ciudad en la que hay un exceso de oferta. Son pocos los días
que llenamos. En un Barcelona, Madrid o Málaga, la mayor parte de
los días están al completo, y eso que estamos defendiendo los hoteles
y sacando buenos resultados con diversas promociones y paquetes.
Y, además, porque los hoteles los gestionamos totalmente nosotros.
Cada hotel tiene su director, pero la dirección general de todos ellos
está en Eizasa y llevamos todo desde aquí: el centro de compras para
todos los establecimientos, la publicidad, el marketing, precios,
personal, promociones… Dirigimos nosotros todos los hoteles y los
restaurantes, lo mismo.
—Son difíciles de gestionar, pero dan mucho juego para el personal.
—Los hoteles dan juego a las empresas, son una herramienta que
favorece al personal. Damos movimiento a los hoteles y beneficiamos
a nuestros trabajadores de todo el Grupo de empresas, ya que tienen
un descuento muy importante en nuestros establecimientos. Les va
bien a las empresas tener este detalle con sus empleados, y por eso, me
gustaría tener alguno en la playa para que fuera nuestro personal, más
para mi gente que para mí. Además, la Fundación envía a los hoteles
a la gente que está necesitada y a jubilados los fines de semana.
Miguel V. Carreras. De vocación, empresario113
Eizasa posee cuatro hoteles de montaña: en el Pirineo aragonés:
Canfranc (en la foto), Jaca y Badaguás; y en Ezcaray (La Rioja).
Los hoteles de Eizasa ofrecen una gran variedad de servicios, desde sus restaurantes a sus
circuitos de spa. En la fotografía: Hotel Golf & SPA Real Jaca Badaguás.
Ana Rioja Jiménez114
—Y la restauración, ¿cómo surgió?
—Tenemos restaurante en todos los hoteles y los llevamos nosotros,
además de uno que tenemos en Plaza Salamero, 13 (Zaragoza). El
tener hoteles de cuatro estrellas nos obliga a que tengan restaurante
y no queremos dejarlos en manos de concesionarios, porque hay que
llevarlos muy bien, ya que es imagen para el hotel. Nosotros tratamos
de no perjudicar al hotel. No hemos querido subcontratar ninguno,
porque los restaurantes son muy complicados, muy difíciles de llevar.
Es una actividad muy personal. Los que van bien son los familiares,
con los propietarios al frente del negocio; o aquellos que dan cientos
de menús al día. En los restaurantes de los hoteles no se gana dinero,
pero se da servicio.
—Y Eizasa tampoco ha parado de construir viviendas como inmobi-
liaria.
—Simultáneamente a todo ello, y al mismo tiempo que crecían
todas las empresas del Grupo, conforme pasaban los años cogimos
velocidad como inmobiliaria. Teníamos obras en cinco provincias.
Estábamos construyendo en Tarragona, Lérida, Huesca, Logroño
y Zaragoza. He hecho cerca de 4.000 viviendas y aún tengo suelo
para otras 1.000. Eizasa Grupo Inmobiliario es una de las empresas
promotoras importantes de Aragón y del Corredor del Ebro, siendo
su crecimiento uno de los más significativos y sólidos de los últimos
años.
—¿Cómo definiría a Eizasa?
—Eizasa tiene casi 50 años de existencia y no ha tenido nunca
problemas de demandas de clientes y es una empresa sólida. Nosotros
decimos «con garantía Eizasa», porque es así, construimos bien, con
buenos materiales y acabados en viviendas variadas, cumpliendo
todos nuestros compromisos y garantías. Hemos realizado Viviendas
de Protección Oficial (VPO), construimos en barrios, en pueblos,
en ciudades, vivienda barata, cara, chalés, adosados… Nos diferencia
Miguel V. Carreras. De vocación, empresario115
nuestra preocupación por promover viviendas a la medida de cada
lugar, de cada familia…, la calidad en la construcción y el respeto por
el entorno.
Eizasa cubre todo el círculo inmobiliario desde la propia empresa,
abarcando suelos y los estudios de mercado, su urbanización,
la preparación de proyectos por parte de la oficina técnica, la
construcción, la comercialización y el servicio de postventa, todo ello
cuidado hasta el mínimo detalle por nuestros profesionales.
—¿Cómo le afectó la crisis del año 2008?
—En ese tiempo tenía 14 promociones en curso, pero las saqué
todas adelante. En 2008, más del 90 % de las inmobiliarias de
España, entre ellas las más grandes, desaparecieron. En Zaragoza
quedamos 4 o 5 fuertes. Los pisos bajaron un 45 %, muy por debajo
del coste de la construcción. Terrenos que se habían comprado con
ánimo de urbanizarlos y desarrollarlos, al producirse el parón, no se
pudieron realizar. La banca se llevó aquellos terrenos, que cuando
pidieron préstamos valían 100 y cuando se los quitaron valían 20.
Se encontraron con que aún debían 80, y tuvieron que pagar con su
patrimonio. Se arruinaron muchos. Eso fue muy duro. A mí no me
tocó porque no fui ambicioso. Iba con moderación. Y eso me salvó.
—¿De qué obra está más orgulloso?
—Respecto a establecimientos hoteleros, estoy muy orgulloso de
Los Girasoles. Era un suelo tan irrepetible que no quería venderlo.
Me lo quedé, quería hacer algo polivalente e hice un apartahotel.
—¿Qué promociones tiene en marcha? ¿Algún proyecto más en mente?
—Ahora tenemos tres promociones en marcha para vender:
en Fraga, avenida Valencia (Zaragoza) y en La Puebla de Alfindén
(adosados), y tengo dos proyectos para empezar, pero los voy a tener
que parar. Entre la pandemia, que el sector no se ha reactivado, y ahora
con los intereses tan elevados y el problema de los alquileres, hemos
Ana Rioja Jiménez116
Fachada de las oficinas de Eizasa junto al Apartahotel Los Girasoles en Zaragoza
(avenida César Augusto).
Miguel Carreras y su esposa, con parte del equipo de Eizasa.
Miguel V. Carreras. De vocación, empresario117
Eizasa ha construido cerca de 4.000 viviendas en cinco provincias, en ciudades,
pueblos, barrios…. promociones a medida de cada lugar.
En las fotografías: edificio avenida Valencia, apartamentos de San Juan de Plan,
apartamentos en Salou, edificio de la calle Heroísmo y en Casetas.
Ana Rioja Jiménez118
echado un poco el freno. En general, en España en estos momentos
ha habido un frenazo en la construcción. La política de alquileres,
limitando y poniendo problemas al propietario, está propiciando
que no se compren pisos para alquilar. Esto, unido al problema de
los «ocupas», hace que haya muy pocos pisos de alquiler y suban los
precios. Una parte de los pisos que se hacen nuevos, los compran
personas que quieren invertir en una segunda vivienda para alquilarla,
por ejemplo, y mejorar su situación económica en la jubilación. Es
un complemento a la pensión para vivir más holgadamente y luego
un patrimonio para dejar a los hijos. Y para eso se vendía mucho,
pero ahora no.
La construcción, el turismo y las fábricas de automóvil son las
tres locomotoras más fuertes de este país. Que frene la construcción
conlleva un parón general de la actividad en España, porque la
construcción acoge todo el abanico de actividades. Si no se construye,
se paraliza la producción y la actividad económica.
—¿De todas las empresas del Grupo, es Eizasa la que más riesgo con-
lleva?
—Sin duda. Para este negocio, como para el resto de empresas,
es muy importante la inversión. Yo he reinvertido siempre todos los
beneficios, pero no solo el 100 % sino hasta el 300 %. No le tengo
apego al dinero, me gusta el riesgo y toda mi vida he puesto en valor
la garantía que he transmitido a las entidades bancarias. He hecho
mucho uso del apalancamiento con talento, con cuidado y con la
seguridad de que siempre he tenido una salida para todo. He debido
muchos millones, pero siempre he dormido, nunca me ha quitado el
sueño.
Sin embargo, las inversiones fuertes en el Grupo Logístico Carreras
nunca me han dado miedo, porque galopaba. Íbamos ampliando
el 10, el 20, el 30 % cada año y he comprado camiones nuevos y
nunca me quitó el sueño No había riesgo financiero. Sin embargo,
Miguel V. Carreras. De vocación, empresario119
en la inmobiliaria es todo más inseguro. La inmobiliaria es lo más
peligroso. Si no se vende, hay que devolver el dinero con la reserva.
Una vez le pregunté al director general de un banco: «Cuál crees
tú que es el tanto por ciento ideal de apalancamiento». «Eso es difícil
de decir, pero, como sé que tú llevas en la cabeza todo y eres muy
responsable, el día que te vayas a la cama y no duermas, te has pasado»,
me contestó.
—¿Y le ha ocurrido alguna vez? ¿Lo que debía le ha quitado el sueño?
—Me ocurrió solo una vez. Quería comprar un solar en el Actur
para 100 viviendas, pero un amigo me dijo que me lo quedara todo y
era casi para 300. Me convenció y me dijo que ya se lo pagaría en 2 o
3 años. Lo compré. Empezaron a vencer las letras, y eran de grandes
cantidades. Me desbordó. No había calculado bien aquella operación.
Empecé con las 100 primeras viviendas y me vi muy apurado. Una
noche tuve una experiencia importante, me asaltó la angustia con
la sensación de responsabilidad de que no iba a poder cumplir. Me
levanté y dije: «Esto se ha acabado». Puse en venta las otras partes y le
pagué al contado. Tan solo esa vez no he podido dormir y he tenido la
sensación de que me había pasado. Otras veces me desvelaba cuando
no me apalancaba suficientemente.
—¿Qué hijos han trabajado o trabajan en Eizasa?
—Carmen, que estudió Derecho, trabajó un tiempo como abogada
en Eizasa, pero ahora lleva la Asesoría Jurídica del Grupo Logístico
Carreras y dirige la Fundación.
Mi hijo Jesús trabaja en Eizasa. Hizo Económicas y pasó por El
Águila para aprender todo el tema de la gestión de hoteles. Después
vino a Eizasa para trabajar en promoción y construcción. Y luego, bajo
su dirección, creamos una empresa, Gestión Integral de Marketing
Hotelero, cuyo principal cometido son las reservas y el call center.
Ahora quiero que se incorpore de nuevo a la inmobiliaria para que
Ana Rioja Jiménez120
se responsabilice del desarrollo de los proyectos, de todo lo que es
promoción, porque esta empresa es muy importante.
Miguel Ángel, que es arquitecto, ha trabajado con Eizasa en
diferentes etapas de su trayectoria profesional, como jefe del
Departamento Técnico de Arquitectura, de Aparejadores y director de
las obras. Desde hace unos años dirige las dos empresas que tenemos
de explotación agrícola.

Miguel V. Carreras. De vocación, empresario121
CAPÍTULO VII
Explotaciones agrícolas
El amor a la tierra y el deseo de un grifo
Miguel Carreras compró su primera finca de regadío en 1983.
Hoy, cuarenta años después, sueña con una tierra llena de olivos
para hacer un buen aceite. Aún falta mucho para ello, quizá más
de diez años, pero este empresario, que disfruta con los grandes
retos, ha recorrido ya ese camino con su imaginación. Como el
guía de su caravana, ha ido y ha vuelto para contarnos que el
Grupo Carreras aún tiene que realizar su gran proyecto: producir
y comercializar millones de litros de aceite de oliva y venderlos
por todo el mundo.
El Grupo Carreras cuenta en la actualidad con dos explotaciones
agrarias: OMPRIO en Peñalba y pueblos cercanos, con 500 hectáreas
de secano de las cuales ha pasado 100 a regadío; y SAT 5007, en
Sariñena y alrededores, una sociedad que posee cerca de 800 hectáreas
de regadío.
—En su infancia, los agricultores de su tierra, Los Monegros, vivían
siempre mirando al cielo. ¿Tanto le marcó esa escasez de agua?
—Mi abuelo y mi padre me inculcaron el amor a la tierra. En
Peñalba teníamos unas tierras de secano y nos las cultivaban. En
verano, los chicos apoyábamos esa agricultura como podíamos.
Siempre se esperaba la gran cosecha que nunca llegaba porque la
escasez de agua lo marchitaba todo. En Los Monegros la falta de
agua ha sido una constante. Por ello, mi ilusión ha sido siempre tener
tierras de regadío.
Ana Rioja Jiménez122
—¿Cuántas explotaciones agrarias tiene en la actualidad?
—De mi familia heredé una sociedad, OMPRIO, con tierras
en Peñalba y algún pueblo de alrededor (Ontiñena, Perdiguera, La
Almolda, Bujaraloz, Candasnos…), en la que están todas las tierras
de secano de mi padre y de mis tíos. Hay 500 hectáreas y 100 las
hemos pasado a regadío. Cultivamos, sobre todo, cebada. Y luego está
SAT 5007, cuyo inicio fue una finca matriz de Sariñena, que había
conseguido una toma de agua del canal solo para sus campos, la M51.
Esa finca grande, que tenía unas 700 hectáreas, se dividió en tres por
diversas circunstancias y las he ido comprando poco a poco. Cuarenta
años me ha costado tener su totalidad y reunirla otra vez.
—Y ese grifo que usted tanto deseaba se acabó convirtiendo en una
fértil tierra de regadío en Sariñena.
—Sí. Todo comenzó en 1983, cuando estaba haciendo con Eizasa
unas viviendas en la calle Sáinz de Varanda. Tenía unos pisos y unos
locales que no vendía y entonces los cambié por un tercio de esa finca
matriz, unas 200 hectáreas. Solo tenía 60 de regadío, con ruedas y
primitivo. Recuerdo que al día siguiente nos marchábamos de viaje
por el 25º aniversario de nuestra boda. Casi no vi la finca, pero me
ilusionaba mucho. Luego compré el segundo tercio y luego la tercera
parte. Este año he comprado un añadido de la primera de 120 hectáreas,
con lo cual ya tengo toda la finca completa. Y toda ella la estamos
pasando a regadío, ya que la toma M51 es solo para sus tierras.
—¿Qué usos le ha dado a esta segunda sociedad, a SAT 5007?
—Es una explotación agraria —cebada, maíz, alfalfa, guisantes,
girasol, árboles frutales y huerta— y fue también ganadera porque
la compré con 100 vacas retintas andaluzas de carne, pero eran
peligrosas y me las quité porque un día casi matan al encargado.
Entonces compré ovejas, hasta 3.000 llegamos a tener. Daban 1,3
corderos por oveja. Casi 4.000 corderos cada año. Pero también me
las quité porque no encontrábamos pastores.
Miguel V. Carreras. De vocación, empresario123
Miguel Carreras ama la naturaleza y el campo desde que era niño.
En esta imagen, recorriendo una de sus fincas.
Ana Rioja Jiménez124
Además, en esas 200 primeras hectáreas, que se podían hacer de
regadío, pero que tenían mucha sal, tuve que tener varios años arroz,
allanarla, hacer terrazas y tapar barrancos. A los cinco años, ya se había
lavado la sal, y entonces tuve que deshacer las terrazas y comenzar a
poner pivots, aspersores fijos y hacer cientos de kilómetros de tubería
para poner el regadío. En total ahora tiene 700 hectáreas de regadío.
He ido transformando y mejorando toda la finca en un regadío
moderno, todo por ordenador, haciendo balsas de riego dentro de la
finca, mejorándola y sin dividir nada.
—¿Cómo acogió su familia su dedicación a la tierra?
—Muy bien, porque cuando compré la primera parte de la finca
había un chalé con piscina que arreglamos y los veranos nos íbamos
allí con los chicos. Nos gustaba mucho. Todos han trabajado en el
campo. Yo hice el camión, pero mis hijos han pasado todos por la
finca a trabajar en los veranos. Les fue bien. Únicamente mi hijo
Miguel Ángel continuó con ese amor por la tierra, que es quien dirige
las dos sociedades.
—Su ilusión por el regadío le ha llevado a seguir comprando fincas.
—Además de la de Sariñena, compré una finca en San Juan de
Flumen de 110 hectáreas de regadío. También compré otra de 300
hectáreas en Velilla de Cinca, de las cuales hay 70 hectáreas de regadío
y el resto de secano y monte. Y otras 40 hectáreas más en Chalamera,
de huerta (regadío). Con todas estas tierras, ahora estamos en una
posición muy fuerte de hectáreas de regadío. Estaremos en cerca de
1.000 hectáreas. Tenemos una ilusión muy grande, porque mis padres
tenían agricultura en secano, pero no de regadío.
—En estas fincas han realizado en los últimos tres años una gran
apuesta por los almendros.
Miguel V. Carreras. De vocación, empresario125
—Sí. Tenemos 200 hectáreas plantadas de almendros con 250.000
árboles y este año vamos a recoger por primera vez una buena
producción de almendras.
—Y ahora hay otro proyecto en marcha: ¿plantar olivos?
—Los almendros los dejamos ya en esas 200 hectáreas ya plantadas,
no vamos a poner más, y hemos comenzado con los olivos. Este año
hemos puesto oliveras en 30 hectáreas de regadío, unos 40.000 árboles,
y también en dos hectáreas de secano para ver cómo se comportan.
De cara al futuro, esta explotación será de olivos y almendros. Vamos
a transformar todo en arbolado. Y en lugar de frutales, pondremos
olivos. En estas tierras también hacemos doble cosecha cada año:
cebada y maíz o guisante y maíz, pero hay poca agua. Para hacer
eso necesitamos 10.000 metros cúbicos por hectárea y este año nos
han dado solo 4.000. Ya no se puede hace doble cosecha y baja el
rendimiento de la finca. Con 4.000 es suficiente para el olivar. Y,
como a la larga, va a escasear el agua, hay que replantear el cultivo y
apostar por los menos exigentes de agua. Hay que transformarlas en
plantación de olivo, porque es un árbol muy resistente. Por ejemplo,
este año va a ser la primera vez que cogemos cosecha de almendras,
porque otros años se han helado, pero el olivo no se hiela. Hay que
jugar con la seguridad, no con el seguro.
—¿Y quiere plantar un millón de olivos?
—Y más. Las cerca de 800 hectáreas que quedan de regadío las
quiero plantar todas de olivos, con lo cual, como ponemos sobre
1.600 olivos en cada hectárea, podríamos llegar a cerca del millón y
medio de olivos. Si es poco para el proyecto que llevamos en mente,
compraremos más hectáreas de regadío para meter dos millones de
olivos para lograr una buena producción de aceite.
De momento, ya está analizada la tierra y preparada para el año
que viene, que plantaremos unos 200.000 olivos y ya los hemos
encargado al vivero. Mientras, también estamos experimentando en
Ana Rioja Jiménez126
La familia Carreras ha realizado una gran apuesta por los almendros, con la plantación
de 200 hectáreas de estos árboles.
Miguel V. Carreras. De vocación, empresario127
Plantación de olivos, uno de los grandes proyectos
de las sociedades agrícolas del Grupo.
Ana Rioja Jiménez128
secano con unas variedades adecuadas y una forma de plantarlas para
que puedan resistir la falta de agua en Los Monegros. Posteriormente,
iremos plantando más. Es un proyecto muy importante. Vamos a
empezar con los olivos para hacer aceite y fabricarlo en Aragón para
el mundo. En 4 o 5 años tendremos olivas. Hacer el aceite es sencillo.
Habrá que poner una almazara o dos, según el volumen, embotellar
el aceite y comercializarlo con un buen marketing y una adecuada
logística. También queremos construir un hotel para temporeros, un
laboratorio, almacenes, etc. Eso será una gozada. Yo no lo veré, pero
ahí está el proyecto. Lo liderará y hará realidad mi hijo, que está tan
ilusionado como yo, con un buen equipo de profesionales.
—¿Plantar olivos en regadío y en secano?
—Todas esas tierras, tanto las de regadío como las de secano,
hay que transformarlas en plantación de olivo, porque se puede dar
hasta en secano. Estamos estudiando todo el tema de variedades con
distintas épocas de madurez. La planta no es cara, lo más costoso
es el riego. Estaremos tres o cuatro años invirtiendo sin producir,
abonando, cuidando, podando, pero estamos dispuestos y preparados
para financiar este proyecto.
—¿Sueña con que el apellido Carreras vaya asociado a ser un buen
productor de aceite de España?
—Claro. Si fuera más joven, vería cómo nos convertimos en
poco más de diez años en grandes productores de aceite. Ya tengo el
nombre, que será «El Molino», como una de las fincas. Y lo haría con
toda facilidad. Tengo esa ambición de que en unos 10 años o poco
más, podamos hacer, al menos, cuatro millones de litros de aceite,
comercializarlos en botellas de un litro, medio litro, cien decilitros, así
como aceitunas de mesa, y venderlos, pero bien vendidos, en Europa.
Yo no lo veré, pero mis hijos y mis nietos, sí. Y, junto a mi hijo, quiero
dejarlo todo muy bien encarrilado. Esta empresa está por crear, pero
Miguel V. Carreras. De vocación, empresario129
tenemos el fundamento, que es la tierra, y reunir todas esas hectáreas
juntas de regadío no es fácil.
Es uno de mis sueños. Respecto a mis empresas, tengo tres ilusio-
nes: que Carreras Grupo Logístico sea en Francia lo que es en España;
tener más hoteles y construir nuevas viviendas en las que la gente sea
feliz y diga: «Qué bien construyen estos de Eizasa»; y vender nuestro
aceite por todo el mundo.
Esas hectáreas de regadío, que Miguel Carreras ha ido reuniendo
a lo largo de 40 años, estaban destinadas a un gran proyecto:
producir olivas que den un un buen aceite y comercializarlo
internacionalmente. Desde Los Monegros, esa tierra árida y seca
que marcó su infancia y su carácter y que él se empeñó en ponerle
grifos para que ya nunca se muriera de sed, al mundo.


Miguel V. Carreras. De vocación, empresario131
CAPÍTULO VIII
La Fundación Carreras y otras obras sociales
Devolver a la sociedad parte de lo que se recibe de ella
Una de las grandes satisfacciones de Miguel Carreras y de sus hijos
es haber creado la Fundación que lleva su nombre: Fundación
Miguel Carreras Calvete e Hijos, que nació en 2019 y aumenta la
Responsabilidad Social Corporativa del Grupo Empresarial. Con
ella, Miguel Carreras quiere hacer partícipe a la sociedad de una parte
de los beneficios de este holding familiar, aportar un capital para
el bien común, sin ningún tipo de finalidad lucrativa, trabajando
de forma desinteresada para impulsar la solidaridad y el bienestar
social.
Con la Fundación quiere agradecerle el éxito y compartir la buena
marcha de sus empresas con la comunidad que ha formado parte
del desarrollo de esta actividad económica: personal, clientes,
proveedores, últimos destinatarios… «No se puede dejar a nadie
en el camino, porque todos somos un equipo —afirma Miguel
Carreras—. Se trata de aportar un capital para el bien común». La
Fundación está presidida por Miguel Carreras y son patronos de la
misma su esposa, María Natividad Lario, y sus seis hijos: Cristina,
Miguel Ángel, Fernando, Carmen (que lleva la dirección), Jesús y
Ricardo.
—¿Cómo surge el proyecto de crear la Fundación Carreras?
—Empecé de auxiliar de conductor con un camión viejo y, aparte
de mi actividad y méritos, entiendo que he tenido suerte y que parte
de mi éxito se lo debo a mis clientes, a mis empleados y a todas
Ana Rioja Jiménez132
estas personas anónimas que han hecho que mis empresas crecieran,
solicitando los productos que mis clientes vendían, a la demanda
indirecta de esas personas. Mediante la Fundación quiero devolverle
a la sociedad una parte de lo que me ha dado.
—¿Cuáles son los fines de la Fundación?
—Somos una Fundación de implantación nacional inscrita en el
Registro Nacional de Fundaciones, financiada por el Grupo Carreras
y que tiene como fines, entre otros, devolver a la sociedad parte de
lo que se recibe de ella, especialmente buscando el apoyo a personas
vulnerables, a la educación y a la salud de los trabajadores. A su vez
ponemos en práctica diversas acciones sociales y culturales, poniendo
en valor el origen de la empresa y la actividad agrícola en el desierto
de Los Monegros, especialmente en La Almolda.
Su misión es fomentar y realizar todo tipo de obras de interés
social, promoviendo el progreso económico, la justicia e inclusión
social, la difusión y promoción de la cultura y el conocimiento,
y, de modo particular, el impulso y desarrollo de actividades
relacionadas con el sector logístico y de transporte. Todos los que
formamos parte del Grupo Carreras podemos sentirnos parte de
esta iniciativa y orgullosos con el trabajo que la Fundación realiza
hasta ahora.
—¿Cómo se financia la Fundación?
—La Fundación recibe una parte de los beneficios del Grupo
Carreras. Cuando creé estas empresas, en lo que pensaba primero
es en que mis hijos no tuvieran que salir fuera de Aragón a trabajar;
y segundo, que trabajaran en empresas que fueran suyas. En esa
generación de los hijos, lo he conseguido. Están trabajando todos
en nuestras empresas. Y ahora pienso en mis nietos. Al diversificar
el Grupo en tantas empresas y actividades, todas rentables, es un
potencial muy importante para que tengan la oportunidad de
trabajar también en su casa. No necesitarán hacer efectivos los
Miguel V. Carreras. De vocación, empresario133
La Fundación Miguel Carreras Calvete e Hijos nació en 2019.
La Fundación Carreras otorga becas de estudio en todas las etapas educativas.
Ana Rioja Jiménez134
patrimonios que ellas conllevan, ni podrán vender. En ese caso,
he pensado que esta Fundación reciba una participación de mis
empresas, que quiero que cada año sea más importante, porque será
el proyecto del que todos estaremos más orgullosos: llevar bienestar
a la gente necesitada.
—Usted afirma que la Fundación es un complemento de la empresa
para sus trabajadores.
—Uno de los fines de la Fundación es ser un complemento para
atender a las familias de nuestros trabajadores en circunstancias
especiales. Ayuda y apoya a los empleados en momentos críticos de sus
vidas por cuestiones de salud (operaciones importantes, tratamientos
especiales), educativas (con apoyo escolar y becas) o separaciones,
viudedad… En 3.000 trabajadores surgen muchas problemáticas de
este tipo. Nuestra política como empresa familiar es ayudar a nuestros
trabajadores y a sus familias, a los de casa. Nuestro personal valora y
agradece mucho la Fundación.
—¿Qué proyectos está desarrollando la Fundación?
—Está realizando acciones educativas, sociales, culturales, de salud
y bienestar, y organiza y otorga el Premio José Luis Carreras Lario a la
Innovación y Sostenibilidad en Logística. Pero aún volará más alto y
desarrollará proyectos de más alto nivel.
—¿Uno de los más importantes es el Premio José Luis Carreras Lario
a la Innovación y Sostenibilidad en Logística?
—Sí, ya se han convocado dos ediciones. Es un premio en
homenaje a mi hijo José Luis, fallecido en 2020, que destacó por
su esfuerzo e implicación en el sector del transporte y la logística
como consejero delegado de Carreras Grupo Logístico, así como en
el ámbito empresarial aragonés. Este premio reconoce y pone en valor
los trabajos de investigación realizados en los estudios universitarios
en España.
Miguel V. Carreras. De vocación, empresario135
—Destaque alguna de las actividades educativas.
—En acciones educativas, otorgamos becas a nivel nacional y en
todos los ciclos formativos. Además, existen becas de desplazamiento
para estudiantes que cursen estudios superiores y que se vean en la
obligación de trasladarse de su domicilio para continuar su educación.
También se dan becas a la Formación Profesional (FP) en Aragón
y tenemos proyectos de ayuda a estos centros, así como contra el
abandono escolar.
Desde el año pasado ofrecemos una línea de apoyo emocional
inmediata para niños y jóvenes de diversos centros educativos en
Aragón. Queremos hacerla extensiva a más centros este año, porque
es un tema muy importante. Patrocinamos y participamos en premios
que buscan conocer ideas frescas e innovadoras de la mano de los más
jóvenes; estamos creando un espacio de coworking para favorecer el
emprendimiento y ayudar a la igualdad de oportunidades. También
fomentamos el deporte con incentivos solidarios, organizamos talleres
culturales —llevamos tres años con el concurso de pintura—, y de
cocina. En definitiva, queremos abrir ventanas a los jóvenes.
—¿Y en acciones sociales y culturales?
—Un proyecto muy importante se centrará en nuestros mayores.
Hemos empezado un programa piloto de actividades relacionadas
con la música dentro de las residencias, que propicia que se conozcan
y disfruten juntos residentes y familiares. Ha sido espectacular, queda
mucho por hacer.
Desde hace tres años, tenemos un programa de reutilización de todos
los ordenadores del Grupo que se van sustituyendo, se limpian digital-
mente, se arreglan y se entregan a los que lo necesitan. Pura economía
circular, aprovechar los recursos, evitar residuos… muy sencillo.
Asimismo, se colabora con distintas entidades sociales y ONGs de
acuerdo a sus necesidades. Una de las más satisfactorias es la donación
mensual al Banco de Alimentos de Aragón. En 2022 se donaron más
Ana Rioja Jiménez136
La colaboración de Carreras durante la pandemia
En los meses que duró el confinamiento en España, a causa
de la pandemia provocada por la Covid-19, surgieron muchas
necesidades en diversos ámbitos de la vida en 2020. Por ello,
desde Carreras Grupo Logístico y desde la Fundación quisimos
colaborar para atenuar al máximo sus efectos. Nuestra transitaria,
Carreras Forwarders, abrió una ruta aérea de importación de
material sanitario y gestionó con éxito el transporte en avión de
3.300.400 mascarillas desde China a España.
También se realizó la gestión altruista del centro logístico de
almacenaje y distribución de materiales contra la Covid-19.
Carreras colaboró en la recogida y transporte de los materiales,
cedió espacio en su nave de PlaZa; y gestionó pro bono la
manipulación y preparación de mascarillas, guantes y otros
materiales para su posterior distribución a hospitales, centros
de salud, residencias e instituciones, en colaboración con
Protección Civil. Asimismo, cedimos algunas de nuestras
plataformas logísticas a la Unidad Militar de Emergencia
(UME) para cargar y transportar material en diferentes muelles
logísticos. El aeropuerto de Zaragoza es un referente a nivel
La transitaria, Carreras Forwarders, abrió una ruta aérea de importación
de material sanitario desde China a España.
Miguel V. Carreras. De vocación, empresario137
de 20.000 kilos. Asimismo, desde la creación de la Fundación se
realiza un mercadillo benéfico y se ha avanzado mucho para evitar «el
desperdicio alimenticio». También patrocinamos eventos en apoyo de
entidades que luchan contra enfermedades diversas.
—¿Un programa del que está muy satisfecho es el de Geni@s para
jubilados?
—Integran Geni@s todas aquellas personas que han formado
parte de nuestra empresa familiar Carreras Grupo Logístico a lo
largo de su vida profesional. Cuando una persona se jubila pasa a
formar parte de este Grupo. La Fundación Carreras organiza para
ellos encuentros, comidas, talleres, concursos, estancias en nuestros
hoteles… Les mostramos así nuestro agradecimiento, porque nos han
ayudado a consolidarnos a lo largo de los años y a llegar a ser el Grupo
empresarial que somos hoy en día.
de transporte de mercancías, hecho que facilitó la llegada al
mismo de gran cantidad de material sanitario. Grupo Carreras
albergó la logística del material sanitario en Aragón durante el
Estado de Alarma.
Desde la Fundación Carreras se colaboró con el Gobierno de
Aragón y el Ayuntamiento de Zaragoza con la donación de
200.000 euros para la compra y suministro de material sanitario.
A su vez, se colaboró con el Ayuntamiento de Zaragoza en
la apertura de líneas de comida. Se donaron alrededor de
3.000 kilos de alimentos y un total de 10.000 euros al Banco
de Alimentos. Además, desde el Grupo Carreras se sufragaron
los portes de gran cantidad de clientes de toda España que
quisieron realizar donación al Banco de Alimentos. También
se donaron un total de 7.000 euros destinados a la compra de
alimentación y sustento para el Refugio de Zaragoza, además
de sufragar los portes, entre otras colaboraciones.
Ana Rioja Jiménez138
—¿Y el Museo Etnográfico de Los Monegros en La Almolda?
—Este es un proyecto muy hermoso, un Museo Etnográfico
ubicado en mi casa familiar y natal de La Almolda, impulsado por
la Fundación Carreras. Su objetivo principal es mostrar la forma de
vida de nuestros antecesores. Recopilamos herramientas, enseres,
documentos e imágenes que plasman la vida rural de Los Monegros
durante el siglo XX y en el que están colaborando numerosos vecinos
de La Almolda, rescatando y restaurando estos objetos. Contamos
con la labor desinteresada y muy generosa de los almoldanos, sin
la cual no sería posible. Para ello se han organizado talleres de
restauración. Es un Museo realizado entre todos y para todos. El
proyecto se presentó el año pasado y como anticipo se estrenó el
documental «Recuerdos vivos», que reúne anécdotas y vivencias
contadas en primera persona para narrar la vida cotidiana que
imperaba hace un siglo en La Almolda. Asimismo, la Fundación
ha editado un libro de recetas tradicionales: «La mesa de antaño y
de siempre. Recetas almodanas», recogidas por los vecinos de esta
localidad.
—Por sus palabras, la Fundación tomará cada año más fuerza con
más proyectos. Hábleme de ellos.
—Sí. Mi idea es que la Fundación tenga cada vez un mayor
presupuesto, vía dividendos de todas las empresas, con lo cual podrá
acometer proyectos muy importantes. Serán proyectos de alto nivel.
Pensamos en hacer una residencia para personas mayores que hayan
sido transportistas, en un edificio de la Ciudad del Transporte que
queremos rehabilitar, porque antes estaba destinado a oficinas. Lo que
quiero es hacer cosas importantes en la Fundación como puede ser ese
centro y cursos para la formación de profesionales del transporte, así
como cuantas oportunidades se nos presenten de ayudar, hasta donde
podamos, a quienes nos necesiten. La Fundación volará más alto, hay
que proyectar y fijarse grandes objetivos y dotarla de personal para
dirigirlos.
Miguel V. Carreras. De vocación, empresario139
En la casa natal y familiar de Miguel Carreras, en La Almolda,
se está creando un Museo-Centro de Interpretación de la vida rural en el siglo XX.
Portada y contraportada del libro
«La mesa de antaño y de siempre. Recetas almoldanas», editado por la Fundación.
Ana Rioja Jiménez140
Ayuda humanitaria por la invasión de Ucrania
Ante la enorme crisis humanitaria provocada por la invasión
de Ucrania por parte de Rusia, la Fundación Carreras colaboró
en 2022 en el envío de ayuda humanitaria para la población,
donando la gestión logística en nuestras instalaciones de las
mercancías que fueron enviadas a Ucrania por el Gobierno
de Aragón de forma centralizada. En Carreras realizamos las
tareas de recepción de material, gestión y organización de
las cajas recogidas, el almacenamiento temporal, la carga y
la descarga de todos los vehículos que acudieron a nuestras
instalaciones. Asimismo, Carreras, como operador logístico y
de transporte, envió un camión con material sanitario y ropa
al campo de refugiados ucranianos en Rumanía. También
colaboramos con la Cruz Roja en una campaña de juguetes
para los niños ucranianos residentes en Zaragoza.
La Fundación donó la gestión logística
de la ayuda humanitaria a Ucrania.
Miguel V. Carreras. De vocación, empresario141

Donación al Cabildo de La Palma
La Fundación Miguel Carreras y Carreras Grupo Logístico donamos
100.000 euros en diciembre de 2021 al Cabildo Insular de La Palma,
con el fin de ayudar a los afectados por la erupción volcánica. Con
esta aportación nos quisimos solidarizar con nuestros compatriotas
y hermanos canarios, especialmente los palmeros, perjudicados
por la desgracia.
Entrega de una donación al presidente
del Cabildo de La Palma.

Miguel V. Carreras. De vocación, empresario143
CAPÍTULO IX
Sostenibilidad y Medio Ambiente
Proyecto Carreras 0,0: la logística en verde
Miguel Carreras sostiene que en el ADN de su Grupo de empresas
hay varias constantes que lo definen: ser aragonés, pero con vocación
internacional, la innovación continua y la sostenibilidad; es decir,
el máximo cuidado por el medio ambiente. En este sentido, hay
que destacar que en 2011, en colaboración con la Universidad de
Zaragoza, creó la Cátedra Carreras de Sostenibilidad e Innovación
Logística. Toda una declaración de intenciones en su apuesta por la
sostenibilidad y la logística verde.
Para conseguir un mayor nivel de sostenibilidad en el área de la
logística y el transporte, mediante la reducción del impacto ambiental,
ha elaborado un Plan de Acción: Carreras 0,0. Esta estrategia de
sostenibilidad 2030 pretende compensar el 90 % de las emisiones
de CO2 para esta fecha, y se basa principalmente en los siguientes
pilares: construcción inteligente, apuesta por las energías renovables,
aumento de una flota propia inteligente y formación eficiente.
Carreras Grupo Logístico, que pertenece a la plataforma de
colaboración Lean & Green, ha logrado ya su primera Estrella por la
reducción de las emisiones derivadas de sus actividades logísticas y de
transporte, y ya está trabajando para conseguir la segunda.
—¿Cómo surge la Cátedra Carreras de Sostenibilidad e Innovación
Logística?
—En 2011 creamos, junto a la Universidad de Zaragoza, la
Cátedra Carreras de Sostenibilidad e Innovación Logística, enfocada
Ana Rioja Jiménez144
a potenciar la investigación y la difusión en el ámbito de la cadena
de suministro, la innovación y logística sostenible. Nace con la
finalidad de promover y organizar actividades que permitan reunir
a profesionales, personal de empresas, investigadores y estudiantes
con interés en adquirir conocimientos en el tema de la logística.
Desde su formación se han llevado a cabos varios eventos con
temáticas diferentes como la huella de carbono, la gran distribución,
las tendencias de la cadena de suministro y la innovación. Este 20
de septiembre de 2023, coincidiendo con nuestro 90 Aniversario, se
celebra el Noveno Encuentro en el Palacio de Congresos de Zaragoza
con el tema «Digitalización en el sector de la logística y transporte».
En este acto también se otorgará el II Premio José Luis Carreras Lario,
a la Innovación y Sostenibilidad en Logística. El tema de esta edición
es «Emprendimiento y desarrollo económico, tecnológico y digital en
el sector del transporte y logística».
—La preocupación por la sostenibilidad medioambiental le ha llevado
a trazar un ambicioso Plan de Acción: Carreras 0,0.
—A través del proyecto Carreras 0,0, trabajamos en la reducción
paulatina de las emisiones de CO 2 de los procesos logísticos en los
próximos años. Pero este proyecto posee una proyección mayor y está
en constante desarrollo hacia una mejora continua y sostenida a lo
largo de todos los años de actividad de la empresa. Se basa princi-
palmente en los siguientes pilares: construcción inteligente, apuesta
por las energías renovables, aumento de una flota propia inteligente
y formación eficiente.
—Construcción inteligente.
—Carreras se caracteriza por la calidad y sostenibilidad de
sus instalaciones, en búsqueda constante de nuevos métodos que
garanticen este compromiso, como el incremento de aislamiento de
las cubiertas de las naves para evitar fugas, la instalación de placas
fotovoltaicas en los techos de los edificios o la reutilización de aguas.
Miguel V. Carreras. De vocación, empresario145
La sostenibilidad y el máximo respeto por el medio ambiente guían a Carreras.
El Grupo Carreras fomenta el uso de las energías renovables.
Ana Rioja Jiménez146
—Apuesta por las energías renovables.
—La empresa fomenta el uso de las energías renovables solar y
eólica. Carreras cuenta con cerca de 300.000 m2 de placas solares en
las instalaciones del Grupo y hemos creado el parque eólico AGON
3, con tres aerogeneradores (que generan 9 megavatios) y se prevé la
construcción de otros tres.
—Aumento de una flota propia inteligente.
—Continuamos con nuestra estrategia de renovación de flota,
invirtiendo en sostenibilidad y seguridad de nuestros vehículos. A lo
largo de este último año, hemos comprado 95 nuevas tractoras, que
cuentan con la certificación Euro VI, e introducen nuevos elementos
que reducen la contaminación y las emisiones. La edad media de flota
actual es de cuatro años de antigüedad; cumpliendo en un 75 % de la
misma con la normativa Euro VI.
—Formación eficiente.
—En Carreras hemos implantado formaciones en técnicas de
conducción eficiente para transportistas y, de este modo, garantizar
el cumplimiento del Plan, así como mejorar la seguridad y elemento
preventivo. Haciendo uso de herramientas de medición se consigue
controlar y gestionar la información de los conductores y los vehículos.
Eficiencia de las rutas; reduciendo la distancia recorrida por los
vehículos en vacío, y creando sinergias con diferentes clientes con
un mismo vehículo. Formación continua por parte de proveedores
y de los responsables operativos, para una conducción eficiente en
su día a día. Análisis de asignación de vehículos; con el objetivo de,
conociendo el servicio y el peso del equipo que vamos a recoger,
asignar vehículos más sostenibles, sin dejar de ser operativos.
Seguimiento y análisis semanal de los datos que emite el sistema de
control como fuente de información, para posteriormente realizar
planes de acción concretos.
Miguel V. Carreras. De vocación, empresario147
Carreras ha iniciado con Repsol un proyecto piloto de uso de biocombustible.
Carreras impulsa un Centro de Producción de Hidrógeno en Masquefa (Barcelona).
Ana Rioja Jiménez148
—Un plan avalado por Lean & Green. ¿En qué consiste esta inicia-
tiva internacional?
—Lean & Green es una plataforma europea de colaboración cuyo
objetivo es ayudar a las empresas de todos los sectores a reducir sus
emisiones de gases de efecto invernadero derivados de las actividades
logísticas, y generar así una red de organizaciones líderes en el campo
de la logística sostenible.
—¿Qué supone para Carreras tener una estrella Lean & Green e ir
camino de otra?
—A finales del año pasado, conseguimos nuestra primera estrella
Lean & Green por alcanzar el objetivo de reducir un 24 % las emisiones
derivadas de muestras actividades logísticas y de transporte. Se trata
de un reconocimiento por los logros hacia una logística sostenible y
supone que vamos avanzando en nuestro camino para la reducción de
emisiones de gases de efecto invernadero, invirtiendo en innovación,
digitalización y sostenibilidad. La meta final es compensar todas
nuestras emisiones a medio plazo.
—Además, Carreras impulsa un Centro de Producción de Hidrógeno
para suministrar energía verde a sus procesos logísticos y de transporte.
—Este Centro de Producción de Hidrógeno estará en nuestras
instalaciones de Masquefa (Barcelona) y será capaz de suministrar
energía 100 % verde para 25 vehículos y 120 carretillas. Es un
proyecto pionero que nos permitirá disponer de energía procedente
de recursos renovables para mantener la actividad logística del
centro y también para su área de transporte. Este centro cuenta con
una instalación fotovoltaica de 5 megavatios que producirá energía
y, a través de una hidrogenera se conseguirá 720 kilos de hidrógeno
al día, suficientes para atender las necesidades de suministro de 120
carretillas y 25 vehículos de distribución capilar. Será el primero
de este tipo por parte de un operador logístico español. En este
proyecto contamos con el apoyo y la colaboración de diferentes
Miguel V. Carreras. De vocación, empresario149
empresas: SHiE (Arpa, equipos móviles de campaña y Zoilo Ríos),
Plug Power, Toyota Material Handling y Tevva.
—¿Qué más iniciativas de este tipo lidera Carreras?
—Hemos iniciado con Repsol un proyecto piloto de uso de
biocombustibles que permitirá la reducción de hasta un 90 % de
las emisiones respecto al gasóleo convencional. La utilización de
este combustible renovable se realizará en algunas de las principales
rutas de transporte. Desde 2015, apostamos por combustibles
sostenibles y una parte importante de nuestra flota está preparada
para la utilización de este combustible en la actualidad. En este
proyecto piloto también colabora el Clúster de Energía de Aragón.
El Grupo avanza imparable en esta carrera por la descarbonización,
innovando en transporte y logística con el uso de biocombustibles,
la inauguración del Centro de Producción de Hidrógeno
reseñado y proyectos de eficiencia y seguridad al volante para
sus conductores. En esta estrategia de ser más sostenibles, todas
sus instalaciones han sido iluminadas con tecnología led, se han
dotado de cargadores de baterías electrónicos y se han optimizado
los movimientos en la logística interna de los almacenes. En
breve creará el segundo parque eólico y también a corto plazo
dotará de más tejados solares a sus instalaciones.
Todo ello sin olvidar la plantación de 200 hectáreas de almendros
y las 800 que se prevén de olivos en Aragón en los próximos
años a través de las sociedades agrarias del Grupo, con toda la
capacidad que tienen los ecosistemas forestales de disminuir el
efecto invernadero. Y es que la sostenibilidad y el cuidado del
medio ambiente guían a Carreras.


Miguel V. Carreras. De vocación, empresario151
CAPÍTULO X
Premios y reconocimientos
Nueve décadas y numerosas distinciones
En estos 90 años de recorrido, son numerosos los premios,
reconocimientos, distinciones y homenajes otorgados al Grupo
Carreras, tanto de forma individual (a su presidente y consejeros
delegados), como a las empresas que conforman este holding
familiar por su aportación al mundo de la economía y a la sociedad.
Miguel Carreras quiere comenzar este recorrido por los más emotivos,
los concedidos a su hijo José Luis Carreras Lario, consejero delegado
de Carreras Grupo Logístico hasta su fallecimiento en octubre de
2020, a los 58 años de edad a consecuencia del Coronavirus.
Homenajes
a Jose´ Luis Carreras Lario
• Enero 2021. El Centro Español de la Logística (CEL) homenajea a
José Luis Carreras, otorgándole un Premio a la Excelencia a título
póstumo e instaurando la primera edición de los premios que llevarán
su nombre.
• Septiembre 2021. Cámara Zaragoza inaugura una Sala en
honor de José Luis Carreras Lario. La Cámara de Comercio de
Zaragoza rinde un emotivo homenaje a José Luis Carreras Lario,
empresario y vocal del Pleno, por su implicación en el sector
empresarial aragonés y su generosa dedicación a la Cámara
hasta su fallecimiento. Manuel Teruel, presidente, y José Miguel
Sánchez, director general, junto al resto de los vocales aprobaron
Ana Rioja Jiménez152
la conversión de la que hasta entonces era conocida como Sala
Multiusos en Sala José Luis Carreras, en un acto muy emotivo que
contó con la presencia del vicepresidente del Gobierno de Aragón,
Arturo Aliaga, familiares y miembros del Pleno.
• Diciembre 2021. ADEA dedica un emotivo homenaje a José Luis
Carreras. La Asociación de Directivos y Ejecutivos de Aragón dedicó
un emotivo homenaje a José Luis Carreras, en el acto de entrega
de los Premios ADEA a los directivos de Aragón.
• Enero 2022. La Fundación Carreras crea el Premio José Luis Ca-
rreras Lario a la Innovación en Logística. Su objetivo es reconocer
y poner en valor la investigación. Tiene una periodicidad anual y ya
se han celebrado dos ediciones.
Premios
a Miguel Carreras Calvete
• Diciembre 2021. Medalla Basilio Paraíso 2021 a la Trayectoria
Empresarial. La Fundación Basilio Paraíso de la Cámara de Comercio
de Zaragoza, concede la Medalla a la Trayectoria Empresarial al
presidente del Grupo Carreras con motivo de todos sus logros
empresariales.
• 2018. Premio Aster ESIC a la Mejor Trayectoria Profesional y Caso
de Éxito Empresarial en 2018, concedido por ESIC Business and
Marketing School.
• 2012. Premio al Directivo de Aragón, que otorga la Asociación de
Directivos y Ejecutivos de Aragón (ADEA) y que reconoce su trayectoria
profesional y la evolución del Grupo empresarial que preside.
• 2011. Premio Actualidad Económica a la mejor trayectoria
empresarial de Aragón, que concede la revista Actualidad Económica.
• 1981. Premio Vendor. Otorgado por el Club de Marketing de Zaragoza,
en la categoría de Director de Empresa.
Miguel V. Carreras. De vocación, empresario153
La Cámara de Comercio de Zaragoza rindió un emotivo homenaje
a José Luis Carreras e inauguró una Sala con su nombre.
La familia de José Luis Carreras, en el homenaje realizado
por la Cámara de Comercio de Zaragoza.
Ana Rioja Jiménez154
Galardones
al Grupo Carreras
• Mayo 2023. ASTIC reconoce a Carreras Grupo Logístico por su
90 Aniversario. ASTIC (Asociación del Transporte Internacional por
Carretera).
• Mayo 2023. Carreras recibe el premio Computing a la Experiencia
Digital del Usuario en el marco de la IX edición de los Premios de
la Era Digital’ que organiza Computing para reconocer los proyectos
más innovadores que han tenido lugar durante el último año.
• Enero 2023. Eizasa Hoteles recibe el premio CaixaBank Hotels &
Tourism, al compromiso con la sostenibilidad de la Territorial Ebro.
• Noviembre 2022. Carreras consigue la Estrella Lean & Green por su
compromiso con la sostenibilidad, dentro de su Plan de Acción para
la reducción de la huella de carbono de sus actividades logísticas.
• Mayo 2022. Cadena de Suministro concede a Fernando Carreras
uno de los Premios 10 a la Excelencia en Logística y Transporte.
• Febrero 2021. Grupo Carreras recibe el Premio Ecosistemas de
Ibercaja y Grupo Henneo. «Un reconocimiento importante y emotivo
que comparte con todo el sector del transporte y la logística, que
ha demostrado su madurez, responsabilidad y profesionalidad en
palabras de Fernando Carreras», consejero delegado.
• 2021. Premios al Grupo Logístico Carreras por su colaboración
durante la pandemia. Carreras Grupo Logístico fue premiado con el
galardón «Empresario de Aragón», por diversas acciones solidarias
durante la pandemia, durante la V Cumbre Empresarial por la
Competitividad de Aragón, organizada por El Consejo Empresarial de
CEOE Aragón, con el patrocinio de Ibercaja. El acto, fue presidido por
S.M el Rey Felipe VI.
• Placa al Mérito de Protección Civil 2020 al Grupo Logístico Carreras
por su labor para combatir el coronavirus y su compromiso con la
sociedad aragonesa durante la pandemia.
• Noviembre 2019. Fernando Carreras recibe el Premio ADEA al
Desarrollo Logístico.
• Enero 2019. El Hotel Golf & Spa Real Badaguás, del Grupo Eizasa,
recibe el premio Hotels & Tourism de Caixabank en Aragón, en la
categoría de Mejor Establecimiento Hotelero e Innovación Turística.
Miguel V. Carreras. De vocación, empresario155
El Rey Don Felipe VI entregó a Fernando Carreras el galardón
por las acciones solidarias del Grupo Carreras durante la pandemia.
Miguel Carreras recibió en 2021 la Medalla Basilio Paraíso
a la Trayectoria Empresarial, que concede la Cámara de Comercio de Zaragoza.
Ana Rioja Jiménez156
• Mayo 2016. Fernando Carreras, Dirigente del Año de la Logística en
la XXIII edición de los Premios Dirigentes del Año, integrado por los
miembros del Club Dirigentes de la Industria y la Distribución.
• Mayo 2016. El Periódico de Aragón otorga al Grupo Carreras el
Premio Pujanza Empresarial 2016 en el marco de la celebración de
Los Aragoneses del Año, premios que eligen los lectores de este diario.
• Junio 2015. Mutua Fraternidad-Muprespa premia a Carreras
por su baja siniestralidad. Carreras Grupo Logístico ha recibido el
reconocimiento BONUS 2012, otorgado por la mutua Fraternidad-
Muprespa, por su esfuerzo en la labor y promoción de la realización
de políticas activas en materia de accidentes de trabajo.
• Julio 2014. Premio IFOY2014. Carreras y Toyota reciben el Premio
«International Forklift Truck of the Year» por la integración en los
sistemas del operador logístico del programa Toyota I_Site, lo que ha
permitido reducir los accidentes y daños de sus carretillas elevadoras
Toyota, disminuir los costes y aumentar tanto la seguridad como la
productividad.
• Abril 2014. Carreras gana el Randstad Award 2014 en la categoría
de Logística, como empresa más atractiva para trabajar en el sector
de la logística.
• Junio 2012. Carreras Grupo Logístico recibe el Premio ICIL a la
excelencia logística. Obtiene el galardón por haber desarrollado e
implantado un sistema de previsión de la demanda.
• Septiembre 2011. La Universidad de Zaragoza crea la Cátedra
Carreras. La Universidad de Zaragoza y Carreras crearon la Cátedra
Carreras de Sostenibilidad e Innovación Logística, enfocada a
contribuir en el ámbito de la Cadena de Suministro, la innovación y
logística sostenible, potenciando la investigación y la difusión del
conocimiento en estas áreas.
• Septiembre 2010. El Grupo Eizasa, premio al Mérito Turístico 2010,
concedido por el Departamento de Industria, Comercio y Turismo del
Gobierno de Aragón.
• 2009. Premio Europa del Transporte. Categoría Logística. Carreras
recibió el Premio Europa del Transporte en la categoría de Logística,
otorgado por la Confederación Española de Transporte de Mercancías
(CETM) y la revista Transporte Profesional, por su crecimiento y excelente
gestión en las áreas de transporte y logística.
Miguel V. Carreras. De vocación, empresario157
Cadena de Suministros concedió a Fernando Carreras uno de los Premios 10
a la Excelencia en mayo 2023.
Carreras recibió el premio Computing a la Experiencia Digital del Usuario
en Logística y Transporte en mayo 2023. En el centro, Ricardo Carreras.
• 2008. Reconocimiento de Expo Zaragoza 2008 a Transportes
Carreras por el apoyo como colaborador oficial de la Exposición
Internacional 2008.
• 2006. Premio Feria de Zaragoza, en la Feria Internacional de
Logísticas Especializadas.
• 2000. Premio Viñas del Vero.
• 1983. Premio ENASA. Galardón por el 50 Aniversario de Carreras.
Entra otros premios, galardones y reconocimientos.
Ana Rioja Jiménez158
Certificaciones
de Carreras Grupo Logístico
Además, Carreras Grupo Logístico disfruta de la certificación de
su sistema de gestión de calidad en base a la norma UNE-EN-ISO
9001:2015. La aplicación de herramientas de mejora continua ha
permitido que su sistema de gestión haya ido evolucionado hasta
convertirse en un sistema de gestión integrado que incluye:
• Primera empresa en Aragón en recibir el Certificado del Sello de
Excelencia Empresarial, en 2014.
• Calidad. Certificado AENOR UNE-EN-ISO 9001:2015. Sistema de
Gestión de la Calidad.
• Prevención. Certificado ISO 45001:2018. Sistemas de Gestión de la
Seguridad y la Salud en el Trabajo.
• Gestión medioambiental. Estrella Lean & Green por su compromiso
con la sostenibilidad, Certificado ISO 14001:2015. Sistema de Gestión
Medioambiental.
• Seguridad de la Información. ISO/IEC 27001:2013. Sistema de
Gestión de la Seguridad de la Información. Fue la primera empresa
española en obtenerlo.
• Procedimientos HACCP. Sistema de Análisis de Puntos Críticos.
• Certificado DUNS. Certificado de solvencia empresarial.
• Desde octubre de 2022, Carreras Grupo Logístico es miembro
de la Red Aragonesa de Empresas Saludables.
• Febrero 2023. Carreras renueva su sello RSA+Aragón Empresa,
gracias a su compromiso social y medio ambiental. El Sello RSA+
impulsa la Responsabilidad Social Corporativa.
• Licencias y Registros Sanitarios del Ministerio de Industria.
Certificado de manipulación.
• Seguridad en la Cadena de Suministros.

Miguel V. Carreras. De vocación, empresario159
EPÍLOGO
Lecciones de vida
Finales de julio de 2023. Nuestras conversaciones para escribir
su biografía, una larga entrevista, han llegado a su fin. Salgo del
despacho de Miguel Carreras con la ilusión de tener un buen material
para su biografía, de haber conocido a un gran hombre, que nació
para ser empresario, con una vocación y una intuición fuera de lo
común, que él ha acompañado de honestidad, sencillez, austeridad,
valentía, esfuerzo y otros dones que le ha dado la vida, mucho de ellos
heredados de su padre, de su familia, forjados en este Aragón que
él tanto ama. Algunos dicen que es un visionario, otros lo tildan de
pionero, pero ante todo es un hombre apasionado por la vida y por
la empresa, y esa pasión le ha hecho sortear todos los obstáculos para
llegar a lo más alto en casi todos los proyectos que ha emprendido, le
ha dado una vida plena.
Antes de irme, le pregunto cuál es la empresa del Grupo Carreras
de la que se siente más satisfecho, a la que está más unido. Como
un padre, al que se le interroga por sus hijos, no puede elegir. «Las
quiero a todas, porque mi trabajo ha sido idear proyectos, estudiarlos,
iniciarlos, gestionarlos y alcanzarlos. Esa planificación y desarrollo es
lo que he perseguido toda mi vida. Hacer las cosas bien suponía que
obtenía una rentabilidad, un beneficio, y así cada proyecto me daba
nuevas alas para el siguiente. Ha sido una carrera continua. Nunca he
parado, por eso no me he jubilado».
Nunca ha trabajado por dinero, su satisfacción no viene de la
cuenta de resultados, tan solo ha sido un medio para llevar a buen
puerto sus empresas. Y me vuelve a hablar de su equipo, del factor
Ana Rioja Jiménez160
humano. «Lo más importante de una empresa es su personal y me he
rodeado de los mejores y he sabido cuidarlos y quererlos». Nunca ha
buscado una recompensa económica, sino una carrera de proyectos,
con ansiedad y ambición sana, de hacer cada vez cosas más difíciles
y estar preparado para hacerlas bien. «Un proyecto me ha llevado a
otro y me ha preparado para el siguiente, que era mayor y requería
más esfuerzo. Y, sin darte cuenta, te realizas y vas dejando proyectos
hechos, diversificas y afianzas empresas, las haces más grandes, con
más fortaleza, con menos riesgo, y con más posibilidades de financiar
nuevos proyectos».
De Carreras Grupo Logístico afirma que «es la empresa más
antigua y la que mayor desarrollo ha tenido. Empezar de ayudante
de conductor con dos camiones viejos y terminar con centenares de
camiones y siendo líderes en logística de alimentación en España y
Portugal, e intentar hacer lo mismo en Francia con proyectos que han
sido pioneros, es muy satisfactorio. Es donde mayores han sido los
logros y donde más nos hemos destacado de la competencia. De no
ser nadie a estar moviendo 1.300 tráilers diarios, con fuerza y acierto.
Es como salir el último en la carrera y llegar de los primeros. Es un
gran orgullo».
Pero construir cerca de 4.000 viviendas con Eizasa ha sido una gran
satisfacción. «Comprar una vivienda es la operación más importante
que hace una persona o una pareja en su vida. Y pensar que yo, sin
haber sido profesional de la inmobiliaria en un principio, haya hecho
de unos solares o campos hermosas viviendas para que la gente viva,
sea feliz y forme una familia en ellas, se desarrollen y tengan años de
felicidad, es maravilloso. Y los polígonos industriales, que eran campos
agrícolas, y ver cómo fueron anidando en ellos empresas productivas...
Los hoteles también han sido una aventura apasionante, porque que
elijan un hotel, que has levantado y preparado tú, para pasar unos
días de descanso, y se encuentren a gusto, es una gran alegría. Es dar
algo tuyo a los demás».
Miguel V. Carreras. De vocación, empresario161
Retrato de Miguel Carreras,
realizado por la pintora Isabel Guerra. Fotografía: Columna Villarroya.
Ana Rioja Jiménez162
Como chico de pueblo que fue, ama el campo, y la agricultura
fue su primer amor, la madre naturaleza. «Sembrar la tierra, regarla,
plantar almendros, olivos en campos donde no había nada y, sobre
todo, con agua, es un placer. Es un trabajo muy bonito y muy noble.
Es un riesgo, porque los campos están expuestos a la voluntad de la
naturaleza, que no cubren los seguros, hay que arriesgar esperando que
no los castiguen mucho las heladas y la sequía. Recoger las cosechas
que nos da la tierra, ver el desarrollo de la planta en sus distintas
etapas es casi un milagro».
Salgo de las oficinas de Eizasa con sentimientos encontrados. La
chica de la recepción ya sabe quién soy y me llama por mi nombre.
Me despido, pero no le digo que ya hemos terminado las entrevistas
para escribir este libro, quizá porque me resisto a que hayan concluido
nuestras reuniones, que han sido verdaderas lecciones de vida, y
porque Miguel Carreras se ha ganado mi admiración y amistad.

Este libro se terminó de preparar en Ediciones·94,
en agosto de 2023,
con motivo de la celebración del
90 Aniversario del GRUPO CARRERAS
el 20 de septiembre de 2023.
LAUS DEO

"""
```
