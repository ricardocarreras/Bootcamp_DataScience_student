```python
import pandas as pd
import os
import zipfile
  
with zipfile.ZipFile("data.zip") as z: 
   
  with z.open("tracks_norm.csv") as a:

    datos = pd.read_csv(a, sep=";") 

  with z.open("albums_norm.csv") as b: 

    datos2=pd.read_csv(b, sep=";") 

  with z.open("artists_norm.csv") as c:
              
    datos3=pd.read_csv(c, sep=";")

```
