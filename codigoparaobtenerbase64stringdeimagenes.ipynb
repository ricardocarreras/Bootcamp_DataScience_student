```python
import base64

# Replace 'path_to_your_image.png' with the path to your image file.
with open('/content/selectron3.png', 'rb') as image_file:
    base64_data = base64.b64encode(image_file.read()).decode('utf-8')

# Now, base64_data contains the Base64 string of your image.
```
