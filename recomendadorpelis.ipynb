```python
import pandas as pd   # Tratamiento dataframes
import numpy as np # Mates
import matplotlib.pyplot as plt #Gráficas
import seaborn as sns #Gráficas aún más chulas
import scipy # Estadísticas, ciencia
import os

nombre_fichero = "/content/Peliculas-220318-180251 (1).csv"

pelis=pd.read_csv(nombre_fichero, sep=",")


pelis.shape
pelis.head(11)
```

```python
pelis["movieId"]=pd.to_numeric(pelis["movieId"])
```

```python
nombre_fichero2 = "/content/Clasificacion-220318-180251 (1).csv"

clasi=pd.read_csv(nombre_fichero2)

clasi.head()

clasi.shape
```

```python
clasi.dtypes
```

```python

datosjuntos=pd.merge(pelis, clasi)

datosjuntos.shape
```

```python
datosjuntos2=datosjuntos.pivot_table(index=["userId"], columns=["title"], values="rating")

datosjuntos2.head()
```

```python
clasificacion_toystory=datosjuntos2["Toy Story (1995)"]
```

```python
datosjuntos2["Toy Story (1995)"]
```

```python
similarpeliculas=datosjuntos2.corrwith(clasificacion_toystory)
```

```python
similarpeliculas.head()
```

```python
corr_similarpeliculas=pd.DataFrame(similarpeliculas, columns=["Correlaciones"])

corr_similarpeliculas=corr_similarpeliculas.dropna()

corr_similarpeliculas.sort_values("Correlaciones", ascending=False)

corr_similarpeliculas_=corr_similarpeliculas.sort_values("Correlaciones", ascending=False).reset_index()

corr_similarpeliculas_
```

```python
listaa=datosjuntos.genres.unique()
lista=[]
for i in range(len(listaa)):
    lista=lista+listaa[i].split("|")
```

```python
lista=list(set(lista))

for i in lista:
    datosjuntos[i]=datosjuntos.genres.str.contains(i)
```

```python
datosjuntos.tail()
```

```python
datosjuntos_aux=datosjuntos.drop_duplicates(subset="title")

datosjuntos_aux.head()
```

```python
df=corr_similarpeliculas_.merge(datosjuntos_aux, how="left")
df=df.drop(columns=["movieId", "genres", "userId", "rating", "timestamp"])

df.head()
```
