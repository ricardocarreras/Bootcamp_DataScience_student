```python
import requests
headers = {
    "Content-type" : "application/json",
    "Authorization" : "TOKEN-KEY"
}
url_anella = "https://opendata-ajuntament.barcelona.cat/data/dataset/50f2bc1f-7066-465a-a530-1e59c4ac17a4/resource/d330f548-c21a-4186-a650-ab7cc508b152/download"
response = requests.get(url_anella, headers=headers)
```

```python
response.status_code
```

```python
response.content
```

```python
import json
data = json.loads(response.content)
data
```

```python
type(data)
```

```python
data2 = data['actes']
for tmp in data2['acte']:
  if tmp['date'][0:4] == '2022' and tmp['acte_type'] == 'Concierto':
    print(tmp['name'])
```
