```python
import warnings
warnings.simplefilter('ignore')
```

```python
def parse_response(response):
  data = None
  if response.status_code == 200:
    # Los datos están formateados en str
    data = json.loads(response.content)
  else:
    raise Exception(f"Error en la respuesta {response.status_code}, {response.reason}")
  return data
```

```python
def get_image_nasa(endpoint, api_key):
  # Realizamos la petición GET
  base_url = endpoint + "api_key=" + api_key

  # Lanzamos la petición GET
  response = requests.get(base_url)

  # Comprobamos el response
  data = parse_response(response)

  return data
```

```python
endpoint = "https://api.nasa.gov/planetary/apod?"
api_key = "DEMO_KEY"
```

```python
data = get_image_nasa(endpoint, api_key)
```

```python
# Probamos que el recurso solicitado es una imagen Alternativa A
if data['media_type'] == 'image':
  tmp = requests.get(data.get('url'))
  img = Image.open(BytesIO(tmp.content))
  plt.imshow(img)
  plt.show()
else:
  print(f"El media-type es {data['media_type']}")
```
