```python
# Importamos las librerías
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
```

```python
# Vinculamos las credenciales del bigquery
creds = service_account.Credentials.from_service_account_file('idbootcamps-project-2022-3f1e7cbf3799.json')
```

```python
# Realizamos la conexión con el cliente bigquery
client = bigquery.Client(credentials=creds)
```

```python
# Creamos la primera query ESTADO
query_1 = """
SELECT
    state as estado
FROM `bigquery-public-data.samples.natality`
WHERE
    state IS NOT NULL
GROUP BY
    estado
ORDER BY
    estado

"""
```

```python
# Lanzamos el job
query_job = client.query(query_1)
df = query_job.to_dataframe()
df
```
